## Cel zajęć 

Celem niniejszych zajęć jest nauczenie się wykorzystywania **kryterium pierwiastkowego Cauchy'ego** i **kryterium ilorazowego d'Alemberta** zbieżności szeregów do badania zbieżności albo rozbieżności szeregów liczbowych o wyrazach nieujemnych oraz **kryterium Leibniza** do badania zbieżności szeregów naprzemiennych.


W wielu zastosowaniach potrzebna jest wiedza o tym, czy dany szereg liczbowy jest zbieżny. Służą do tego specjalne narzędzia - twierdzenia podające kryteria na zbieżność szeregów liczbowych.

## Kryterium pierwiastkowe Cauchy'ego

Rozpocznijmy od przypomnienia klasycznej wersji kryterium pierwiastkowego Cauchy'ego sformułowanego w oparciu o teorię granic.

>**Twierdzenie** (Kryterium pierwiastkowe Cauchy'ego w postaci limesowej).
>
>Niech $a_{n}\geq 0$ dla wszystkich $n\in\BN$ i niech ciąg $(\sqrt[n]{a_{n}})$ ma granicę g. Jeżeli $g<1,$ to szereg $\underset{n=1}{\overset{\infty}{\sum}} a_{n}$ jest zbieżny, a jeżeli $g>1$, to szereg $\underset{n=1}{\overset{\infty}{\sum}} a_{n}$ jest rozbieżny.

### Uwaga: 
Kryterium nie rozstrzyga o zbieżności szeregu, w przypadku, gdy granica $g=1$. Należy wówczas użyć innego kryterium zbieżności szeregów.

### Przykład
Zbadamy zbieżność szeregu: 

$$
\displaystyle\sum_{n=1}^\infty\frac{n^2}{(2+1 / n)^n}.
$$

Mamy: $a_n=\frac{n^2}{(2+1/ n )^n}>0$ dla każdego $n\in\BN$ oraz 

\begin{eqnarray*}
\lim\limits_{n\tow\infty} \sqrt[n]{a_n}&=& \lim\limits_{n\tow\infty}  \sqrt[n]{\frac{n^2}{(2+1 / n )^n}}=\lim\limits_{n\tow\infty} \frac{\sqrt[n]{n^2}}{\sqrt[n]{(2+1 / n )^n}}=\lim\limits_{n\tow\infty} \frac{\sqrt[n]{n}\cdot \sqrt[n]{n}}{2+\frac{1}{n} }=\frac{1}{2}<1.
\end{eqnarray*}

Na mocy kryterium pierwiastkowego Cauchy'ego szereg $\displaystyle\sum_{n=1}^\infty\frac{n^2}{(2+1 / n )^n}$ jest zbieżny.


### Zadanie 1. 
Korzystając z kryterium Cauchy'ego, zbadaj zbieżność następujących szeregów:

a) $\displaystyle\sum_{n=1}^\infty (\operatorname{arctg}(n^2+1))^n$

$\color{green}{Rozwiązanie:}$

Mamy: $a_n=(\operatorname{arctg}(n^2+1))^n>0$ dla każdego $n\in\BN$ oraz 

\begin{eqnarray*}
\lim\limits_{n\tow\infty} \sqrt[n]{a_n}&=& \lim\limits_{n\tow\infty}  \sqrt[n]{(\operatorname{arctg}(n^2+1))^n }=\lim\limits_{n\tow\infty}\operatorname{arctg}(n^2+1)=\frac{\pi}{2}>1 \hspace{0.1cm}(\pi\approx 3.14).
\end{eqnarray*}

Na mocy kryterium pierwiastkowego Cauchy'ego szereg $\displaystyle\sum_{n=1}^\infty (\operatorname{arctg}(n^2+1))^n$ jest rozbieżny.

$$
\color{blue}{\text{Odp. Szereg rozbieżny.}}
$$

b) $\displaystyle\sum_{n=1}^\infty\left(\frac{n}{2n+1}\right)^n$

$\color{green}{Rozwiązanie:}$

Mamy: $a_n= \left(\frac{n}{2n+1}\right)^n>0$ dla każdego $n\in\BN$ oraz 

\begin{eqnarray*}
\lim\limits_{n\tow\infty} \sqrt[n]{a_n}&=& \lim\limits_{n\tow\infty}  \sqrt[n]{\left(\frac{n}{2n+1}\right)^n}=\lim\limits_{n\tow\infty} \frac{n}{2n+1}=\frac{1}{2}<1.
\end{eqnarray*}

Na mocy kryterium pierwiastkowego Cauchy'ego szereg $\displaystyle\sum_{n=1}^\infty\left(\frac{n}{2n+1}\right)^n$ jest zbieżny.

$$
\color{blue}{\text{Odp. Szereg zbieżny.}}
$$

c) $\displaystyle\sum_{n=1}^\infty\frac{(\frac{n+1}{n})^n}{3^n}$

$\color{green}{Rozwiązanie:}$

Mamy: $a_n= \frac{(\frac{n+1}{n})^n}{3^n}>0$ dla każdego $n\in\BN$ oraz 

\begin{eqnarray*}
\lim\limits_{n\tow\infty} \sqrt[n]{a_n}&=& \lim\limits_{n\tow\infty}  \sqrt[n]{\frac{(\frac{n+1}{n})^n}{3^n}}=\lim\limits_{n\tow\infty} \frac{\frac{n+1}{n}}{3}=\lim\limits_{n\tow\infty} \frac{1+\frac{1}{n}}{3}=\frac{1}{3} <1.
\end{eqnarray*}

Na mocy kryterium pierwiastkowego Cauchy'ego szereg $\displaystyle\sum_{n=1}^\infty\frac{(\frac{n+1}{n})^n}{3^n}$ jest zbieżny.

$$
\color{blue}{\text{Odp. Szereg zbieżny.}}
$$

d) $\displaystyle\sum_{n=1}^\infty\left(\frac{n}{n+1}\right)^{n^2}2^n$

$\color{green}{Rozwiązanie:}$

Mamy: $a_n= \left(\frac{n}{n+1}\right)^{n^2}\cdot 2^n>0$ dla każdego $n\in\BN$ oraz 

\begin{eqnarray*}
\lim\limits_{n\tow\infty} \sqrt[n]{a_n}&=& \lim\limits_{n\tow\infty}  \sqrt[n]{\left(\frac{n}{n+1}\right)^{n^2}\cdot 2^n}=\lim\limits_{n\tow\infty} 2\left(\frac{n}{n+1}\right)^{n}= \lim\limits_{n\tow\infty} 2\left(\frac{n+1}{n}\right)^{-n}\\
&=& \lim\limits_{n\tow\infty} 2\left(1+\frac{1}{n}\right)^{-n}=\frac{2}{e}<1.
\end{eqnarray*}

Na mocy kryterium pierwiastkowego Cauchy'ego szereg $\displaystyle\sum_{n=1}^\infty\left(\frac{n}{n+1}\right)^{n^2}2^n$  jest zbieżny.

$$
\color{blue}{\text{Odp. Szereg zbieżny.}}
$$

e) $\displaystyle\sum_{n=1}^\infty\frac{(n!)^n}{n^{n^2}}$

$\color{green}{Rozwiązanie:}$

Mamy: $a_n=\frac{(n!)^n}{n^{n^2}}>0$ dla każdego $n\in\BN$ oraz 

\begin{eqnarray*}
\lim\limits_{n\tow\infty} \sqrt[n]{a_n}&=& \lim\limits_{n\tow\infty} \sqrt[n]{\frac{(n!)^n}{n^{n^2}}}=\lim\limits_{n\tow\infty} \frac{n!}{n^{n}}
\end{eqnarray*}

Zastosujemy teraz 

>**Twierdzenie (*).** Jeżeli $\lim\limits_{n\rightarrow\infty}\vert\frac{a_{n+1}}{a_n}\vert=q<1$ i $a_n\neq 0$ dla  każdego $n\in\BN$, to $\lim\limits_{n\rightarrow\infty}a_n=0$.

do ciągu o wyrazie ogólnym $b_n=\frac{n!}{n^{n}}$, $n\in\BN$. Mamy  $b_n=\frac{n!}{n^{n}}>0$ dla każdego $n\in\BN$ oraz

\begin{eqnarray*}
\lim\limits_{n\tow\infty} \left|\frac{b_{n+1}}{b_n}\right|=\lim\limits_{n\tow\infty} \left|\frac{\frac{(n+1)!}{(n+1)^{n+1}}}{\frac{n!}{n^{n}}}\right|=\lim\limits_{n\tow\infty} \frac{n!(n+1)}{(n+1)^{n}(n+1)}\cdot \frac{n^n}{n!}=\lim\limits_{n\tow\infty} \left(\frac{n}{n+1}\right)^n=e^{-1}=\frac{1}{e}<1,
\end{eqnarray*}

więc na mocy Twierdzenia (*) otrzymujemy: 

$$
\lim\limits_{n\tow\infty} \frac{n!}{n^{n}}=0 \text{ (i jest }<1).
$$

Na mocy kryterium pierwiastkowego Cauchy'ego szereg $\displaystyle\sum_{n=1}^\infty\frac{(n!)^n}{n^{n^2}}$  jest zbieżny.

$$
\color{blue}{\text{Odp. Szereg zbieżny.}}
$$

## Kryterium ilorazowe d'Alemberta

>**Twierdzenie** (Kryterium ilorazowe d'Alemberta w postaci limesowej).
>
>Niech $a_{n}>0$ i niech ciąg $(\frac{a_{n+1}}{a_{n}})$ ma granicę g. Jeżeli $g<1$, to $\underset{n=1}{\overset{\infty}{\sum}} a_{n}<\infty$, a jeżeli $g>1$, to $\underset{n=1}{\overset{\infty}{\sum}} a_{n}=\infty$.

### Uwagi:

1. Kryterium nie rozstrzyga o zbieżności szeregu, w przypadku, gdy granica $g=1$. Należy wówczas użyć innego kryterium zbieżności szeregów.

2. Warto pamiętać, że $\lim\limits_{n\rightarrow\infty}\sqrt[n]{n!}=\infty$, tzn. ciąg $(\sqrt[n]{n!})$ jest rozbieżny.

### Przykład 

Zbadamy zbieżność następującego szeregu:

$$
\displaystyle\sum_{n=1}^\infty\frac{10^n}{n!}.
$$

Mamy: $a_n=\frac{10^n}{n!}>0$ dla każdego $n\in\BN$, 

$$
a_{n+1}=\frac{10^{n+1}}{(n+1)!}
$$ 

oraz

$$
\lim\limits_{n\tow\infty}\frac{a_{n+1}}{a_{n}}=\lim\limits_{n\tow\infty}\frac{\frac{10^{n+1}}{(n+1)!}}{\frac{10^n}{n!}}=\lim\limits_{n\tow\infty} \frac{10^{n+1}}{(n+1)!}\cdot \frac{n!}{10^n}=\lim\limits_{n\tow\infty} \frac{10^n\cdot 10}{n!\cdot (n+1)}\cdot \frac{n!}{10^n}= \lim\limits_{n\tow\infty} \frac{10}{n+1}=0<1,        
$$


więc na mocy kryterium d'Alemberta szereg $\displaystyle\sum_{n=1}^\infty\frac{10^n}{n!}$ jest zbieżny.

### Zadanie 2. 
Korzystając z kryterium d'Alemberta, zbadaj zbieżność następujących szeregów:

a) $\displaystyle\sum_{n=1}^\infty\frac{n!}{n^n}$

$\color{green}{Rozwiązanie:}$

Mamy: $a_n=\frac{n!}{n^n}>0$ dla każdego $n\in\BN$, 

$$
a_{n+1}=\frac{(n+1)!}{(n+1)^{n+1}}
$$

oraz

\begin{eqnarray*}
\lim\limits_{n\tow\infty}\frac{a_{n+1}}{a_{n}}&=&\lim\limits_{n\tow\infty} \frac{\frac{(n+1)!}{(n+1)^{n+1}}}{{\frac{n!}{n^n}}}= \lim\limits_{n\tow\infty}\frac{(n+1)!}{(n+1)^{n+1}}\cdot \frac{n^n}{n!}= \lim\limits_{n\tow\infty}\frac{n!\cdot (n+1)}{(n+1)^n\cdot (n+1)}\cdot \frac{n^n}{n!}\\
&=& (\text{ po redukcji wyrażeń podobnych })
=\lim\limits_{n\tow\infty}\frac{n^n}{(n+1)^n}=\lim\limits_{n\tow\infty}\left(\frac{n}{n+1}\right)^n=\frac{1}{e}.
\end{eqnarray*}

 

Ponieważ 

$$
\lim\limits_{n\tow\infty} \frac{\frac{(n+1)!}{(n+1)^{n+1}}}{{\frac{n!}{n^n}}}=\frac{1}{e}<1,
$$

na mocy kryterium d'Alemberta szereg $\displaystyle\sum_{n=1}^\infty\frac{n!}{n^n}$ jest zbieżny.

$$
\color{blue}{\text{Odp. Szereg zbieżny.}}
$$

b) $\displaystyle\sum_{n=1}^\infty\frac{(n!)^2}{2^{n^2}}$

$\color{green}{Rozwiązanie:}$

Mamy: $a_n=\frac{(n!)^2}{2^{n^2}}>0$ dla każdego $n\in\BN$, 

$$
a_{n+1}=\frac{((n+1)!)^2}{2^{(n+1)^2}}
$$

oraz 

\begin{eqnarray*}
\lim\limits_{n\tow\infty}\frac{a_{n+1}}{a_{n}}&=&\lim\limits_{n\tow\infty} \frac{\frac{((n+1)!)^2}{2^{(n+1)^2}}}{\frac{(n!)^2}{2^{n^2}}}=\lim\limits_{n\tow\infty} \frac{((n+1)!)^2}{2^{(n+1)^2}}\cdot \frac{2^{n^2}}{(n!)^2}=\lim\limits_{n\tow\infty} \frac{(n!)^2\cdot (n+1)^2}{2^{n^2+2n+1}}\cdot \frac{2^{n^2}}{(n!)^2}\\
&=& (\text{ po redukcji wyrażeń podobnych })
% &=&\lim\limits_{n\tow\infty} \frac{ \hbox{\sout{$(n!)^{2}$}}\cdot (n+1)^2}{2^{n^2+2n+1}}\cdot \frac{2^{n^2}}{\hbox{\sout{$(n!)^2$}}}
=\lim\limits_{n\tow\infty} \frac{ (n^2+2n+1)2^{n^2}}{2^{n^2}\cdot 2^{2n+1}}\\
&=& (\text{ po redukcji wyrażeń podobnych })
% &=&\lim\limits_{n\tow\infty} \frac{ (n^2+2n+1)\hbox{\sout{$2^{n^2}$}}}{\hbox{\sout{$2^{n^2}$}}\cdot 2^{2n+1}}
=\lim\limits_{n\tow\infty} \frac{ n^2+2n+1}{2\cdot 2^{2n}}
\end{eqnarray*}

Teraz stosujemy

>**Twierdzenie (*).** Jeżeli $\lim\limits_{n\rightarrow\infty}\vert\frac{a_{n+1}}{a_n}\vert=q<1$ i $a_n\neq 0$ dla  każdego $n\in\BN$, to $\lim\limits_{n\rightarrow\infty}a_n=0$.

do ciągu o wyrazie ogólnym $b_n=\frac{ n^2+2n+1}{2\cdot 2^{2n}}$. Ponieważ

\begin{eqnarray*}
\lim\limits_{n\tow\infty}\left|\frac{b_{n+1}}{b_{n}}\right|&=&\lim\limits_{n\tow\infty} \frac{\frac{ (n+1)^2+2(n+1)+1}{2\cdot 2^{2(n+1)}}}{\frac{ n^2+2n+1}{2\cdot 2^{2n}}}=\lim\limits_{n\tow\infty}\frac{n^2+4n+4}{8\cdot 2^{2n}}\cdot \frac{2\cdot 2^{2n}}{n^2+2n+1}\\
&=&\lim\limits_{n\tow\infty}\frac{n^2+4n+4}{4\cdot (n^2+2n+1)}=\frac{1}{4}<1,
\end{eqnarray*}

na mocy Twierdzenia (*), dostajemy

$$
\lim\limits_{n\tow\infty}  \frac{ n^2+2n+1}{2\cdot 2^{2n}}=0.
$$

Z uwagi na fakt, iż

$$\lim\limits_{n\tow\infty}  \frac{ n^2+2n+1}{2\cdot 2^{2n}}=0<1,
$$

na mocy kryterium d'Alemberta szereg $\displaystyle\sum_{n=1}^\infty\frac{(n!)^2}{2^{n^2}}$  jest zbieżny.

$$
\color{blue}{\text{Odp. Szereg zbieżny.}}
$$

c) $\displaystyle\sum_{n=1}^\infty\frac{n^5}{2^n+3^n}$

$\color{green}{Rozwiązanie:}$

Oczywiście $a_n=\frac{n^5}{2^n+3^n}>0$ dla każdego $n\in\BN$. Mamy:

\begin{eqnarray*}
\lim\limits_{n\tow\infty}\frac{a_{n+1}}{a_{n}}&=&\lim\limits_{n\tow\infty} \frac{\frac{(n+1)^5}{2^{n+1}+3^{n+1}}}{\frac{n^5}{2^n+3^n}}=\lim\limits_{n\tow\infty}\frac{(n+1)^5}{2^{n+1}+3^{n+1}}\cdot \frac{2^n+3^n}{n^5}=\lim\limits_{n\tow\infty} \frac{(n+1)^5}{n^5}\cdot \frac{2^n+3^n}{2\cdot 2^{n}+3\cdot 3^{n}}\\
&=& \lim\limits_{n\tow\infty} \left(1+\frac{1}{n}\right)^5\cdot \frac{\left(\frac{2}{3}\right)^n+1}{2\cdot \left(\frac{2}{3}\right)^{n}+3}=\frac{1}{3}<1.
\end{eqnarray*}

Na mocy kryterium d'Alemberta szereg $\displaystyle\sum_{n=1}^\infty\frac{n^5}{2^n+3^n}$ jest zbieżny.

$$
\color{blue}{\text{Odp. Szereg zbieżny.}}
$$

d) $\displaystyle\sum_{n=1}^\infty\frac{(n!)^2\cdot 5^n}{(2n)!}$.

$\color{green}{Rozwiązanie:}$

Mamy: $a_n=\frac{(n!)^2\cdot 5^n}{(2n)!}>0$ dla każdego $n\in\BN$, 

$$
a_{n+1}=\frac{((n+1)!)^2\cdot 5^{n+1}}{(2(n+1))!}=\frac{(n!)^2\cdot (n+1)^2\cdot 5\cdot 5^{n}}{(2n)!\cdot(2n+1)\cdot (2n+2)}\;\;\text{ dla każdego }n\in\BN
$$

oraz 

$$
\lim\limits_{n\tow\infty}\frac{a_{n+1}}{a_{n}} = \lim\limits_{n\tow\infty} \frac{\frac{(n!)^2\cdot (n+1)^2\cdot 5\cdot 5^{n}}{(2n)!\cdot(2n+1)\cdot (2n+2)}}{\frac{(n!)^2\cdot 5^n}{(2n)!}}=\lim\limits_{n\tow\infty}  \frac{5(n+1)^2}{(2n+1)\cdot 2(n+1)}=\lim\limits_{n\tow\infty}  \frac{5(n+1)}{2(2n+1)}=\frac{5}{4}>1.
$$

Na mocy kryterium d'Alemberta szereg $\displaystyle\sum_{n=1}^\infty\frac{(n!)^2\cdot 5^n}{(2n)!}$ jest rozbieżny.

$$
\color{blue}{\text{Odp. Szereg rozbieżny.}}
$$

## Kryterium Raabe'go

Jest ono mocniejsze od kryterium  Cauchy'ego, a więc również mocniejsze od
kryterium d'Alemberta.


>__Kryterium Raabe'go.__ (Patrz wykład) Niech dany będzie szereg $ \sum_{n=1}^{\infty}a_n$, którego wyrazy są dodatnie 
i niech spełniony będzie następujący warunek
>                                                                                     
>$$
\lim_{n\to\infty}n\left (\frac {a_n}{a_{n+1}}-1\right )=\gamma \ .
$$
>                                                                                     
>Jeżeli $\gamma>1$, to szereg $ \sum_{n=1}^{\infty}a_n$ jest zbieżny, 
jeżeli $\gamma <1$, to szereg $ \sum_{n=1}^{\infty}a_n$ jest rozbieżny, natomiast w przypadku 
gdy $\gamma=1$ kryterium nie daje rozstrzygnięcia.

### Przykład

Rozważmy szereg 

$$
\underset{n=1}{\overset{\infty}{\sum}}\frac{1}{n^\alpha}
$$

przy założeniu, że $\alpha\in \BN$. 

Zauważmy, że ani kryterium Cauchy'ego ani kryterium d'Alemberta nie rozstrzygają zaganienia zbieżności tego szeregu ponieważ, przyjmując 

$$
a_n=\frac{1}{n^\alpha},
$$

mamy $a_n>0$ dla wszystkich $n\in\BN$ oraz:

$$
\sqrt[n]{a_n}=\left(\frac{1}{\sqrt[n]{n}}\right)^\alpha\rightarrow 1
$$

i

$$
\frac{a_{n+1}}{a_n}=\left(\frac{n}{n+1}\right)^\alpha\rightarrow 1
$$

przy $n\rightarrow\infty$. W obu przypadkach dostajemy $g=1$, tzw. przypadek wątpliwy. 

Zastosujmy wobec tego kryterium Raabego. Dla $\alpha$ całkowitych dodatnich mamy

$$
n\left(\frac{a_n}{a_{n+1}}-1\right)=\frac{(n+1)^\alpha-n^\alpha}{n^{\alpha-1}}=\alpha+\left(\stackrel{\alpha}{2}\right)\frac{1}{n}+...+\frac{1}{n^{\alpha-1}}\rightarrow \alpha
$$

przy $n\rightarrow\infty$. 

Zatem:

1. dla $\alpha=2,3,...,$ szereg $\underset{n=1}{\overset{\infty}{\sum}}\frac{1}{n^\alpha}$ jest zbieżny na mocy kryterium Raabego ponieważ $\alpha>1$, 

2. a dla $\alpha=1$ otrzymujemy szereg harmoniczny $\underset{n=1}{\overset{\infty}{\sum}}\frac{1}{n}$, który, jak pamiętamy, jest rozbieżny. W tym przypadku kryterium Raabego nie rozstrzyga o zbieżności/rozbieżności szeregu.


## Kryterium Leibniza

Przejdźmy teraz do kryterium pozwalającego zbadać zbieżność szeregu naprzemiennego.

>**Twierdzenie** (Kryterium Leibniza).
>
>Jeżeli $a_{n}\geq 0$, $(a_{n})$ jest ciągiem nierosnącym i $a_{n}\rightarrow0$ przy $n\rightarrow\infty$, to szereg naprzemienny $\underset{n=1}{\overset{\infty}{\sum}} (-1)^{n+1}a_{n}$ jest zbieżny. 
>Co więcej, n-ta reszta tego szeregu $$r_{n}=\underset{k=n+1}{\overset{\infty}{\sum}} (-1)^{k+1}a_{k}$$ 
spełnia nierówność $$\left|r_{n}\right|\leq a_{n+1}.$$

### Uwagi:

1. Zauważmy, że $r_n$ oznacza różnicę pomiędzy sumą szeregu naprzemiennego $s$, a $s_n:=\underset{k=1}{\overset{n}{\sum}} (-1)^{k+1}a_{k}$.

2. Druga część kryterium Leibniza umożliwia **obliczenie sumy zbieżnego szeregu naprzemiennego z dowolną dokładnością**!

### Zadanie 3.

Korzystając z kryterium Leibniza, uzasadnić zbieżność szeregu:

a) $\displaystyle\sum_{n=1}^\infty(-1)^{n+1}\frac{1}{n^2}$

Mamy: 

$1^0$ $a_n=\frac{1}{n^2}>0$ dla każdego $n\in\BN$.

$2^0$ Zauważmy, że wraz ze wzrostem $n$ wyrażenie $n^2$ rośnie, czyli $\frac{1}{n^2}$ maleje, a zatem ciąg $(\frac{1}{n^2})$  jest malejący.

$3^0$ $\lim\limits_{n\rightarrow\infty} \frac{1}{n^2}=0$.

Na mocy kryterium Leibniza szereg $\displaystyle\sum_{n=1}^\infty(-1)^{n+1}\frac{1}{n^2}$ jest zbieżny.

$$
\color{blue}{\text{Odp. Szereg jest zbieżny.}}
$$

b) $\displaystyle\sum_{n=1}^\infty(-1)^{n+1}(\sqrt[n]{19}-1)$.

Mamy: 

$1^0$ $a_n=\sqrt[n]{19}-1>0$ dla każdego $n\in\BN$.

$2^0$ Sprawdzamy, czy ciąg $(a_n)$ jest malejący, tzn. czy $a_{n+1}< a_n$ dla dowolnego $n\in\BN$. Równoważnie: czy  $a_{n+1}-a_n<0$ dla dowolnego $n\in\BN$. Mamy:

$$
a_{n+1}-a_n=\sqrt[n+1]{19}-1 -( \sqrt[n]{19}-1)=\sqrt[n+1]{19}-\sqrt[n]{19}<0
$$

ponieważ warunek 

$$
\sqrt[n+1]{19}-\sqrt[n]{19}<0
$$

oznacza równoważnie, że 

$$
\sqrt[n+1]{19}<\sqrt[n]{19},
$$

czyli równoważnie,

$$
19<19^{\frac{n+1}{n}}=19^{1+\frac{1}{n}},\hspace{0.2cm} n\in\BN,
$$

a ostatni warunek jest prawdziwy.

$3^0$ $\lim\limits_{n\rightarrow\infty} (\sqrt[n]{19}-1)=\lim\limits_{n\rightarrow\infty} \sqrt[n]{19}-1=1-1=0$.

Na mocy kryterium Leibniza szereg $\displaystyle\sum_{n=1}^\infty(-1)^{n+1}(\sqrt[n]{19}-1)$ jest zbieżny.

$$
\color{blue}{\text{Odp. Szereg jest zbieżny.}}
$$

### Przykład
Obliczymy sumę szeregu  

$$
\displaystyle\sum_{n=1}^\infty(-1)^{n+1}\frac{1}{n^2}
$$

z dokładnością do $0.01$.

Wykorzystamy w tym celu kryterium Leibniza. Zgodnie z nim, resztę $r_n$ możemy oszacować w następujący sposób: $|r_n|\leq a_{n+1}$, czyli

$$
|r_n|\leq a_{n+1}=\frac{1}{(n+1)^2}.
$$

Skoro chcemy obliczyć sumę szeregu z dokładnością do $0.01$, to 

$$
\frac{1}{(n+1)^2}<0.01.
$$

Rozwiązując tę nierówność mamy:

$$
\frac{1}{n+1}<0.1=\frac{1}{10},
$$

czyli $n+1>10\Leftrightarrow n>9$. Zatem dla $n\geq 10, n\in\BN$ dostajemy, że

$$
{ \color{red}{\sum_{i=1}^\infty(-1)^{i+1}\frac{1}{i^2}}}=\sum_{i=1}^n(-1)^{i+1}\frac{1}{i^2}+r_n{\color{red}{\approx \sum_{i=1}^{10}(-1)^{i+1}\frac{1}{i^2}}} \text{ z błędem mniejszym niż }0.01.
$$

Liczymy przybliżając sumę $s$ danego szeregu zbieżnego sumą jedynie 10 pierwszych wyrazów powyższego szeregu:

$$
\sum_{i=1}^\infty(-1)^{i+1}\frac{1}{i^2}\approx \sum_{i=1}^{10}(-1)^{i+1}\frac{1}{i^2}\approx 0.8179621756.
$$

### Uwaga: 

Oczywiście można zsumować więcej niż 10 pierwszych wyrazów tego szeregu i tą wartością przybliżać dokładną sumę zadanego szeregu naprzemiennego. Dokładność obliczeniowa nie pogorszy się.

Poniżej skrypt w Pythonie obliczający przybliżoną sumę powyższego szeregu naprzemiennego: 

```python
from __future__ import division
import math
import numpy as np
from decimal import *

getcontext().prec = 28


def sz(n_0):
    s = float(0.0)
    m = int(math.floor(1 / np.sqrt(n_0)))
    for i in range(1, m + 1):
        s += (-1) ** (i + 1) * float(1 / i ** 2)
    return s


print('Przybliżona suma 10 pierwszych wyrazów szeregu naprzemiennego to:', sz(0.01))
print('Przybliżona suma 100 pierwszych wyrazów szeregu naprzemiennego to:', sz(0.0001))
print('Przybliżona suma 10000 pierwszych wyrazów szeregu naprzemiennego to:', sz(0.00000001))
```

A teraz inny sposób obliczenia przybliżonej sumy tego samego szeregu (skrypt Python) z dokładnością do 0.0001:

```python
def evaluate_sum():
    total_sum = 0
    n = 1
    while True:
        term = (-1) ** (n + 1) / (n ** 2)
        total_sum += term
        if abs(term) < 0.0001:
            break
        n += 1
    return total_sum

sum_result = evaluate_sum()
print(f"The sum of the series is approximately {sum_result:.4f}")
```

## Dodatek

W literaturze znane są ogólniejsze kryteria zbieżności szeregów o wyrazach nieujemnych. Poniżej zostały one przytoczone.

### Ogólniejsza postać kryterium Cauchy'ego

>**Twierdzenie** (Kryterium pierwiastkowe Cauchy'ego w języku granic częściowych). Jeżeli:
>
>(1) $\limsup\limits_{n\rightarrow\infty}\sqrt[n]{|a_n|}<1$, to szereg $\underset{n=1}{\overset{\infty}{\sum}} a_{n}$ jest bezwzględnie zbieżny;
>
>(2) $\limsup\limits_{n\rightarrow\infty}\sqrt[n]{|a_n|}>1$, to szereg $\underset{n=1}{\overset{\infty}{\sum}} a_{n}$ nie jest zbieżny;
>
>(3) $\limsup\limits_{n\rightarrow\infty}\sqrt[n]{|a_n|}=1$, to nie można stwierdzić zbieżności bezwzględnej szeregu $\underset{n=1}{\overset{\infty}{\sum}} a_{n}$.

### Ogólniejsza postać kryterium d'Alemberta 

>**Twierdzenie (słabsze od kryterium Cauchy'ego).** Niech dany będzie szereg $\underset{n=1}{\overset{\infty}{\sum}} a_{n}$, gdzie $a_n\neq 0$ dla prawie wszystkich $n\in\BN$. Jeżeli:
>
>(1) istnieją liczby $\rho\in (0,1)$ oraz $n_0\in\BN$ takie, że $|\frac{a_{n+1}}{a_n}|\leq \rho$ dla wszystkich $n\geq n_0$, to szereg $\underset{n=1}{\overset{\infty}{\sum}} a_{n}$ jest bezwzględnie zbieżny;
>
>(2) jeśli $|\frac{a_{n+1}}{a_n}|\geq 1$ dla prawie wszystkich $n\in\BN$, to szereg $\underset{n=1}{\overset{\infty}{\sum}} a_{n}$ nie jest zbieżny.

Inne, równoważne sformułowanie tego kryterium:
>**Wniosek.**
Jeżeli $a_n\neq 0$ dla prawie wszystkich $n\in\BN$ oraz 
>
>(1) $\limsup\limits_{n\rightarrow\infty}|\frac{a_{n+1}}{a_n}|<1$, to szereg $\underset{n=1}{\overset{\infty}{\sum}} a_{n}$ jest bezwzględnie zbieżny;
>
>(2) $\liminf\limits_{n\rightarrow\infty}|\frac{a_{n+1}}{a_n}|\leq 1\leq \limsup\limits_{n\rightarrow\infty}|\frac{a_{n+1}}{a_n}|$, to kryterium nie rozstrzyga zbieżności szeregu $\underset{n=1}{\overset{\infty}{\sum}} a_{n}$;
>
>(3) $\liminf\limits_{n\rightarrow\infty}|\frac{a_{n+1}}{a_n}|>1$, to szereg $\underset{n=1}{\overset{\infty}{\sum}} a_{n}$ nie jest zbieżny.

Dlaczego warto znać te kryteria? 

Ponieważ dzięki nim możemy sprawdzać/weryfikować zbieżność albo rozbieżność w znacznie szerszej klasie szeregów liczbowych o wyrazach nieujemnych.

**Przykład** (o przewadze kryterium Cauchy'ego nad kryterium d'Alemberta). Rozważmy szereg $\underset{n=1}{\overset{\infty}{\sum}} a_{n}$, gdzie

$$
a_{2n-1}=\frac{1}{2^n}\hspace{0.4cm}\text{ oraz }\hspace{0.4cm}a_{2n}=\frac{1}{3^n},\;\;n\in\BN.
$$

Wtedy

$$
\liminf\limits_{n\rightarrow\infty}|\frac{a_{n+1}}{a_n}|=\lim\limits_{n\rightarrow\infty} \left(\frac{2}{3}\right)^n=0; \hspace{1cm} \limsup\limits_{n\rightarrow\infty}|\frac{a_{n+1}}{a_n}|=\lim\limits_{n\rightarrow\infty} \left(\frac{3}{2}\right)^n=+\infty
$$

oraz

$$
\liminf\limits_{n\rightarrow\infty}\sqrt[n]{|a_n|}=\lim\limits_{n\rightarrow\infty}\sqrt[2n]{\frac{1}{3^n}}=\frac{1}{\sqrt{3}};\hspace{1cm}\limsup\limits_{n\rightarrow\infty}\sqrt[n]{|a_n|}=\lim\limits_{n\rightarrow\infty}\sqrt[2n]{\frac{1}{2^n}}=\frac{1}{\sqrt{2}}<1.
$$

 **Wniosek:** kryterium d'Alemberta nie daje rozstrzygnięcia, a kryterium Cauchy'ego implikuje zbieżność (nawet bezwzględną) szeregu $\underset{n=1}{\overset{\infty}{\sum}} a_{n}$.
