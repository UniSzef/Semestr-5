## Granica ciągu

Fundamentalnym pojęciem dla Analizy Matematycznej jest pojęcie granicy ciągu liczbowego. Mówi ono o tym jak zachowują się wyrazy tego ciągu odpowiadające "bardzo dużym" indeksom.

### Definicja ciągu zbieżnego

>Ciąg $(a_n)_{n\in\mathbb{N}}$ nazywamy zbieżnym do liczby $g$, gdy dla każdego $\varepsilon>0$ istnieje takie $N$, że dla każdego $n\geq N$ mamy
>
>$$
|a_n-g| < \varepsilon.
$$
>
>Piszemy wtedy, że  $\displaystyle\lim_{n\to_\infty}a_n=g$ lub $a_n\xrightarrow{n\to\infty}g$. Jeśli ciąg $(a_n)$ jest zbieżny do pewnej liczby $g$, to nazywamy go ciągiem zbieżnym.


__Uwaga:__ 


***To, że liczba $g$ jest granicą ciągu $a_n$ oznacza, że dla każdego $\varepsilon>0$  w przedziale $(g-\varepsilon,g+\varepsilon)$ znajdują się prawie wszystkie wyrazy ciągu $a_n$.***

### Przykład ciągu zbieżnego

Wykres pierwszych $100$ wyrazów ciągu 

$$
a_n=\frac{2n+3}{n+1}
$$

sugeruje, że  wraz ze wzostem indeksu wyrazy tego ciągu znajdują się coraz bliżej 2.

```python
import matplotlib.pyplot as plt
import math   
x = range(1, 100,1)

# Calculate the values of the sequence for n = 1 to n = 50
y = [(2*n-5)/(n+1) for n in range(1, 100,1)]

# Set the axis range
plt.ylim(-2, 3)

plt.scatter(x, y)

# Add the red dashed line at y = 2
plt.axhline(y=2, color='r', linestyle='--')


plt.show()
```

Spróbujmey pokazać zatem, że granicą ciągu $(a_n)_{n\in\mathbb{N}}$ jest liczba $2$.


Dla dowolnego $\varepsilon>0$ chcemy znaleźć taką liczbę naturalną  $N$, aby dla każdego $n \geq N$  zachodziło

$$
\left|\frac{2n+3}{n+1} - 2 \right| < \varepsilon.
$$

Zauważmy, że

$$
\left| \frac{2n+3}{n+1} - 2 \right| = \left| \frac{2n+3}{n+1} - \frac{2(n+1)}{n+1} \right| = \left| \frac{1}{n+1} \right| = \frac{1}{n+1}.
$$

Wybierzmy zatem dowolną liczbę  $N$  większą niż $\frac{1}{\varepsilon}$. Dla każdego  $n \geq N$ mamy:

$$
\frac{1}{n+1} \leq \frac{1}{N+1} < \varepsilon.
$$


Pokazaliśmy, że 

$$
\lim_{n\to\infty}\frac{2n+3}{n+1}=2.
$$

### Ciągłość pierwiastka

Niech $(a_n)_{n\in\mathbb{N}}$ będzie ciągiem liczb nieujemnych zbieżnym do $g$.
Pokażemy, że ciąg $\left(\sqrt{a_n}\right)_{n\in\mathbb{N}}$ jest zbieżny do $\sqrt{g}$.

__Przypadek 1:__ $g=0$

Niech $\varepsilon > 0$ będzie dowolne. Z założenia istnieje takie $N$, że dla wszystkich $n\geq N$ mamy

$$
\left|a_n \right| < \varepsilon^2.
$$

Zatem dla $n\geq N$ mamy

$$
\sqrt{a_n} < \varepsilon.
$$

To pokazuje, że 

$$
\lim_{n\to\infty} \sqrt{a_n}=0.
$$

__Przypadek 2:__ $g > 0$ 

Ponieważ ciąg $(a_n)_{n\in\mathbb{N}}$ jest zbieżny do $g$ to możemy założyć bez utraty ogólności, że 
dla każdego $n\in\mathbb{N}$ mamy

$$
a_n > \frac{g}{4}.
$$
Zauważmy, że 

$$
\left|\sqrt{a_n} - \sqrt{g}\right| = \left|\left(\sqrt{a_n} - \sqrt{g}\right) \cdot \frac{\sqrt{a_n} + \sqrt{g}}{\sqrt{a_n} + \sqrt{g}} \right| = \left| \frac{a_n - g}{\sqrt{a_n} + \sqrt{g}} \right| <  \frac{\left|a_n - g\right|}{\frac{\sqrt{g}}{2} + \sqrt{g}}= 
\frac{2}{3\sqrt{g}}\left|a_n - g\right|.
$$

Niech $\varepsilon > 0$ będzie dowolne. Z założenia istnieje takie $N$, że dla wszystkich $n\geq N$ mamy

$$
\left|a_n-g\right| < \frac{3\sqrt{g}}{2}\varepsilon.
$$

Zatem dla $n\geq N$ mamy

$$
|\sqrt{a_n} - \sqrt{g}|< \varepsilon.
$$

To pokazuje, że 

$$
\lim_{n\to\infty} \sqrt{a_n}=\sqrt{g}.
$$

### Zadanie do samodzielnego rozwiązania

Niech $(a_n)_{n\in\mathbb{N}}$ będzie ciągiem o wyrazach dodatnich, zbieżnym do liczby dodatniej $g$.
Pokaż, że ciąg $\left(\frac{1}{a_n}\right)_{n\in\mathbb{N}}$ jest zbieżny do $\frac{1}{g}$.

## Obliczanie granic wybranych ciągów

### Ważne granice

>__Twierdzenie__
>1. Dla dowolnego $\alpha > 0$
>
>$$
\lim_{n\to\infty}\frac{1}{n^\alpha}=0.
$$
>
>2. Dla dowolnego $|a| < 1$ 
>
>$$
\lim_{n\to\infty} a^n=0.
$$
>
>3. Dla dowolnego $a > 0$ 
>
>$$
\lim_{n\to\infty} \sqrt[n]{a}=1.
$$
>
>4. Mamy  
>
>$$
\lim_{n\to\infty}\sqrt[n]{n}=1.
$$
>
> 5. Mamy  
>
>$$
\lim_{n\to\infty}\left(1+\frac{1}{n}\right)^n=e.
$$

```python
import matplotlib.pyplot as plt
import math   
x = range(1, 50,1)

# Calculate the values of the sequence for n = 1 to n = 50
y = [n ** (1/n) for n in range(1, 50,1)]

# Set the axis range
plt.ylim(-2, 3)

plt.scatter(x, y)

# Add the red dashed line at y = 1
plt.axhline(y=1, color='r', linestyle='--')

# Set text under the x-axis using xlabel()
plt.xlabel('Pierwsze 50 wyrazów ciągu n^{1/n}', fontsize=12, color='blue')


plt.show()
```

### Arytmetyczne własności granicy

>__Twierdzenie__
>
> Niech $(a_n)_{n\in\mathbb{N}}$ i  $(b_n)_{n\in\mathbb{N}}$ będą ciągami zbieżnymi takimi, że $\displaystyle\lim_{n\to_\infty}a_n=A$ i $\displaystyle\lim_{n\to_\infty}b_n=B$. Wtedy:
>1) $\displaystyle\lim_{n\to_\infty}c\cdot a_n=c\cdot A$ dla dowolnego $c\in\mathbb{R}$;
>
>2) $\displaystyle\lim_{n\to_\infty}(a_n+b_n)=A+B$;
>
>3) $\displaystyle\lim_{n\to_\infty}(a_n-b_n)=A-B$;
>
>4) $\displaystyle\lim_{n\to_\infty}(a_n\cdot b_n)=A\cdot B$.
>
>5) Jeżeli $B\not=0$ i  $b_n\not=0$ dla każdego $n\in\mathbb{N}$, to
>  $\displaystyle\lim_{n\to_\infty}\frac{a_n}{b_n}=\frac{A}{B}$.
>
>6) Jeżeli $A > 0$ i $a_n > 0$ dla każdego $n\in\mathbb{N}$, to 
> $\displaystyle\lim_{n\to_\infty}a_n^{b_n}=A^B$.

### Przykłady
1. Obliczymy granicę ciągu 

$$
a_n=\frac{4n}{7n+1}.
$$

Mamy 

$$
a_n=\frac{4n}{7n+1}=\frac{\frac{4n}{n}}{\frac{7n}{n}+\frac{1}{n}}=\frac{4}{7+\frac{1}{n}}.
$$

Ponieważ

$$ 
\lim_{n\to\infty} 4=4 \text{ oraz } \lim_{n\to\infty}7+\frac{1}{n}=7,
$$

to

$$
\lim_{n\to\infty}a_n=\frac{4}{7}.
$$

2. Obliczymy granicę ciągu 

$$
a_n=\frac{n^3+n^2}{4n^4+ 14n}.
$$

Dzieląc licznik i mianownik przez $n^4$ otrzymujemy, że 

$$
a_n= \frac{n^3+n^2}{4n^4+ 14n}=\frac{\frac{1}{n}+\frac{1}{n^2}}{4+\frac{14}{n^3}}.
$$

Ponieważ

$$ 
\lim_{n\to\infty}\frac{1}{n}+\frac{1}{n^2}=0 \text{ oraz } \lim_{n\to\infty}4+\frac{14}{n^3}=4,
$$

to

$$
\lim_{n\to\infty}a_n=0.
$$



3. Obliczymy granicę ciągu

$$
a_n=\frac{25^n+3^n}{5^{2n+1}+4^n}.
$$

Dzieląc licznik i mianownik przez $25^n$
dostajemy, że 

$$
a_n=\frac{25^n+3^n}{5^{2n+1}+4^n}=\frac{1+\left(\frac{3}{25}\right)^n}{5+\left(\frac{4}{25}\right)^n}.
$$

Ponieważ

$$
\lim_{n\to\infty} 1+\left(\frac{3}{25}\right)^n=1 \text{ oraz } \lim_{n\to\infty}5+\left(\frac{4}{25}\right)^n=5,
$$

to

$$
\lim_{n\to\infty}a_n=\frac{1}{5}.
$$

4. Obliczymy granicę ciągu

$$
a_n=\frac{n^2}{\sqrt{n^4+1}}.
$$




Dzieląc licznik i mianownik przez $n^2$ otrzymujemy, że 

$$
a_n=\frac{n^2}{\sqrt{n^4+1}}=\frac{1}{\sqrt{1+\frac{1}{n^4}}}.
$$

Ponieważ

$$
\lim_{n\to\infty} 1=1 \text{ oraz } \lim_{n\to\infty}\sqrt{1+\frac{1}{n^4}}=1,
$$

to 

$$
\lim_{n\to\infty}a_n=1.
$$

5. Obliczymy granicę ciągu

$$
a_n=\sqrt{n^2+4n+1}-n.
$$

Mamy

$$
a_n=\sqrt{n^2+4n+1}-n=\left(\sqrt{n^2+4n+1}-n\right)\cdot \frac{\sqrt{n^2+4n+1}+n}{\sqrt{n^2+4n+1}+n}=
\frac{4n+1}{\sqrt{n^2+4n+1}+n}=\frac{4+\frac{1}{n}}{\sqrt{1+\frac{4}{n}+\frac{1}{n^2}}+1}.
$$

Ponieważ 

$$
\lim_{n\to\infty} 4+\frac{1}{n}=4 \text{ oraz } \lim_{n\to\infty}\sqrt{1+\frac{4}{n}+\frac{1}{n^2}}+1=2,
$$

to 

$$
\lim_{n\to\infty}a_n=2.
$$

## Twierdzenie o ciągu monotonicznym i ograniczonym

>__Twierdzenie__
>
> Niech  $(a_n)_{n\in\mathbb{N}}$ będzie ciągiem monotonicznym i ograniczonym. Wtedy $(a_n)_{n\in\mathbb{N}}$ jest ciągiem zbieżnym.

__Uwagi:__
> Wiadomo, że każdy ciąg zbieżny jest ciągiem ograniczonym. Nie musi być jednak ciągiem monotonicznym. Dla przykładu:
ciąg $\left(\frac{(-1)^n}{n}\right)_{n\in\mathbb{N}}$ jest zbieżny ale nie jest monotoniczny.

### Przykład
Niech $(a_n)$ będzie ciągiem określonym nastepująco:

$$
a_1=2, a_{n+1}=2-\frac{1}{a_n}.
$$

```python
a = 2
for i in range(30):
    print(a)
    a= 2-1/a
```

Powyższe rachunki sugerują trzy rzeczy: wyrazy ciągu $(a_n)_{n\in\mathbb{N}}$ znajdują się między $1$ a $2$, ciąg ten jest malejący i jego granicą chyba jest 1. Teraz to pokażemy.

1. Oczywiście $1 < a_1 \leq 2$. Ponadto jeśli dla pewnego $n\in\mathbb{N}$ mamy, $1 < a_{n}\leq 2$, to 

$$
1 < 2-\frac{1}{a_n} < 2.
$$

Zatem na mocy Indukcji Matematycznej nasz ciąg jest ograniczony, co więcej wiemy że jego wyrazy począwszy od drugiego  znajdują się w przedziale $(1,2)$.


2. Dla $n\in\mathbb{N}$ mamy 

$$a_{n+1}-a_n=2-\frac{1}{a_n}-a_n=\frac{-a_n^2+2a_n-1}{a_n}=-\frac{(a_n-1)^2}{a_n} < 0$$

zatem nasz ciąg jest malejący.

3. Na mocy *Twierdzenie o ciągu monotonicznym i ograniczonym* ciąg $(a_n)_{n\in\mathbb{N}}$ jest zbieżny. Niech $g$ będzie jego granicą. Ponieważ 

$$
a_{n+1}=2-\frac{1}{a_n},
$$

to 

$$
g=2-\frac{1}{g}.
$$

Rozwiązując to łatwe równanie dostajemy, że $g=1$.

### Przykład

Niech $(s_n)$ będzie ciągiem danym wzorem 

$$
s_n=\sum_{i=1}^n\frac{1}{i(i+1)}.
$$

Zauważmy, że 

$$
\begin{align}
s_n =&\frac{1}{1\cdot 2}+
\frac{1}{2\cdot 3}+
\ldots +\frac{1}{n\cdot n+1}
\\
=&1-\frac{1}{2}+\frac{1}{2}-\frac{1}{3}+
\ldots +\frac{1}{n}-\frac{1}{n+1}\\
=& 1-\frac{1}{n+1}.
\end{align}
$$

Stąd

$$
\lim_{n\to\infty}s_n=1.
$$

__Uwaga:__


> Tak naprawdę pokazaliśmy, że szereg $\displaystyle \sum_{i=1}^\infty\frac{1}{i(i+1)}$ jest zbiezny i jego sumą jest $1$.
