# Całka Riemanna i jej zastosowania

## Cel zajęć

Opanowanie definicji całki Riemanna, jej związku z całką nieozaczoną oraz zastosowań takich jak obliczanie pól powierzchni, długości krzywych i objętości brył obrotowych.

## Całka Riemanna

### Przykład funkcji całkowalnej
Pokażemy, że funkcja $f(x)=x$ jest całkowalna na $[0,1]$ i obliczymy
wprost z definicji 

$$
\int_0^1xdx.
$$

Przypomnijmy, że funkcja jest całkowalna w sensie Riemanna, gdy
całka dolna z tej funkcji jest równa całce górnej z tej funkcji.

Niech $n$ będzie dowolną liczbą naturalną większą bądź równą od dwóch.
Niech

$$
\mathcal{P}_n=\{x_0=0,x_1=\frac{1}{n},x_2=\frac{2}{n},\ldots,x_{n-1}=\frac{n-1}{n},x_n=1\}
$$

będzie podziałem
odcinka $[0,1]$ (dzielimy odcinek $[0,1]$ na $n$ równych części).
Dla każdego $1\leq j\leq n$ mamy (korzystamy z notacji z wykładu):

$$
\Delta x_j=x_{j}-x_{j-1}=\frac{j}{n}-\frac{j-1}{n}=\frac 1 n;
$$

$$
M_j=\sup_{x_{j-1}\leq x\leq x_j}x=x_j=\frac{j}{n}
$$

oraz

$$
m_j=\inf_{x_{j-1}\leq x\leq x_j}x=x_{j-1}=\frac{j-1}{n}.
$$

Zatem

$$
U(f,\mathcal{P}_n)=\sum_{j=1}^n M_j\Delta x_j=
\sum_{j=1}^n \frac{j}{n}\cdot\frac{1}{n}
=\frac{1}{n^2}\sum_{j=1}^nj=\frac{1}{n^2}\frac{n(n+1)}{2}=\frac{n+1}{2n},
$$

gdzie przedostatnia równość wynika ze wzoru na sumę ciągu arytmetycznego. Podobnie

$$
L(f,\mathcal{P}_n)=\sum_{j=1}^n m_j\Delta x_j=
\sum_{j=1}^n \frac{j-1}{n}\cdot \frac{1}{n}
=\frac{1}{n^2}\sum_{j=1}^n(j-1)=\frac{1}{n^2}\frac{n(n-1)}{2}=\frac{n-1}{2n}.
$$

Stąd

$$
\underline{\int_0^1xdx}=\sup_{\mathcal{P}} L(f,\mathcal{P})
\geq L(f,\mathcal{P}_n)=\frac{n-1}{2n}.
$$

Ponieważ $n$ było dowolne, to w powyższej nierówności możemy przejść obustronnie do granicy przy $n\to\infty$. Dzięki temu otrzymujemy, że 

$$
\underline{\int_0^1xdx}\geq \frac{1}{2}.
$$

Podobnie otrzymujemy, że

$$
\overline{\int_0^1xdx}=\inf_{\mathcal{P}} U(f,\mathcal{P})
\leq U(f,\mathcal{P}_n)=\frac{n+1}{2n}.
$$

Stąd

$$
\overline{\int_0^1xdx}\leq \frac{1}{2}.
$$

Mamy zatem (całka dolna jest zawsze mniejsza bądź równa niż całka górna)


$$
\frac{1}{2}\leq \underline{\int_0^1xdx}\leq\overline{\int_0^1xdx}
\leq \frac 1 2.
$$

Więc

$$
\underline{\int_0^1xdx}=\overline{\int_0^1xdx}
= \frac 1 2.
$$

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the function to integrate
def f(x):
    return x

# Define the interval [a, b] and the number of subintervals (n)
a = 0
b = 1
n = 15  # You can adjust the number of subintervals as needed

# Create a list of sample points within the interval [a, b]
x_values_n = [a + i * (b - a) / n for i in range(n+1)]

# Calculate the width of each subinterval
dx = (b - a) / n

# Calculate the corresponding function values
y_values_n = [f(x) for x in x_values_n]



# Calculate the lower Riemann sum
lower_riemann_sum = sum(y_values_n[:-1]) * dx

# Create a bar plot to visualize the lower Riemann sum
plt.bar(x_values_n[:-1], y_values_n[:-1], width=dx, align='edge',  edgecolor='b', color='blue', alpha=0.2)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title(f'Lower Riemann Sum for $\int_{{0}}^{{1}} x dx$ = {lower_riemann_sum:.4f}')

# Create a list of sample points within the interval [a, b]
x_values = np.linspace(a, b, 1000)  # For smoother function plot
y_values = [f(x) for x in x_values]
plt.plot(x_values, y_values, color='red', label='$f(x) = x$')

# Show the plot
plt.show()

```

### Przykład funkcji niecałkowalnej

Niech $f\colon\mathbb{R}\to\mathbb{R}$ będzie dana wzorem

$$
f(x)=\begin{cases}
1, & x\in \mathbb{Q}\\
0, & x\not\in \mathbb{Q}.
\end{cases}
$$


Pokażemy, że funkcja $f$ nie jest całkowalna na żadnym przedziale $[a,b]$, gdzie $a<b$.
By to zrobić ustalmy dowolny podział $\mathcal{P}$


$$
a=x_0<x_1<x_2<\ldots<x_{n-1}<x_n=b
$$
odcinka $[a,b]$.
Ponieważ w każdym przedziale znajduje się liczba wymierna jak i liczba niewymierna, 
to dla każdego $1\leq j\leq n$ mamy

$$
M_j=\sup_{x_{j-1}\leq x\leq x_j} f(x)=1
$$

oraz 

$$
m_j=\inf_{x_{j-1}\leq x\leq x_j} f(x)=0
$$

Zatem

$$
U(f,\mathcal{P})=\sum_{j=1}^n M_j\Delta x_j=\sum_{j=1}^n \Delta x_j=b-a
$$
oraz

$$
L(f,\mathcal{P})=\sum_{j=1}^n m_j\Delta x_j=0.
$$

Ponieważ podział $\mathcal{P}$ był dowolny, to 

$$
\underline{\int_a^bf(x)dx}=0
$$
oraz

$$
\overline{\int_a^bf(x)dx}=(b-a).
$$

Zatem $f$ nie jest całkowalna na przedziale $[a,b]$.

$$

## Całka Riemanna a całka nieoznaczona

>__Twierdzenie (Wzór Newtona-Leibniza)__ 
>Niech $f\colon[a,b]\to\mathbb{R}$ będzie funkcją całkowalną na przedziale $[a,b]$. Jeśli istnieje funkcja różniczkowalna $F$ 
> taka, że 	$F'(x)=f(x)$ dla
>$x\in [a,b]$, to 
>
>$$
\int_a^bf(x)dx=F(b)-F(a).
$$

__Uwaga:__ dla ułatwienia  zapisu będziemy używać następującej notacji:

$$
F(x)\biggr|_a^b=F(b)-F(a).
$$


### Przykłady
1. Wiemy, że

$$
\int xdx=\frac{x^2}{2}+C.
$$

Zatem dla dowolnych $a,b$ mamy na mocy wzoru Newtona-Leibniza

$$
\int_a^bxdx=\frac{x^2}{2}\biggr|_a^b=\frac{b^2}{2}-\frac{a^2}{2}.
$$

W szczególności

$$
\int_0^1xdx=\frac{x^2}{2}\biggr|_0^1=\frac{1}{2}.
$$

2. Wiemy, że (łatwe całkowanie przez części)

$$
\int \operatorname{ln}xdx=x\operatorname{ln}x-x+C
$$

Zatem

$$\int_1^e \operatorname{ln}xdx=x\operatorname{ln}x-x\biggr|_1^e=
(e\operatorname{ln}e-e)-(1\operatorname{ln}1-1)=1.
$$

3. Wiemy, że (łatwe całkowanie przez podstawienie), że

$$
\int 2\sin x\cos xdx=\sin^2x+C.
$$

Zatem

$$
\int_0^\frac{\pi}{2}2\sin x\cos xdx=\sin^2x\biggr|_0^{\frac{\pi}{2}}=
\sin^2\frac{\pi}{2}-\sin^20=1.
$$

4. Obliczymy

$$
\int_1^3|x-2|dx.
$$

Mamy 

$$
\begin{align*}
\int_1^3|x-2|dx&=\int_1^2|x-2|dx+\int_2^3|x-2|dx\\
&=\int_1^22-xdx+\int_2^3x-2dx\\
&=\left(2x-\frac{x^2}{2}\right)\biggr|_1^2+\left(\frac{x^2}{2}-2x\right)\biggr|_2^3\\
&=(4-2)-(2-\frac{1}{2})+(\frac{9}{2}-6)-(2-4)\\
&=1.
\end{align*}
$$

### Przykłady do samodzielnego rozwiązania

\int_{-2}^2 x^3+x^5+x^7dx, \quad \int_0^4|x-1|dx.

## Zastosowania

### Obliczanie pól

#### Pole zbioru znajdującego się między wykresami dwóch funkcji

>Niech $f,g\colon [a,b]\to\mathbb{R}$ będą funkcjami ciągłymi takimi, że 
>$f(x)\leq g(x)$ dla $x\in[a,b]$. Wtedy pole zbioru
>
>$$
D=\{(x,y)\colon a\leq x\leq b, f(x)\leq y\leq g(x)\}
$$
>wyraża się wzorem
>
>$$
\int_a^bg(x)-f(x)dx.
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the interval
x = np.linspace(0, 2*np.pi, 100)

# Define two functions (you can replace these with your own functions)
function1 = x
function2 = np.cos(x)

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the two functions
plt.plot(x, function1, label='x', color='red')
plt.plot(x, function2, label='cos(x)', color='green')

# Shade the area between the two functions in blue
plt.fill_between(x, function1, function2, where=(function1 > function2), interpolate=True, color='blue', alpha=0.3)


# Show the plot
plt.grid(True)
plt.show()
```

#### Pole koła

Koło to zbiór punktów 
znajdujących się między wykresami funkcji
$g(x)=\sqrt{1-x^2}$ oraz $f(x)=-\sqrt{1-x^2}$, gdzie $x\in [-1,1]$.
Zatem pole jest równe (całkę, która się teraz pojawi kilka razy już liczyliśmy)

$$
\begin{align*}
2\int_{-1}^1\sqrt{1-x^2}dx=\left(x\sqrt{1-x^2}+\operatorname{arcsin}x \right)\biggr|_{-1}^1=\pi.
\end{align*} 
$$

__Przykład__
Obliczymy pole figur ograniczonej przez wykresy funkcji 
$g(x)=e^{-x}$, $f(x)=e^x$ oraz prostą $x=1$. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the interval
x = np.linspace(-1, 3, 400)

# Define the functions
def f(x):
    return np.exp(x)

def g(x):
    return np.exp(-x)

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the functions
plt.plot(x, f(x), label='f(x) = e^x', color='red')
plt.plot(x, g(x), label='g(x) = e^(-x)', color='green')

# Fill the area between the two functions and the line x=1
x_fill = np.linspace(0, 1, 400)
plt.fill_between(x_fill, f(x_fill), g(x_fill), interpolate=True, color='blue', alpha=0.3)

# Add the vertical line x=1
plt.axvline(x=1, color='purple', linestyle='--', label='x = 1')


# Show the plot
plt.grid(True)
plt.ylim(0, 10)  # You can adjust the y-axis limits if needed
plt.xlim(-1, 3)  # You can adjust the x-axis limits if needed
plt.show()
```

Z rysunku łatwo zobaczyć, że należy policzyć

$$
\begin{align*}
\int_0^1 e^{x}-e^{-x}dx=e^x+-e^{-x}\biggr|_0^1=e+e^{-1}-2.
\end{align*}
$$


__Przykład__

Obliczymy pole figury ograniczonej przez 
parabolę $y=x^2$ oraz parabolę $y=8-x^2$. 
Rozwiązując odpowiedni układ równań dostajemy, że 
parabole te przecinają się w punktach $(-2,4)$ i $(2,4)$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the interval
x = np.linspace(-3, 3, 400)

# Define the functions
def f(x):
    return x**2

def g(x):
    return 8 - x**2

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the functions
plt.plot(x, f(x), label='y = x^2', color='red')
plt.plot(x, g(x), label='y = 8 - x^2', color='green')

# Fill the area between the two functions
plt.fill_between(x, f(x), g(x), where=(f(x) < g(x)), interpolate=True, color='blue', alpha=0.3)

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Show the plot
plt.grid(True)
plt.title('Area between y = x^2 and y = 8 - x^2')
plt.ylim(0, 10)  # Adjust the y-axis limits if needed
plt.xlim(-3, 3)  # Adjust the x-axis limits if needed
plt.show()
```

Zatem pole figury jest równe
$$
\begin{align*}
\int_{-2}^28-x^2-x^2dx=\int_{-2}^28-2x^2dx=
8x-\frac{2}{3}x^3\biggr|_{-2}^2=\left(16-\frac{16}{3}\right)-\left(-16+\frac{16}{3}\right)=32-\frac{32}{3}.
\end{align*}
$$

#### Pole pod krzywą zadaną parametrycznie

>Niech $\Gamma\colon[a,b]\to\mathbb{R}^2$, $\Gamma(t)=(x(t), y(t))$, $t\in[a,b]$ będzie łukiem gładkim taki, że 
$x'(t)>0$ dla $t\in[a,b]$.
 Pole figury $D$ zawartej
pomiędzy tą krzywą a osią $OX$ i odcinkami równoległymi do osi $OY$ poprowadzonymi
przez odcięte w punktach końcowych krzywej, wyraża się wzorem
>
>$$
\int_a^by(t)x'(t)dt.
$$


>Niech $\Gamma\colon[a,b]\to\mathbb{R}^2$, $\Gamma(t)=(x(t), y(t))$, $t\in[a,b]$ będzie łukiem gładkim taki, że 
$x'(t)<0$ dla $t\in[a,b]$.
 Pole figury $D$ zawartej
pomiędzy tą krzywą a osią $OX$ i odcinkami równoległymi do osi $OY$ poprowadzonymi
przez odcięte w punktach końcowych krzywej, wyraża się wzorem
>
>$$
-\int_a^by(t)x'(t)dt.
$$

#### Przykład: pole pod cykloidą
Policzymy pole zbioru ograniczonego osią $0X$ oraz fragmentem cykloidy 
zadanej przez równania parametryczne

$$
x(t)=t-\sin(t),\quad y(t)=1-\cos t, \quad t\in[0,2\pi].
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameter t
t = np.linspace(0, 2*np.pi, 100)

# Define the parametric equations
x = t - np.sin(t)
y = 1 - np.cos(t)

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the parametric curve
plt.plot(x, y, color='blue')




# Show the plot
plt.grid(True)
plt.title('Parametric Curve: x(t) = t - sin(t), y(t) = 1 - cos(t)')
plt.xlim(0, 7)  # Adjust the x-axis limits to match the interval
plt.ylim(0, 2.5)  # Adjust the y-axis limits if needed
plt.show()
```

Zauważmy, że 

$$
x'(t)=1-\cos t, y'(t)=\sin(t).
$$

Ponieważ $x'(t)\geq 0$ dla wszystkich $t$, to 
pole jest równe 

$$
\begin{align*}
\int_0^{2\pi}y(t)x'(t)dt=\int_{0}^{2\pi}(1-\cos t )^2dt
=\frac{3}{2}t-2 \sin(t)+\frac{1}{4}\sin(2t)\biggr|_0^{2\pi}=3\pi.
\end{align*}
$$

#### Pole zbioru ograniczonego przez krzywą we współrzędnych biegunowych
>Jeśli krzywa ma we współrzędnych biegunowych równanie	
>
>$$
r=r(\varphi),\, 0\leq\alpha\leq\varphi\leq\beta\leq 2\pi,
$$
>
>to pole zbioru ograniczonego przez półproste nachylone do osi $OX$ pod kątem $\alpha$ i $\beta$ oraz tą krzywą
wyraża się wzorem
>
>$$
\frac{1}{2}\int_\alpha^\beta r^2(\varphi)d\varphi.
$$

#### Przykład: pole kardioidy

Obliczymy pole figury ograniczonej kardioidą

$$
r(\varphi)=\cos\varphi+1,\quad\varphi\in[0,2\pi].
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate values for Φ (phi) from 0 to 2π
phi = np.linspace(0, 2 * np.pi, 1000)

# Calculate the corresponding values of r
r = np.cos(phi) + 1

# Create the polar plot
plt.figure(figsize=(8, 6))
plt.polar(phi, r, label='r(Φ) = cos(Φ) + 1', color='blue')

# Add labels and legend
plt.legend()

# Show the polar plot
plt.title('Polar Plot: r(Φ) = cos(Φ) + 1')
plt.show()
```

Pole to jest równe 
\begin{align}
\frac{1}{2}\int_0^{2\pi}r^2(\varphi)d\varphi=
\frac{1}{2}\int_0^{2\pi}(1+\cos t)^2d\varphi=\frac{3}{2}t+2 \sin(t)+\frac{1}{4}\sin(2t)\biggr|_0^{2\pi}=3\pi=3\pi.
\end{align}

### Obliczanie dlugości krzywych

#### Długość krzywej będącej wykresem funkcji

>Niech $f\colon[a,b]\to\mathbb{R}$ będzie funkcją klasy $
\mathcal{C}^1$. Wtedy długość wykresu funkcji $f$ wyraża się wzorem
>
>$$
\int_a^b\sqrt{1+(f'(x))^2}dx.
$$

#### Przykład: obwód okręgu
Policzymy teraz obwód okręgu o promieniu $R$ i środku w punkie $(0,0)$.
Wystarczy policzyć długość części okręgu znajdującej się w pierwszej i drugiej ćwiartce układu współrzędnych. Jest  to tak naprawdę wykres funkcji

$$
f\colon[-R,R]\to\mathbb{R}
$$

danej wzorem

$$
f(x)=\sqrt{R^2-x^2}.
$$

Mamy

$$
f'(x)=\frac{-x}{\sqrt{R^2-x^2}}.
$$

Stąd długość tego fragmentu okręgu jest równa

$$
\begin{align*}
\int_{-R}^R\sqrt{1+(f'(x))^2}dx=&
\int_{-R}^R\sqrt{1+\frac{x^2}{R^2-x^2}}dx=
\int_{-R}^R\sqrt{\frac{R^2}{R^2-x^2}}dx\\
=&\int_{-R}^R\frac{1}{\sqrt{1-\left(\frac{x}{R}\right)^2}}dx=R\cdot\operatorname{arcsin}\left(\frac x R\right)\biggr|_{-R}^R\\
=&R\cdot\frac{\pi}{2}-\left(-R\cdot\frac{\pi}{2}\right)=\pi R.
\end{align*}
$$

Zatem obwód okręgu to 

$$
2\pi R.
$$

#### Długość krzywej  zadanej parametrycznie

> Niech $\Gamma\colon[a,b]\to\mathbb{R}^2$, $\Gamma(t)=(x(t), y(t))$, będzie łukiem gładkim. Jego długość wyraża się wzorem
>
$$
\int_{a}^{b}\sqrt{(x'(t))^2+(y'(t))^2}dt.
$$

#### Przykład: długość asteroidy
Obliczymy długość krzywej zadanej parametrycznie przez równania

$$
x(t)=\cos^3t,\quad y(t)=\sin^3t, \quad t\in[0,2\pi].
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameter t
t = np.linspace(0, 2*np.pi, 100)

# Define the parametric equations
x = (np.sin(t))**3
y = (np.cos(t))**3

# Create the plot
plt.figure(figsize=(8, 6))

# Plot the parametric curve
plt.plot(x, y, color='blue')




# Show the plot
plt.grid(True)
plt.xlim(-2, 2)  # Adjust the x-axis limits to match the interval
plt.ylim(-2, 2)  # Adjust the y-axis limits if needed
plt.show()
```

Krzywa ta zwana jest asteroidą.
Mamy 

$$
x'(t)=-3\cos^2 t\sin t, \quad y'(t)=3\sin^2t\cos t.
$$
Zatem 

$$
(x'(t))^2+(y'(t))^2=\frac{9}{4}\sin^22t.
$$

Stąd długość tej krzywej jest równa

$$
\int_0^{2\pi}\sqrt{\frac{9}{4}\sin^22t}dt=\frac{3}{2}\int_{0}^{2\pi}
|\sin(2t)|dt=\ldots=6.
$$

#### Długość łuku zadanego we współrzędnych biegunowych

>Jeśli krzywa ma we współrzędnych biegunowych równanie	
>$r=r(\varphi)$, $\alpha\leq\varphi\leq\beta$, gdzie funkcja $r(\varphi)$ jest nieujemna,  to jej długość wyraża się wzorem

$$
\int_\alpha^\beta\sqrt{r^2(\varphi)+(r'(\varphi))^2}d\varphi.
$$

#### Przykład: długość kardioidy

Obliczymy długośc krzywej zadanej we współrzędnych biegunowych równaniem

$$
r(\varphi)=\cos\varphi+1,\quad\varphi\in[0,2\pi].
$$



Mamy 

$$
r'(\varphi)=-\sin\varphi.
$$

Zatem

$$
(r(\varphi))^2+(r'(\varphi))^2=2+2\cos \varphi=4\cos^2\frac{\varphi}{2},
$$

gdzie przedostatnia równość wynika ze wzoru na cosinus podwojonego kąta.
Stąd długość naszej krzywej to

$$
\int_0^{2\pi}\sqrt{(r(\varphi))^2+(r'(\varphi))^2}d\varphi

$$

### Obliczanie objętości i pola powierzchni bryły obrotowej

>Niech $f\colon[a,b]\to\mathbb{R}$ będzie nieujemną funkcją całkowalną.
>Objętość bryły obrotowej powstałej przez obrót wykresu funkcji 
>$f$ dookoła osi $OX$ wyraża się wzorem
>
>$$
\pi\int_a^bf^2(x)dx.
$$
>Pole powierzchni tej bryły obrotowej wyraża się wzorem
>
$$
2\pi\int_a^bf(x)\sqrt{1+(f'(x))^2}dx.
$$

#### Przykład: objętośći i pole powierzchni kuli
Obliczymy objętość kuli o promieniu $R$ i jej pole powierzchni.
Zauważmy, że by otrzymać kulę jako bryłę obrotową, trzeba obrócić dookoła osi  $OX$ wykres funkcji $f(x)=\sqrt{R^2-x^2}$, $x\in[-R,R]$.
Zatem objętość tej bryły jest równa

$$
\begin{align*}
\pi\int_{-R}^{R}R^2-x^2dx=\pi\left(R^2x-\frac{x^3}{3}\right)\biggr|_{-R}^R=
\pi \left(R^3-\frac{R^3}{3}-\left(-R^3+\frac{R^3}{3}\right)\right)=\frac{4}{3}R^3\pi.
\end{align*}
$$

Ponadto 

$$
f'(x)=\frac{-x}{\sqrt{R^2-x^2}}.
$$

Zatem pole jest równe

$$
\begin{align*}
2\pi\int_{-R}^Rf(x)\sqrt{1+(f'(x))^2}dx=&
2\pi\int_{-R}^R\sqrt{R^2-x^2}\sqrt{1+\frac{x^2}{R^2-x^2}}dx\\
=&2\pi\int_{-R}^RRdx=4\pi R^2.
\end{align*}
$$
