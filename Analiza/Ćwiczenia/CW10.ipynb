## Cel zajęć

Rachunek różniczkowy jako narzędzie służące do dowodzenia prawdziwości nierówności oraz rozwiązywania pewnych zagadnień natury optymalizacyjnej. Obliczanie granic funkcji z reguły de l'H$\hat{\text{o}}$spitala.

Narzędziami pomocnymi w dowodzeniu prawdziwości nierówności, w których pojawiają się funkcje zmiennej rzeczywistej są: badanie ekstremów lokalnych funkcji i badanie monotoniczoności funkcji. Warto pamiętać, że 

$$
\inf\limits_{x\in D_f}f(x)\leq f(x)\leq \sup\limits_{x\in D_f}f(x)
$$

dla wszystkich $x\in D_f$.

### Przykład

Pokażemy, że dla każdego $x\in\mathbb{R}$ prawdziwa jest nierówność

$$
2x\operatorname{arctg}x\geq \ln(1+x^2).
$$

Nierówność ta, dla każdego $x\in\mathbb{R}$, jest równoważna następującej

$$
2x\operatorname{arctg}x- \ln(1+x^2)\geq 0.
$$

Aby udowodnić jej prawdziwość wystarczy pokazać, że najmniejszą wartością funkcji 

$$
f(x)=2x\operatorname{arctg}x- \ln(1+x^2)
$$

dla argumentów rzeczywistych jest $0$.

W tym celu sprawdzimy, czy funkcja $f$ ma ekstremum lokalne. Mamy 

$$
f^\prime(x)=(2x\operatorname{arctg}x- \ln(1+x^2))^\prime=2 \operatorname{arctg}x+2x\cdot \frac{1}{1+x^2}-\frac{1}{1+x^2}\cdot 2x=2 \operatorname{arctg}x
$$

i $D_{f^\prime}=\BR$. Dalej

$$
f^\prime(x)=0 
$$

wtedy i tylko wtedy, gdy $\operatorname{arctg}x=0$, tzn., gdy $x=0$ - w tym punkcie może być ekstremum lokalne. 

Ponieważ 

$$
f^{\prime\prime}(x)=\frac{2}{1+x^2}, \hspace{1cm} D_{f^{\prime\prime}}=\BR
$$

oraz 

$$
f^{\prime\prime}(0)=2>0,
$$

więc w punkcie $x=0$ funkcja $f$ ma jedyne minimum lokalne właściwe wynoszące $f_{\min}=f(0)=2\cdot 0\cdot\operatorname{arctg}0-\ln(1+0)=0$. Zauważmy przy tym, że analiza monotoniczności funkcji na podstawie znaku pierwszej pochodnej pokazuje, że funkcja $f$ maleje w $(-\infty,0)$ oraz rośnie w $(0,\infty)$.

Z uwagi na fakt, że

$$
f(x)\geq \inf\limits_{x\in\BR}f(x)=\min\limits_{x\in\BR}f(x)=0
$$

dla każdego $x\in\BR$ (infimum jest w tym przypadku osiągnięte), pokazaliśmy, że 

$$
f(x)=2x\operatorname{arctg}x- \ln(1+x^2)\geq 0.
$$

### Przykład 2

Pokażemy prawdziwość nierówności

$$
2\ln x<x-\frac{1}{x}
$$

dla $x>1.$

Nierówność ta jest równoważna następującej

$$
2\ln x-x+\frac{1}{x}<0
$$

dla każdego $x>1$. Oznaczając

$$
f(x)=2\ln x-x+\frac{1}{x},\hspace{1cm}x>1,
$$

sprawdzamy, czy funkcja $f$ ma ekstremum. Mamy

$$
f^\prime(x)=\frac{2}{x}-1-\frac{1}{x^2}, \hspace{1cm}D_{f^{\prime}}=(1,\infty)
$$

oraz 

$$
f^\prime(x)=0 \Leftrightarrow \frac{2}{x}-1-\frac{1}{x^2}=0 \Leftrightarrow \frac{2x-x^2-1}{x^2}=0 \Leftrightarrow \frac{-(x-1)^2}{x^2}=0,
$$

czyli $x=1$ jest miejscem zerowym pierwszej pochodnej. Ale $1\notin D_{f^{\prime}}=(1,\infty)$, więc nie może tam być ekstremum.

Jednak dla $x\in(1,\infty)$ funkcja $f$ jest malejąca ponieważ $f^{\prime}(x)<0$ dla tych argumentów.

W konsekwencji (funkcja malejąca)

$$
f(x)<\sup\limits_{x>1}f(x)=f(1)=2\ln 1-1+1 =0
$$

dla każdego $x>1$ i w ten sposób pokazaliśmy prawdziwość nierówności

$$
2\ln x<x-\frac{1}{x}
$$

dla $x>1.$

dla każdego $x\in [0,\infty)$.

Przejdźmy teraz do przykładowego zdania optymalizacyjnego. Zobaczymy, że rachunek różniczkowy będzie pełnił istotną rolę w rozwiązaniu przedstawionego poniżej w formie pytania problemu.

### Przykład

Odpowiemy na pytanie:

>Jakie wymiary powinno mieć naczynie w kształcie otwartego walca o pojemności $V$ 
i grubości ścianek $a$, tak aby ilość materiału potrzebnego do jego produkcji była najmniejsza? 

Oznaczmy przez $h$ wysokość ścianek wewnętrznego walca naszego naczynia, a przez $r$ promień tego walca. Wówczas wymiary walca zewnętrznego tworzącego naczynie wynoszą (zobacz rysunek): 

- promień walca zewnętrznego: r+a, 

- a wysokość walca zewnętrznego: a+h, 

gdzie $a$ oznacza grubość ścianek naczynia. 


![alt text](naczynie1.png "naczynie")

Pojemność, a raczej objętość walca wewnętrznego wynosi $V$ i wyraża się wzorem $V=\Pi r^2 h$.

![alt text](naczynie2.png "naczynie")

Niech $y$ oznacza funkcję mierzącą ilość materiału potrzebnego do wykonania naczynia. Jej wzór to:

$$
y=\Pi(r+a)^2(a+h)-\Pi r^2 h,
$$

gdzie $a,V>0$ są dane, zaś $r>0$ i $h>0$ są niewiadomymi.

Przekształcając wzór $V=\Pi r^2 h$, otrzymujemy, że 

$$
h=\frac{V}{\Pi r^2}.
$$

Podstawiając za $h$ to wyrażenie do wzoru $y=\Pi(r+a)^2(a+h)-\Pi r^2 h,$ dostajemy wzór na funkcję $y$ jednej zmiennej $r>0$, a mianowicie

$$
y=y(r)=\Pi(r+a)^2\left(a+\frac{V}{\Pi r^2}\right)-V.
$$

Odnotujmy, że $D_y=(0,\infty)$. Dalej

$$
\begin{eqnarray*}
y^\prime(r) & = & \left(\Pi(r+a)^2\left(a+\frac{V}{\Pi r^2}\right)-V\right)^\prime \hspace{1cm}(\text{ pochodna względem zmiennej }r).\\
&=& 2\Pi (r+a)\left(a+\frac{V}{\Pi r^2}\right)+\Pi(r+a)^2\left(-\frac{2V}{\Pi r^3}\right)
\end{eqnarray*}
$$

ponieważ $\left(\frac{V}{\Pi r^2}\right)^\prime _r=\frac{V}{\Pi }(r^{-2})^\prime=\frac{V}{\Pi }(-2r^{-3})$.

Następnie

$$
y^\prime(r) =0 \Leftrightarrow 2\Pi (r+a)\left(a+\frac{V}{\Pi r^2}\right)+\Pi(r+a)^2\left(-\frac{2V}{\Pi r^3}\right)=0 \Leftrightarrow a+\frac{V}{\Pi r^2}-\frac{(r+a)V}{\Pi r^3}=0, 
$$

co jest równoważne, po przemnożeniu przez $\Pi r^3$, temu, że

$$
a\Pi r^3+Vr -rV-aV=0 \Leftrightarrow a(\Pi r^3-V)=0 \Leftrightarrow r^3=\frac{V}{\Pi}. 
$$

Zatem w punkcie $r=\sqrt[3]{\frac{V}{\Pi}}$ może być ekstremum lokalne funkcji $y(r)$, gdzie $r>0$. 

Pokażemy, że jest to punkt minimum lokalnego. Mamy

$$
y^\prime(r)>0  \Leftrightarrow 2\Pi (r+a)\left(a+\frac{V}{\Pi r^2}-\frac{V(r+a)}{\Pi r^3}\right)>0 \Leftrightarrow \frac{2\Pi(r+a)}{\Pi r^3}(a\Pi r^3-Va)>0.
$$

Widać, że przy przejściu z lewostronnego otoczenia punktu $\sqrt[3]{\frac{V}{\Pi}}$ na prawostronne zmienia się monotoniczność funkcji $y(r)$ (z malejącej na rosnącą), zatem w punkcie $r=\sqrt[3]{\frac{V}{\Pi}}$ funkcja $y$ posiada minimum lokalne właściwe. 

Wówczas 

$$
h=\frac{V}{\Pi r^2}=\frac{V}{\Pi \cdot\frac{V}{\Pi}^\frac{2}{3}}=\frac{V}{\pi^\frac{1}{3}\cdot V^\frac{2}{3}}= \left(\frac{V}{\Pi}\right)^\frac{1}{3}.
$$

Widać też, że gdy $r\rightarrow \infty$, to $y(r)\rightarrow \infty$.

Zatem naczynie w kształcie walca powinno mieć wymiary zewnętrzne: 

- $a+r=a+\sqrt[3]{\frac{V}{\Pi}}$ (promień)

- i $a+h=a+ \sqrt[3]{\frac{V}{\Pi}}$ (wysokość).

Przy obliczaniu pewnych granic funkcji w punkcie pomocną okazuje się być pochodna funkcji. Mówi o tym twierdzenie de l'H$\hat{\text{o}}$spitala, zwane również regułą de l'H$\hat{\text{o}}$spitala.

## Reguła de l'H$\hat{\textbf{o}}$spitala

Przypomnijmy najpierw, że przez wydrążone otoczenie $U_\ve$ punktu $a$ o promieniu $\ve>0$ rozumiemy zbiór:

$$
\{x\in \BR\colon 0<\left|x-a\right|<\ve\}=(a-\ve,a)\cup(a,a+\ve), \text{ gdy a jest liczbą rzeczywistą},
$$

oraz zbiór

$$
\{x\in\BR\colon \left|x\right|>\frac{1}{\ve}\}, \text{ gdy a oznacza $\infty$}.
$$

### Twierdzenie 1 
__Pierwszy przypadek reguły de l'H$\hat{\textbf{o}}$spitala $($wyrażenia typu $\frac{0}{0})$.__

Jeżeli: 

a) funkcje $f(x)$ i $g(x)$ są określone i ciągłe w pewnym $($wydrążonym$)$ otoczeniu $U_\ve$ punktu $a$, gdzie $a$ jest liczbą lub symbolem $\infty$ oraz $\lim\limits_{x\rightarrow a}f(x)=\lim\limits_{x\rightarrow a}g(x)=0$,

b) w otoczeniu $U_\ve$ punktu $a$ $($za wyjątkiem, być może, punktu $a)$ istnieją pochodne $f'(x)$ i $g'(x)$ nie zerujące się jednocześnie dla $x\neq a$,

c) istnieje skończona lub nieskończona granica $\lim\limits_{x\rightarrow a}\frac{f'(x)}{g'(x)}$, 
to 

$$
\lim\limits_{x\rightarrow a}\frac{f(x)}{g(x)}=\lim\limits_{x\rightarrow a}\frac{f'(x)}{g'(x)}.
$$

### Uwaga

Analogiczne twierdzenie dla granic jednostronnych jest również prawdziwe.

### Przykład

Obliczymy granicę funkcji

$$
\lim_{x\rightarrow 0}\frac{\ln (1+x)}{x}.
$$

Zauważmy, że: 

a) obie funkcje $f(x)=\ln (1+x)$ oraz $g(x)=x$ są (dobrze) określone i ciągłe w wydrążonym otoczeniu punktu $a=0$. Mamy też $\lim\limits_{x\rightarrow 0}\ln (1+x) =0$ oraz $\lim\limits_{x\rightarrow 0}x =0$. To ostatnie oznacza wystąpienie symbolu nieoznaczonego typu $\frac{0}{0}$.

b) Ponadto $f^{\prime}(x)=\frac{1}{1+x}\neq 0$ i $g^{\prime}(x)=1\neq 0$ dla wszystkich argumentów $x\neq 0$ z (wydążonego) otoczenia punktu $0$, które nie zerują się jednocześnie dla $x\neq 0$, tzn. $(f^\prime(x))^2+(g^\prime(x))^2=\frac{1}{(1+x)^2}+1\neq 0$ dla $x\neq 0$.

Dalej 

$$
\lim\limits_{x\rightarrow 0} \frac{f^{\prime}(x)}{g^{\prime}(x)}=\lim\limits_{x\rightarrow 0} \frac{(\ln (1+x))^\prime}{(x)^\prime}=\lim\limits_{x\rightarrow 0} \frac{\frac{1}{1+x}}{1} = 1,
$$

czyli 

c) istnieje skończona granica $\lim\limits_{x\rightarrow 0} \frac{f^{\prime}(x)}{g^{\prime}(x)}=1$ (i jest równa $1$).

Zatem, na mocy pierwszej reguły de l'H$\hat{\text{o}}$spitala dla symbolu nieoznaczonego typu $\frac{0}{0}$ (ozn. $\left[\frac{0}{0}\right]$), otrzymujemy, że 

$$
\lim\limits_{x\rightarrow 0}\frac{\ln (1+x)}{x}=\lim\limits_{x\rightarrow 0}\frac{(\ln (1+x))^\prime}{(x)^\prime}=1.
$$

### Uwaga
W praktyce granicę taką często obliczamy od razu, po sprawdzeniu założeń twierdzenia, pisząc

$$
\lim\limits_{x\rightarrow 0}\frac{\ln (1+x)}{x}\overset{[\frac{0}{0}]}{=}\lim\limits_{x\rightarrow 0}\frac{(\ln (1+x))^\prime}{(x)^\prime}=...=1.
$$




### Przykład 2

Obliczymy 

$$
\lim_{x\to 0}\frac{x-\sin x}{x^3}.
$$

Występuje tutaj symbol nieoznaczony $[\frac{0}{0}]$. Wobec spełnionych założeń Twierdzenia 1, stosujemy regułę de l'H$\hat{\text{o}}$spitala dla symbolu nieoznaczonego $[\frac{0}{0}]$ i otrzymujemy

$$
\lim_{x\to 0}\frac{x-\sin x}{x^3}\overset{[\frac{0}{0}]}{=} \lim_{x\to 0}\frac{(x-\sin x)^\prime}{(x^3)^\prime}=\lim_{x\to 0}\frac{1-\cos x}{3x^2}.
$$

Znowu rozpoznajemy symbol nieoznaczony $[\frac{0}{0}]$. Ponownie stosujemy regułę de l'H$\hat{\text{o}}$spitala dla symbolu nieoznaczonego $[\frac{0}{0}]$ dostając

$$
\lim_{x\to 0}\frac{1-\cos x}{3x^2}\overset{[\frac{0}{0}]}{=}\lim_{x\to 0}\frac{(1-\cos x)^\prime}{(3x^2)^\prime}=\lim_{x\to 0}\frac{\sin x}{6x}=\frac{1}{6},
$$

gdzie ostatnia równość wynika ze wzoru $\lim\limits_{x\to 0}\frac{\sin x}{x}=1$ z teorii granic funkcji.

Zatem 

$$
\lim_{x\to 0}\frac{x-\sin x}{x^3}=\frac{1}{6}.
$$

### Przykład 3

Obliczymy granicę

$$
\lim_{x\to 0}\frac{e^x-e^{-x}-2x}{x-\sin x}.
$$

Rozpoznajemy symbol nieoznaczony $[\frac{0}{0}]$. Mamy 

$$
\begin{eqnarray*}
\lim_{x\to 0}\frac{e^x-e^{-x}-2x}{x-\sin x} &\overset{[\frac{0}{0}]}{=}&\lim_{x\to 0}\frac{(e^x-e^{-x}-2x)'}{(x-\sin x)'}=\lim_{x\to 0}\frac{e^x+e^{-x}-2}{1-\cos x} \overset{[\frac{0}{0}]}{=} \lim_{x\to 0}\frac{(e^x+e^{-x}-2)'}{(1-\cos x)'} \\
&=& \lim_{x\to 0} \frac{e^x-e^{-x}}{\sin x}\overset{[\frac{0}{0}]}{=}\lim_{x\to 0} \frac{e^x+e^{-x}}{\cos x}=\frac{1+1}{1}=2.
\end{eqnarray*}
$$

Zatem $\lim\limits_{x\to 0}\frac{e^x-e^{-x}-2x}{x-\sin x}=2$.

### Uwaga

Zauważmy, że chcąc obliczyć daną granicę, trzeba niekiedy wielokrotnie zastosować regułę de l'H$\hat{\text{o}}$spitala dla tego samego symbolu nieoznaczonego lub dla różnych symboli nieoznaczonych. W przykładzie 3 stosowaliśmy trzykrotnie tę samą regułę dla symbolu nieoznaczonego $[\frac{0}{0}]$.

### Twierdzenie 2
__Drugi przypadek reguły de l'H$\hat{\textbf{o}}$spitala $($wyrażenia typu $\frac{\infty}{\infty})$.__

Jeżeli:

a) funkcje $f(x)$ i $g(x)$ przy $x\rightarrow a$ dążą do nieskończoności, tzn. 

$$
\lim\limits_{x\rightarrow a}f(x)=\lim\limits_{x\rightarrow a}g(x)=\infty,
$$

b) w otoczeniu $U_\ve$ punktu $a$ $($za wyjątkiem, być może, punktu $a$) istnieją pochodne $f'(x)$ i $g'(x)$ oraz $({f'}(x))^2+({g'}(x))^2\neq 0$ dla $x\in U_\ve$ i $x\neq a$,

c) istnieje skończona lub nieskończona granica $\lim\limits_{x\rightarrow a}\frac{f'(x)}{g'(x)}$, 

to 

$$
\lim\limits_{x\rightarrow a}\frac{f(x)}{g(x)}=\lim\limits_{x\rightarrow a}\frac{f'(x)}{g'(x)}.
$$

### Uwagi:
1. Analogiczne twierdzenie dla granic jednostronnych jest również prawdziwe.

2. Dla symbolu nieoznaczonego "$[-\frac{\infty}{\infty}]$" (tzn. typu $-\frac{\infty}{\infty}$) stosujemy również Twierdzenie 2.

### Przykład 

Obliczymy granicę funkcji

$$
\lim_{x\to \infty}\frac{x^3}{e^{2x}}.
$$

Rozpoznajemy symbol nieoznaczony typu $\frac{\infty}{\infty}$ (ozn. $[\frac{\infty}{\infty}]$). Wobec spełnionych założeń a) i b) Twierdzenia 2, stosując trzykrotnie regułę de l'H$\hat{\text{o}}$spitala dla symbolu nieoznaczonego $[\frac{\infty}{\infty}]$, dostajemy

$$
\lim_{x\to \infty}\frac{x^3}{e^{2x}}\overset{[\frac{\infty}{\infty}]}{=}\lim_{x\to \infty}\frac{(x^3)^\prime}{(e^{2x})^\prime}=\lim_{x\to \infty}\frac{3x^2}{2e^{2x}}\overset{[\frac{\infty}{\infty}]}{=}\lim_{x\to \infty}\frac{(3x^2)^\prime}{(2e^{2x})^\prime}=\lim_{x\to \infty}\frac{6x}{4e^{2x}}\overset{[\frac{\infty}{\infty}]}{=}\lim_{x\to \infty}\frac{(6x)^\prime}{(4e^{2x})^\prime}=\lim_{x\to \infty}\frac{6}{8e^{2x}}=\lim_{x\to \infty}\frac{3}{4e^{2x}}=0
$$

czyli $\lim\limits_{x\to \infty}\frac{x^3}{e^{2x}}=0$.

__Trzeci przypadek: wyrażenia nieoznaczone "$[0\cdot \infty]$", czyli typu $0\cdot \infty$.__


>__Definicja.__ Iloczyn $f(x)g(x)$ nazywamy wyrażeniem nieoznaczonym postaci $0\cdot \infty$ w punkcie $x=a$, jeżeli $\lim\limits_{x\tow a}f(x)=0$ i jednocześnie $\lim\limits_{x\tow a}g(x)=\infty$ lub $\lim\limits_{x\tow a}g(x)=-\infty$.

Metoda postępowania -przekształcamy:

$$
f(x)g(x)=\frac{f(x)}{\frac{1}{g(x)}}\hspace{0.5cm}\text{ albo }\hspace{0.5cm}f(x)g(x)=\frac{g(x)}{\frac{1}{f(x)}},
$$

otrzymując symbole nieoznaczone typu $\frac{0}{0}$ albo $\frac{\infty}{\infty}$ odpowiednio.

### Przykład

Obliczymy granicę

$$
\lim_{x\to \frac{\pi}{2}^-}\left(x-\frac{\pi}{2}\right)\tg x.
$$

Zauważmy, że $x-\frac{\pi}{2}\rightarrow 0^-$, gdy $x\to \frac{\pi}{2}^-$ oraz $\tg x\tow \infty$, gdy $x\to \frac{\pi}{2}^-$. Zatem mamy tutaj symbol nieoznaczony $[0\cdot\infty]$.

Po rozpoznaniu typu symbolu nieoznaczonego, przystępujemy do obliczenia granicy funkcji zgodnie z powyżej opisaną metodą. Mamy

$$
\begin{eqnarray*}
\lim_{x\to \frac{\pi}{2}^-}\left(x-\frac{\pi}{2}\right)\tg x &\overset{[0\cdot\infty]}{=}&\lim_{x\to \frac{\pi}{2}^-}\frac{\tg x}{\frac{1}{x-\frac{\Pi}{2}}} =\lim_{x\to \frac{\pi}{2}^-}\frac{\tg x}{(x-\frac{\Pi}{2})^{-1}} \overset{\left[\frac{\infty}{-\infty}\right]}{=}\lim_{x\to \frac{\pi}{2}^-}\frac{(\tg x)'}{((x-\frac{\Pi}{2})^{-1})'}= \lim_{x\to \frac{\pi}{2}^-} \frac{\frac{1}{\cos^2 x}}{-(x-\frac{\Pi}{2})^{-2}\cdot 1} \\
&=& \lim_{x\to \frac{\pi}{2}^-} \frac{-(x-\frac{\Pi}{2})^{2}}{\cos^2 x}\overset{\left[\frac{0}{0}\right]}{=} \lim_{x\to \frac{\pi}{2}^-} \frac{(-(x-\frac{\Pi}{2})^{2})'}{(\cos^2 x)'}=\lim_{x\to \frac{\pi}{2}^-} \frac{-2(x-\frac{\Pi}{2})}{2\cos x(-\sin x)}=\lim_{x\to \frac{\pi}{2}^-} \frac{x-\frac{\Pi}{2}}{\cos x\sin x}\overset{\left[\frac{0}{0}\right]}{=} \\
& =& \lim_{x\to \frac{\pi}{2}^-} \frac{(x-\frac{\Pi}{2})'}{(\cos x\sin x)'}=\lim_{x\to \frac{\pi}{2}^-} \frac{1}{\cos^2 x-\sin^2 x}=\lim_{x\to \frac{\pi}{2}^-} \frac{1}{\cos 2x}=-1.
\end{eqnarray*}
$$

Zatem $\lim\limits_{x\to \frac{\pi}{2}^-}\left(x-\frac{\pi}{2}\right)\tg x=-1$.

**Inne rozwiązanie** (technika podstawienia).

Dokonajmy podstawienia 

$$
y=x-\frac{\Pi}{2}.
$$

Wtedy $x=y+\frac{\Pi}{2}$ oraz, jeżeli $x\tow {\frac{\Pi}{2}}^-$, to $y\tow 0^{-}$. Wobec tego, dokonując przejścia na nową zmienną $y$ oraz pamiętając o wzorze 

$$
\tg\left(\frac{\Pi}{2}+x\right)=-\ctg x
$$ 

dla argumentów $x$ z pierwszej ćwiartki układu współrzędnych, mamy 

$$
\lim_{x\to \frac{\pi}{2}^-}\left(x-\frac{\pi}{2}\right)\tg x = \lim_{y\tow 0^{-}} y\tg \left(y+\frac{\Pi}{2}\right)=\lim_{y\tow 0^{-}} -y\ctg y=\lim_{y\tow 0^{-}} \frac{-y}{\tg y}=\lim_{y\tow 0^{-}} \frac{-1}{\frac{\tg y}{y}}=-1 
$$

ponieważ $\lim\limits_{y\tow 0^{-}}\frac{\tg y}{y}=1 $.

__Czwarty przypadek: wyrażenia nieoznaczone "$[\infty - \infty]$".__

Rozpatrzmy wyrażenie nieoznaczone postaci $f(x)-g(x)$, takie, że gdy $x\tow a$, to $f(x)\tow\infty$ i $g(x)\tow\infty$ albo $f(x)\tow-\infty$ i $g(x)\tow-\infty$.

Metoda postępowania -przekształcamy do symbolu typu $\frac{0}{0}$:

$$
f(x)-g(x)=\frac{f(x)g(x)}{g(x)}-\frac{g(x)f(x)}{f(x)}=\frac{\frac{1}{g(x)}-\frac{1}{f(x)}}{\frac{1}{f(x)g(x)}}.
$$

### Przykład 

Obliczymy granicę

$$
\displaystyle \lim_{x\to 0}\left(\frac{1}{x}-\frac{1}{\sin x}\right).
$$

Rozpoznajemy symbol nieoznaczony $[\infty-\infty]$. Stosując metodę opisaną w czwartym przypadku, dostajemy

$$
\begin{eqnarray*}
\lim_{x\to 0}\left(\frac{1}{x}-\frac{1}{\sin x}\right) &\overset{[\infty-\infty]}{=}& \lim_{x\to 0} \frac{\sin x -x}{x\sin x} \overset{\left[\frac{0}{0}\right]}{=}\lim_{x\to 0} \frac{(\sin x -x)'}{(x\sin x)'}
= \lim_{x\to 0} \frac{\cos x -1}{\sin x+x\cos x}\overset{\left[\frac{0}{0}\right]}{=}\\
&=& \lim_{x\to 0} \frac{(\cos x -1)'}{(\sin x+x\cos x)'}=\lim_{x\to 0} \frac{-\sin x}{\cos x+\cos x-x\sin x}=\frac{0}{2}=0.
\end{eqnarray*}
$$

Zatem $\lim\limits_{x\to 0}\left(\frac{1}{x}-\frac{1}{\sin x}\right)=0$.

### Uwaga

Pojawienie się symbolu nieoznaczonego $[\infty-\infty]$ nie oznacza, że wartość granicy z różnicy dwóch funkcji w tym samym punkcie będzie zawsze równa zeru.

__Piąty przypadek: wyrażenia nieoznaczone "$[0^0]$", "$[\infty^0]$", "$[1^\infty]$".__

Niech dane będą funkcje $f(x)$ i $g(x)$ określone w pewnym sąsiedztwie (ewentualnie jednostronnym) punktu $x=a$, przy czym $f(x)>0$ dla wszystkich $x\neq a$ należących do tego otoczenia. 

Weźmy pod uwagę wyrażenie 

$$
(f(x))^{g(x)}.
$$

$1^0$ Jeżeli przy $x\tow a$  mamy $f(x)\tow 0$, $g(x)\tow 0$, to mówimy, że wyrażenie $(f(x))^{g(x)}$ w punkcie $x=a$ jest wyrażeniem nieoznaczonym typu $0^0$.

$2^0$ Jeżeli przy $x\tow a$  mamy $f(x)\tow \infty$ i jednocześnie $g(x)\tow 0$, to mówimy, że wyrażenie $(f(x))^{g(x)}$ w punkcie $x=a$ jest wyrażeniem nieoznaczonym typu $\infty^0$.

$3^0$ Jeżeli przy $x\tow a$  mamy $f(x)\tow 1$ i jednocześnie $g(x)\tow \infty$ lub $g(x)\tow -\infty$ , to mówimy, że wyrażenie $(f(x))^{g(x)}$ w punkcie $x=a$ jest wyrażeniem nieoznaczonym typu $1^\infty$.


Metoda postępowania -ponieważ $f(x)>0$ dla $x$ z sąsiedztwa punktu $a$, przekształcamy:

$$
(f(x))^{g(x)}=e^{\ln(f(x))^{g(x)}}=e^{g(x)\ln f(x)},
$$

a następnie obliczamy odpowiednią granicę korzystajac z poprzednich reguł.

Inna metoda: 

ponieważ $f(x)>0$ dla $x$ z sąsiedztwa punktu $a$, równanie $y(x)=(f(x))^{g(x)}$ logarytmujemy stronami logarytmem naturalnym $\ln$ dostając:

$$
\ln y(x)=g(x)\ln f(x),
$$

a następnie obliczamy granicę z symbolem nieoznaczonym "$[0\cdot\infty]$". Jeżeli $\lim\limits_{x\tow a}\ln y(x)=A$, to $\lim\limits_{x\tow a}y(x)=e^A$.


### Przykład

Obliczymy granicę 

$$
\displaystyle \lim_{x\to 0^+}x^x.
$$

Rozpoznajemy symbol nieoznaczony $\left[0^{0}\right]$. Stosując metodę opisaną w piątym przypadku, dostajemy

$$
\begin{eqnarray*}
\lim_{x\to 0^+}x^x &\overset{\left[0^{0}\right]}{=}& \lim_{x\to 0^+}e^{\ln {x}^{x}}=\lim_{x\to 0^+} e^{x\ln x}=e^0=1
\end{eqnarray*}
$$

ponieważ

$$
\lim_{x\to 0^+} x\ln x \overset{[0\cdot\infty]}{=}\lim_{x\to 0^+} \frac{\ln x}{\frac{1}{x}}\overset{[-\frac{\infty}{\infty}]}{=}\lim_{x\to 0^+} \frac{(\ln x)'}{(\frac{1}{x})'}=\lim_{x\to 0^+} \frac{\frac{1}{x}}{-\frac{1}{x^2}}=\lim_{x\to 0^+} -x=0.
$$

Zatem $\lim\limits_{x\to 0^+}x^x=1$.

### Uwaga

Przy obliczaniu granicy $\lim\limits_{x\to 0^+} x\ln x$ (symbol nieoznaczony $[0\cdot\infty]$), tzn. granicy z iloczynu funkcji dążącej do zera i funkcji logarytmicznej dążącej do $-\infty$, logarytm zostawiamy w liczniku.

### Przykład 2

Obliczymy granicę 

$$
\displaystyle \lim_{x\to 0}(\cos x)^{\frac{1}{x^2}}.
$$

Rozpoznajemy symbol nieoznaczony $\left[1^{\infty}\right]$. Stosując metodę opisaną w piątym przypadku, dostajemy

$$
\begin{eqnarray*}
\lim_{x\to 0}(\cos x)^{\frac{1}{x^2}}\overset{\left[1^\infty\right]}{=}\lim_{x\to 0} e^{\cos x{^{\frac{1}{x^2}}}}=\lim_{x\to 0} e^{\frac{1}{x^2}\ln\cos x}=e^{-\frac{1}{2}}=\frac{\sqrt{e}}{e}.
\end{eqnarray*}
$$

ponieważ

$$
\lim_{x\to 0} \frac{1}{x^2}\ln\cos x =\lim_{x\to 0} \frac{\ln\cos x}{x^2}\overset{[\frac{0}{0}]}{=}\lim_{x\to 0}\frac{\frac{-\sin x}{\cos x}}{2x}=\lim_{x\to 0} \frac{-\tg x}{2x}=\frac{-1}{2}
$$

z uwagi na wzór $\lim\limits_{x\to 0} \frac{\tg x}{x}=1$.

### Uwaga

Nie należy bezkrytycznie korzystać z reguły de l'H$\hat{\text{o}}$spitala. Dla przykładu granica 

$$
\lim_{x\tow \infty}\frac{x}{2x+\sin x}=\lim_{x\tow \infty}\frac{1}{2+\frac{\sin x}{x}}=\frac{1}{2}
$$

istnieje, lecz granica 

$$
\lim_{x\tow \infty} \frac{(x)'}{(2x+\sin x)'}=\lim_{x\tow \infty} \frac{1}{2+\cos x}
$$

nie istnieje ponieważ nie istnieje granica z funkcji $\cos x$, gdy $x\tow \infty$!

### Przykład 

Rozważmy granicę

$$
\lim_{x\tow 0} \frac{x^2\sin \frac{1}{x}}{\sin x}.
$$

Zauważmy, że w celu obliczenia tej granicy nie możemy zastosować reguły de l'H$\hat{\text{o}}$spitala ponieważ nie jest spełnione założenie o istnieniu granicy ilorazu pochodnych. Rzeczywiście mamy

$$
\lim_{x\tow 0} \frac{\left(x^2\sin \frac{1}{x}\right)'}{(\sin x)'}=\lim_{x\tow 0} \frac{2x\sin \frac{1}{x}-\cos \frac{1}{x}}{\cos x}.
$$

Ta ostatnia granica nie istnieje ponieważ przyjmując $x_n=\frac{1}{2n\Pi}$ otrzymamy, że $x_n\tow 0$ oraz

$$
\frac{2 x_n \sin \frac{1}{x_n}-\cos \frac{1}{x_n}}{\cos x_n}\rightarrow -1,
$$

gdy $n\tow \infty$. Przyjmując natomiast $\hat{x}_n=\frac{1}{\Pi+2n\Pi}$ otrzymamy, że $\hat{x}_n\tow 0$ oraz

$$
\frac{2 \hat{x}_n \sin \frac{1}{\hat{x}_n}-\cos \frac{1}{\hat{x}_n}}{\cos \hat{x}_n}\rightarrow 1,
$$

gdy $n\tow \infty$, czyli nie istnieje granica.

Zauważmy jednak, że można obliczyć tę granicę stosując inną metodę, a mianowicie

$$
\lim_{x\tow 0} \frac{x^2\sin \frac{1}{x}}{\sin x}= \lim_{x\tow 0} \frac{x}{\sin x} \cdot x \sin \frac{1}{x}=1 \cdot 0 = 0.
$$
