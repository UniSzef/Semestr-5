## Twierdzenie o monotoniczności funkcji

>### Twierdzenie
>Niech funkcja $f$ o wartościach rzeczywistych będzie różniczkowalna w każdym punkcie przedziału $(a,b)$. Wówczas:
>
>a) $f$ jest niemalejąca w $(a,b)$ wtedy i tylko wtedy, gdy $f^\prime(x)\geq 0$ w $(a,b)$,
>
>b) $f$ jest nierosnąca w $(a,b)$ wtedy i tylko wtedy, gdy $f^\prime(x)\leq 0$ w $(a,b)$.
>
>Ponadto, jeżeli $f^\prime(x)>0$ w $(a,b)$, to $f$ jest (ściśle) rosnąca w $(a,b)$, a gdy $f^\prime(x)<0$ w $(a,b)$, to $f$ jest (ściśle) malejąca w $(a,b)$.

### Uwagi:
1. Druga część powyższego twierdzenia, podająca związek pomiędzy znakiem pierwszej pochodnej funkcji, a jej ścisłą monotonicznością, jest często stosowana w praktyce.

2. Funkcja rosnącą (odp. malejąca) i różniczkowalna nie musi mieć pochodnej $f^{\prime}(x)$ stale dodatniej (odp. stale ujemnej, tzn. $f^{\prime}(x)<0)$ w całym przedziale $(a,b)$. Dla przykładu, wystaczy rozważyć funkcję $f(x)=x^3$ w przedziale $(-1,1)$. Dla tej funkcji mamy:

$$
f^\prime(x)=3x^2,\;\;x\in(-1,1),
$$

więc $f^\prime(0)=0$ oraz $f^\prime(x)>0$ dla $x\in (-1,0)\cup (0,1)$.

### Przykład
Zbadamy monotoniczność funkcji $f(x)=\frac{1}{3}x^3-5x$. Mamy $D_f=\BR$ oraz

$$
f^\prime(x)=x^2-5, \hspace{0.5cm}D_{f^\prime}=\BR.
$$

Dalej

$$
f^\prime(x)>0\Leftrightarrow x^2-5>0\Leftrightarrow (x-\sqrt{5})(x+\sqrt{5})>0.
$$

Rozwiązaniem tej ostatniej nierówności są $x\in(-\infty,-\sqrt{5})\cup(\sqrt{5},\infty)$. Zatem funkcja $f$ jest rosnąca w $(-\infty,-\sqrt{5})$ oraz w $(\sqrt{5},\infty)$ (faktycznie, skoro np. $x\in (-\infty,-\sqrt{5})$, to $f^\prime(x)>0$, czyli funkcja $f$ jest rosnąca w $(-\infty,-\sqrt{5})$; analogicznie dla drugiego przedziału). Podobnie, skoro rozwiązaniem nierówności:

$$
f^\prime(x)<0\Leftrightarrow (x-\sqrt{5})(x+\sqrt{5})<0
$$

są $x\in(-\sqrt{5},\sqrt{5})$,
więc funkcja $f$ jest malejąca w $(-\sqrt{5},\sqrt{5})$.

Zatem:

- funkcja $f$ jest rosnąca w $(-\infty,-\sqrt{5})$;

- funkcja $f$ jest malejąca w $(-\sqrt{5},\sqrt{5})$;

- funkcja $f$ jest rosnąca w $(\sqrt{5},\infty)$.

### Przykład 2
Niech $f\colon\mathbb{R}\to\mathbb{R}$ będzie funkcją daną wzorem 

$$
f(x)=2x^3-15x^2+36x+1.
$$

Wtedy 

$$
f'(x)=6x^2-30x+36.
$$

Znajdziemy miejsca zerowe funkcji $f'$.
Równanie

$$ 
6x^2-30x+36=0
$$

jest równoważne równaniu

$$
x^2-5x+6=0.
$$

Jego jedynymi rozwiązaniami są liczby 
$x_1=2$ i $x_2=3$.
Łatwo widzimy (bo $f'$ jest funkcją kwadratową), że

- $f'(x)>0$ dla $x\in (-\infty,2)$;

- $f'(x)<0$ dla $x\in (2,3)$;

- $f'(x)>0$ dla $x\in (3,\infty)$.

Zatem:

- funkcja $f$ jest rosnąca w $(-\infty,2)$;

- funkcja $f$ jest malejąca w $(2,3)$;

- funkcja $f$ jest rosnącą w $(3,\infty)$.

## Ekstrema lokalne funkcji

Rozpocznijmy od przedstawienia definicji minimum oraz maksimum lokalnego właściwego funkcji w punkcie ponieważ będziemy z niej korzystać w przykładach i zadaniach.

>### Definicja
>
>Niech funkcja $f$ o wartościach rzeczywistych będzie określona w pewnym otoczeniu punktu $x_0\in\BR$, tj. w pewnym przedziale $(x_0-a;x_0+a)$, gdzie $a>0$. 
>
>1. Mówimy, że funkcja $f$ ma w punkcie $x_0$ **maksimum lokalne**, gdy istnieje taka liczba $\delta>0$, $\delta\leq a$, że jeżeli $\left|x-x_0\right|<\delta$, to $f(x)\leq f(x_0)$. 
>
>2. Mówimy, że funkcja $f$ ma w punkcie $x_0$ **minimum lokalne**, gdy istnieje taka liczba $\delta>0$, $\delta\leq a$, że jeżeli $\left|x-x_0\right|<\delta$, to $f(x)\geq f(x_0)$. 
>
>Maksima lokalne i minima lokalne obejmujemy wspólną nazwą -- ekstremum lokalnego.
>
>3. Mówimy, że funkcja $f$ ma w punkcie $x_0$ **maksimum lokalne właściwe**, gdy istnieje taka liczba $\delta>0$, $\delta\leq a$, że jeżeli $0<\left|x-x_0\right|<\delta$, to $f(x)< f(x_0)$. 
>
>4. Mówimy, że funkcja $f$ ma w punkcie $x_0$ **minimum lokalne właściwe**, gdy istnieje taka liczba $\delta>0$, $\delta\leq a$, że jeżeli $0<\left|x-x_0\right|<\delta$, to $f(x)>f(x_0)$.
>
>W przeciwnym razie mówimy o maksimum (lub odpowiednio minimum) lokalnym niewłaściwym.

### Uwaga 
Nie należy mylić pojęcia minimum (odp. maksimum) lokalnego z pojęciem wartości najmniejszej (odp. największej) funkcji! 

Dla przykładu: 

to, że w punkcie $x_0$ funkcja $f$ przyjmuje wartość największą na zbiorze $A\subset \BR$ oznacza, że dla **wszystkich** $ x\in A$  mamy $f(x)\leq f(x_0)$. To, że $x_0$ jest maksimum lokalnym oznacza, że ta nierówność zachodzi dla tych $x\in A$, które są **blisko** $x_0$ (tzn. w pewnym otoczeniu punktu $x_0$).

### Przykład 3

Funkcja $f(x)=x$ nie posiada ekstremów lokalnych.

Poszukiwanie ekstremów lokalnych funkcji jednej zmiennej $y=f(x)$ przebiega dwuetapowo. 

I. Najpierw szukamy wszystkich "kandydatów" na ekstrema lokalne funkcji f. Są to tzw. punkty krytyczne. Ekstremum może być w tych punktach, ale nie musi.

II. Następnie, w oparciu o odpowiednie narzędzia matematyczne, tzn. jedno z twierdzeń podających warunki dostateczne istnienia ekstremum lokalnego funkcji, weryfikujemy, czy w każdym z osobna punkcie krytycznym jest ekstremum lokalne, czy też go w danym punkcie nie ma.

Gdy poszukujemy kandydatów na ekstrema lokalne funkcji, pomocnym jest następujące twierdzenie. 

## Warunek konieczny istnienia ekstremum funkcji jednej zmiennej

>### Twierdzenie
>Jeżeli funkcja $f$ o wartościach rzeczywistych, określona w pewnym otoczeniu punktu $x_0\in\BR$ i mająca pochodną w punkcie $x_0$ posiada ekstremum lokalne w punkcie $x_0$, to pochodna $f^\prime(x_0)=0$.

### Uwaga

Powyższe twierdzenie umożliwia znalezienie wszystkich tzw. punktów stacjonarnych funkcji $f$ - to punkty, które są miejscami zerowymi pierwszej pochodnej, czyli spełniają warunek $f^\prime(x)=0$. Punkty stacjonarne są najczęściej spotykanymi punktami krytycznymi.

Istnieją jednak przykłady funkcji, które posiadają także inne punkty krytyczne (patrz punkt 2. poniżej). Warto o niech pamiętać, aby rozwiązanie zadania polegającego na znalezieniu ekstremów funkcji było kompletne.

I. Punkty krytyczne, w których mogą (choć nie muszą) być ekstrema lokalne funkcji $f$:

1. punkty stacjonarne,

2. punkty, które należą do dziedziny funkcji $f$, ale nie należą do dziedziny pierwszej pochodnej $f^\prime$ tej funkcji. 

### Przykład 
Funkcja $f(x)=|x|$ ma w punkcie $x=0$ minimum lokalne (a nawet osiąga w tym punkcie swoją wartość najmniejszą)  i nie ma w tym punkcie pochodnej. 

$x=0$ jest tutaj przykładem punktu krytycznego typu 2.

### Uwaga
Jeśli  funkcja $f$ ma w punkcie $x_0$ pochodną i $f'(x_0)=0$ to z powyższego twierdzenia jeszcze **nie wynika**, że $f$ ma tym punkcie ekstremum lokalne. 

Dla przykładu: jeśli $f(x)=x^3$, to  $f'(0)=0$ oraz $f$ nie posiada ekstremów lokalnych (bo jest funkcją rosnącą w $\BR$). 

II. Aby stwierdzić, czy we wskazanym punkcie krytycznym jest ekstremum i jakiego jest ono typu, musimy znać **warunki dostateczne na istnienie ekstremum lokalnego funkcji f**. 

Poniżej zaprezentowano trzy odpowiednie, ważne twierdzenia - narzędzia. 

## Warunek dostateczny istnienia ekstremum lokalnego funkcji przy pomocy pierwszej pochodnej

>### Twierdzenie 
>Niech funkcja $f$ o wartościach rzeczywistych będzie różniczkowalna w pewnym wydrążonym otoczeniu punktu $x_0\in\BR$.
>
>a) Jeżeli przy pewnym $\delta>0$, $f^\prime(x)>0$ dla $x_0-\delta<x<x_0$ i $f^\prime(x)<0$ dla $x_0<x<x_0+\delta$, to funkcja $f$ ma maksimum lokalne właściwe w punkcie $x_0$.
>
>b) Jeżeli przy pewnym $\delta>0$, $f^\prime(x)<0$ dla $x_0-\delta<x<x_0$ i $f^\prime(x)>0$ dla $x_0<x<x_0+\delta$, to funkcja $f$ ma minimum lokalne właściwe w punkcie $x_0$.

### Uwaga
Podane powyżej twierdzenie jest uniwersalne - można je stosować **dla dowolnego**, wyznaczonego **punktu krytycznego**.

### Przykład 

Niech 

$$
f(x)=x^3+3x^2-9x-2.
$$

$D_f=\BR$. Mamy

$$
f'(x)=3x^2+6x-9=3(x^2+2x-3),\hspace{0.5cm}D_{f^{\prime}}=\BR.
$$

Jedynymi miejscami zerowymi $f'$ są liczby $x=-3$ i $x=1$.
Widzimy, że

- $f$ jest rosnąca na $(-\infty,-3)$, bo $f'$ jest tam dodatnia;

- $f$ jest malejąca na $(-3,1)$, bo $f'$ jest tam ujemna;

- $f$ jest rosnąca na $(1,\infty)$, bo $f'$ jest tam dodatnia.

Stąd wynika, że  $f$ ma w punkcie $x=-3$ maksimum lokalne (wynoszące $y_{\max}=f(-3)=25$), a w punkcie $x=1$ minimum lokalne (wynoszące $y_{\min}=f(1)=-7 $).

### Przykład 2

Niech 

$$
f(x)=x^{\frac{1}{3}}(1-x)^\frac{2}{3}.
$$

Zbadamy monotoniczność tej funkcji oraz wyznaczymy jej ekstrema lokalne.

Dziedziną funkcji $f$ są wszystkie liczby rzeczywiste, tzn. $D_f=\BR$. 

Obliczamy 

$$
f^\prime(x)= \left(x^\frac{1}{3}(1-x)^\frac{2}{3}\right)^\prime=\frac{1}{3}x^{-\frac{2}{3}}(1-x)^\frac{2}{3}+x^\frac{1}{3}\left(-\frac{2}{3}(1-x)^{-\frac{1}{3}}\right)=\frac{(1-x)^\frac{2}{3}}{3x^\frac{2}{3}}-\frac{2x^\frac{1}{3}}{3(1-x)^\frac{1}{3}}=\frac{1-x-2x}{3x^\frac{2}{3}(1-x)^\frac{1}{3}}=\frac{1-3x}{3x^\frac{2}{3}(1-x)^\frac{1}{3}}
$$

Wyznaczamy $D_{f^\prime}=\BR\backslash\{0,1\}$, a następnie trzy punkty krytyczne (są one potencjalnymi kandydatami na ekstrema):

- punkty stacjonarne: 

$$
f^{\prime}(x)=0\Leftrightarrow 1-3x=0 \Leftrightarrow x=\frac{1}{3},
$$

czyli $x=\frac{1}{3}$ jest jedynym punktem stacjonarnym.

- punkty: $x=0$ oraz $x=1$ należą do $D_f$, ale nie należą do $D_{f^\prime}$. Są to punkty krytyczne typu 2.

Badamy monotoniczność funkcji. Mamy

$$
f^{\prime}(x)>0\Leftrightarrow \frac{1-3x}{3x^\frac{2}{3}(1-x)^\frac{1}{3}}>0 \Leftrightarrow (1-3x)3x^\frac{2}{3}(1-x)^\frac{1}{3}>0.
$$

Biorąc pod uwagę, że $D_f=\BR$, $D_{f^\prime}=\BR\backslash\{0,1\}$ oraz, że $x=0$ jest pierwiastkiem podwójnym równania $(1-3x)3x^\frac{2}{3}(1-x)^\frac{1}{3}=0$, dostajemy rozwiązanie powyższej nierówności: 

$$
x\in(-\infty,0)\cup \left(0,\frac{1}{3}\right)\cup (1,\infty).
$$


Podobnie, rozwiązaniem nierówności $f^{\prime}(x)<0$ są $x\in \left(\frac{1}{3},1\right)$.

Zatem:

- funkcja $f$ rośnie w przedziale $(-\infty,0)$, rośnie w przedziale $\left(0,\frac{1}{3}\right)$, maleje w przedziale $\left(\frac{1}{3},1\right)$ oraz ponownie rośnie w przedziale $(1,\infty)$.

Biorąc pod uwagę powyższe informacje oraz twierdzenie podające warunek dostateczny istnienia ekstremum lokalnego funkcji przy pomocy pierwszej pochodnej otrzymujemy, że:

- w punkcie krytycznym $x=0$ nie ma ekstremum (ponieważ przy przejściu z lewostronnego otoczenia tego punktu na prawostronne pierwsza pochodna nie zmienia swojego znaku (tzn. $f^\prime(x)>0$ dla $x$ z lewostronnego otoczenia punktu $x=0$ oraz dla $x$ z prawostronnego otoczenia punktu $x=0$),

- w punkcie stacjonarnym $x=\frac{1}{3}$ funkcja $f$ posiada maksimum lokalne właściwe ponieważ przy przejściu z lewostronnego otoczenia tego punktu na prawostronne pierwsza pochodna zmienia swój znak (z "+" na "-"). Wynosi ono 

$$
y_\max=f\left(\frac{1}{3}\right)=\frac{1}{3}\sqrt[3]{4}.
$$

- w punkcie krytycznym $x=1$ funkcja $f$ posiada minimum lokalne właściwe ponieważ przy przejściu z lewostronnego otoczenia tego punktu na prawostronne pierwsza pochodna zmienia swój znak (z "-" na "+"). Wynosi ono 
$$
y_\min=f(1)=0.
$$

Podsumowując: 

- funkcja $f$ rośnie w przedziale $(-\infty,0)$, rośnie w przedziale $\left(0,\frac{1}{3}\right)$, maleje w przedziale $\left(\frac{1}{3},1\right)$ oraz ponownie rośnie w przedziale $(1,\infty)$.

- w $x=\frac{1}{3}$ funkcja posiada maksimum lokalne $y_\max=\frac{1}{3}\sqrt[3]{4}$ i w $x=1$ funkcja $f$ posiada minimum lokalne $y_\min=0$.

### Uwaga

Zauważmy, że funkcja posiada punkt należący do dziedziny funkcji, w którym nie ma ekstremum.

## Pochodne wyższych rzędów

Niech $f\colon (a,b)\to\mathbb{R}$.
Jeśli $f$ jest różniczkowalna w $(a,b)$
to $f'$ jest funkcją określoną na $(a,b)$.
Jej pochodną nazywamy drugą pochodną funkcji $f$ i oznaczamy symbolem $f''$.
W analogiczny sposób definiuje się pochodne wyższych rzędów.

### Uwaga
Do oznaczenia $n$-tej pochodnej funkcji $f$ używamy symbolu $f^{(n)}$.
Dla małych $n$ używa się też notacji:
$f'(x)$, $f''(x)$, $f'''(x)$.
 
### Przykład
Niech 

$$
f(x)=x^3+2x^2+x.
$$

Wtedy:

$$
f'(x)=3x^2+4x,
$$

$$
f''(x)=6x+4,
$$

$$
f'''(x)=6
$$

oraz

$$
f^{(n)}(x)=0\quad\text{dla}\quad n\geq 4.
$$

## Warunek dostateczny istnienia ekstremum lokalnego funkcji przy pomocy drugiej pochodnej

>### Twierdzenie 
>Niech funkcja $f$ o wartościach rzeczywistych ma w punkcie $x_0\in\BR$ skończoną pochodną rzędu drugiego $f^{\prime\prime}(x_0)$. Jeżeli $f^\prime(x_0)=0$ i $f^{\prime\prime}(x_0)\neq 0$, to $f$ ma w punkcie $x_0$ ekstremum lokalne właściwe, przy czym, gdy $f^{\prime\prime}(x_0)<0$, jest to maksimum lokalne, a gdy $f^{\prime\prime}(x_0)>0$, jest to minimum lokalne.

### Uwagi:

1. W przypadku wielomianów łatwo policzyć pochodną drugiego rzędu $f^{\prime\prime}$, więc można stosować powyższe twierdzenie podające  warunek dostateczny istnienia ekstremum lokalnego funkcji przy pomocy drugiej pochodnej, aby szybko wyznaczać ekstrema lokalne zadanej funkcji $f$.

2. Twierdzenia tego **nie wolno** stosować dla punktów krytycznych typu 2., tzn. takich, że nie należą one do $D_{f^\prime}$, choć należą do $D_f$!

### Przykład 
Wyznaczymy ekstrema lokalne funkcji 

$$
f(x)=\frac{1}{3}x^3-5x.
$$

Mamy $D_f=\BR$ oraz

$$
f^\prime(x)=x^2-5, \hspace{0.5cm}D_{f^\prime}=\BR.
$$

Dalej

$$
f^\prime(x)=0\Leftrightarrow x^2-5=0\Leftrightarrow (x-\sqrt{5})(x+\sqrt{5})=0.
$$

Zatem jedynymi punktami krytycznymi tej funkcji są punkty stacjonarne, tzn. miejsca zerowe pierwszej pochodnej. Są to $x=-\sqrt{5}$ i $x=\sqrt{5}$.

Sprawdzimy teraz, czy funkcja $f$ posiada ekstremum lokalne w każdym z tych punktów. Wykorzystamy w tym celu twierdzenie podające warunek dostateczny istnienia ekstremum lokalnego funkcji przy pomocy drugiej pochodnej.

Obliczamy

$$
f^{\prime\prime}(x)=2x, \hspace{0.5cm}D_{f^{\prime\prime}}=\BR,
$$

oraz 

$$
f^{\prime\prime}(-\sqrt{5})=-2\sqrt{5}<0,
$$

i

$$
f^{\prime\prime}(\sqrt{5})=2\sqrt{5}>0.
$$

Zatem funkcja $f$ posiada:

- w punkcie $x=-\sqrt{5}$ **maksimum lokalne** (właściwe) wynoszące $y_\max=f(-\sqrt{5})=-\frac{5}{3}\sqrt{5}+5\sqrt{5}=3\frac{1}{3}\sqrt{5}$, 

- a w punkcie $x=\sqrt{5}$ **minimum lokalne** (właściwe) wynoszące $y_\min=f(\sqrt{5})=\frac{5}{3}\sqrt{5}-5\sqrt{5}=-3\frac{1}{3}\sqrt{5}$.

### Uwaga

Pochodne wielomianów liczy się szybko i łatwo. W tym przypadku twierdzenie podające warunek dostateczny istnienia ekstremum lokalnego funkcji przy pomocy drugiej pochodnej błyskawicznie rozstrzyga, jakiego typu ekstremum jest w danym punkcie stacjonarnym.

Poniższe twierdzenie uogólnia warunek dostateczny istnienia ekstremum lokalnego.

## Warunek dostateczny istnienia ekstremum lokalnego funkcji przy pomocy pochodnych wyższych rzędów

>### Twierdzenie 
>
>Niech funkcja $f$ o wartościach rzeczywistych ma w punkcie $x_0\in\BR$ skończoną pochodną $f^{(n)}(x_0)$ przy pewnych $n>1$ i niech
>
>$$
f^\prime(x_0)=...=f^{(n-1)}(x_0)=0,\hspace{1cm} f^{(n)}(x_0)\neq 0.
$$
>
>a) Gdy $n$ jest liczbą parzystą, to funkcja $f$ ma w punkcie $x_0$ ekstremum lokalne właściwe, przy czym gdy $f^{(n)}(x_0)<0$, jest to maksimum lokalne, a gdy $f^{(n)}(x_0)>0$, jest to minimum lokalne.
>
>b) Gdy $n$ jest liczbą nieparzystą, funkcja $f$ nie ma w punkcie $x_0$ ekstremum lokalnego.

### Przykład

Niech

$$
f(x)=\frac{1}{4}x^4-x^3+\frac{3}{2}x^2-x-2.
$$

Wyznaczymy ekstrema lokalne tej funkcji korzystając z twierdzenia podającego warunek dostateczny istnienia ekstremum lokalnego funkcji przy pomocy pochodnych wyższych rzędów. 

Mamy $D_f=\BR$ oraz

$$
f^\prime(x)=x^3-3x^2+3x-1=(x-1)^3, \hspace{0.5cm} D_{f^\prime}=\BR.
$$

Szukamy punktu krytycznego, który w tym przypadku jest punktem stacjonarnym (i nie ma innych punktów krytycznych!).

$$
f^\prime(x)=0\Leftrightarrow x=1,
$$

czyli w $x=1$ może być ekstremum.

Dalej:

$$
f^{\prime\prime}(x)= 3(x-1)^2, \hspace{0.5cm} D_{f^{\prime\prime}}=\BR,
$$

ale 

$$
f^{\prime\prime}(1)=0.
$$

Nie możemy więc zastosować twierdzenia podającego warunek dostateczny na istnienie ekstremum lokalnego funkcji przy pomocy drugiej pochodnej funkcji $f$. Możemy jednak zastosować twierdzenie ogólniejsze. Mamy

$$
f^{\prime\prime\prime}(x)=6(x-1),\hspace{0.5cm} D_{f^{\prime\prime\prime}}=\BR, \hspace{0.5cm} f^{\prime\prime\prime}(1)=0,
$$

$$
f^{(4)}(x)=6,\hspace{0.5cm} D_{f^{(4)}}=\BR,\hspace{0.5cm} f^{(4)}(1)=6>0,
$$

zatem, na mocy twierdzenia podającego warunek dostateczny istnienia ekstremum lokalnego przy pomocy pochodnych wyższych rzędów wnioskujemy, że w $x=1$ funkcja $f$ posiada minimum lokalne wynoszące $y_\min=f(1)=1\frac{3}{4}$.

### Uwaga

Przy sprawdzaniu, czy w punkcie stacjonarnym $x=1$ jest ekstremum, można było również wykorzystać twierdzenie podające warunek dostateczny istnienia ekstremum lokalnego funkcji przy pomocy pierwszej pochodnej i odpowiedź byłaby taka sama.

## Wartość najmniejsza i największa funkcji

W poniższych przykładach wyjaśnimy jak można wyznaczyć wartość najmniejszą i największą funkcji.

### Uwaga 
Niech $f\colon [a,b]\to\mathbb{R}$ będzie funkcją różniczkowalną. Aby wyznaczyć wartość wartość najmniejszą i największą funkcji $f$ na przedziale $[a,b]$ wystarczy zbadać wartość tej funkcji  na końcach tego przedziału oraz w punktach $x$ spełniających warunek: $f'(x)=0$.  

### Przykład 1
Niech 

$$
f(x)=x^3+3x^2.
$$

Wyznaczymy wartość najmniejszą i największą tej funkcji na przedziale $[-3,1]$.
Mamy

$$
f'(x)=3x^2+6x=3x(x+2).
$$

Jedynymi rozwiązaniami równania $f'(x)=0$ są liczby $x=-2$ i $x=0$.

Ponieważ 
$f(-3)=0$, $f(-2)=4$, $f(0)=0$ i  $f(1)=4$, to największą wartością funkcji $f$ na przedziale $[-3,1]$ jest $4$ a najmniejszą wartością jest $0$. 

### Uwaga 
Może się zdarzyć, że wartość minimalna albo maksymalna jest generowana przez więcej niż jeden argument $x$ funkcji z zadanego przedziału.

### Przykład 2
Niech 

$$
f(x)=e^{-x^4+4x}.
$$

Funkcja ta jest określona dla wszystkich liczb rzeczywistych. Znajdziemy jej wartość największą.

Mamy

$$
f'(x)=(-4x^3+4)e^{-x^4+4x}.
$$

Jedynym miejscem zerowym $f'$ jest $x=1$.
Ponadto $f'(x)>0$ dla $x\in (-\infty,1)$ i $f'(x)<0$ dla $x\in (1,\infty)$.
Zatem $f$ jest rosnąca w $(-\infty,1)$ i malejąca w $(1,\infty)$.
Stąd dla $x=1$ funkcja $f$ przyjmuje wartość największą równą $1$.

Zauważmy teraz, że $f(x)\not=0$ dla $x\in \mathbb{R}$, $\lim\limits_{x\to\infty}f(x)=0$
i $\lim\limits_{x\to-\infty}f(x)=0$.
Zatem $f$ nie przyjmuje wartości najmniejszej (jej zbiór wartości jest równy $(0,1]).$

$$

## Wyznaczanie ekstremów lokalnych przez program komputerowy

Znajdowanie wartości ekstremalnych (lokalnych i globalnych) funkcji jest jednym z najważniejszych zastosowań rachunku różniczkowego. Niezłe pakiety graficzne upościły znajdowanie wartości ekstremalnych. Stosując techniki graficzne (rysowanie wykresów funkcji, zmiana skali i następnie znalezienie współrzędnych odpowiednich punktów) można otrzymać akceptowalne przybliżenia odpowiedzi.

Są jednak co najmniej cztery przyczyny, dla których warto rozumieć podejście analityczne:

a) Rozwiązanie graficzne jest rozwiązaniem przybliżonym. Kiedy wymaga się dokładnego rozwiązania, konieczne jest podejście analityczne.

b) Szukanie wartości ekstremalnych funkcji trzech lub więcej zmiennych zdecydowanie wymaga podejścia analitycznego. Zauważmy, że aby narysować wykres funkcji jednej zmiennej $y=f(x)$ potrzebujemy dwuwymiarowej przestrzeni $\BR^2$; aby narysować wykres funkcji dwóch zmiennych $z=f(x,y)$, gdzie $x,y$ są zmiennymi niezależnymi, potrzebujemy trójwymiarowej przestrzeni $\BR^3$. Jednak **nie da się rysować w czterech i więcej wymiarach!**

c) Rysunek nie dostarcza nam żadnych informacji na temat zachowania się funkcji poza dziedziną rysunku. Należy więc analitycznie wyznaczyć dziedzinę funkcji.

d) Rysunek może nie być dostatecznie jasny. Dla przykładu, poniżej skrypt generujący wykres funkcji $y=\sin \left(\frac{1}{x}\right)$. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate values for x-axis
x = np.linspace(-5, 5, 1000)

# Calculate corresponding values for y-axis
y = np.sin(1 / x)

# Plot the function
plt.plot(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = sin(1/x)')

# Show the plot
plt.show()
```

Po zmianie skali wykres wygląda tak:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate values for x-axis
x = np.linspace(-0.05, 0.05, 1000)

# Calculate corresponding values for y-axis
y = np.sin(1 / x)

# Plot the function
plt.plot(x, y)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = sin(1/x)')

# Show the plot
plt.show()
```

Niestety, nie da wiele powiedzieć o zachowaniu się funkcji $y=\sin \left(\frac{1}{x}\right)$ w pobliżu zera. Widać jedynie, że wartości funkcji oscylują między $-1$ oraz $1$ i należą do przedziału domkniętego $[-1,1]$.

Omówimy teraz metody poszukiwania ekstremów lokalnych funkcji przy pomocy Pythona. Najpierw metoda graficzna.

Rozpocznijmy od przypomnienia wykresu funkcji 

$$
y=x^3+3x^2-9x-2
$$

z przykładu powyżej, a następnie, bazując na wykresie i teorii, zlokalizujemy ekstrema lokalne tej funkcji. 

Poniżej skrypt w Pythonie generujacy wykres tej funkcji.

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt


def fun(x):
    return x**3+3*x**2-9*x-2


plt.axis([-5, 2, -20, 30])
x = np.linspace(-5, 2, 1000) 
vfun = np.vectorize(fun)
y = vfun(x)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Wykres funkcji f(x)=x**3+3*x**2-9*x-2")
plt.grid(True, axis='both')  # włącza siatkę
plt.plot(x, y, '-')
plt.show() 
```

Widać, że funkcja ma w punkcie $x=-3$ maksimum lokalne (wynoszące $y_{\max}=f(-3)=25$), a w punkcie $x=1$ minimum lokalne (wynoszące $y_{\min}=f(1)=-7 $). Wartości $y_\min$ oraz $y_\max$ obliczamy w oparciu o wzór funkcji.

Poniżej skrypt z wykresem pochodnej $f^\prime(x)=3x^2+6x-9$ funkcji $y=x^3+3x^2-9x-2$, by upewnić się co do współrzędnych punktów, w których funkcja $f$ ma ekstremum lokalne.

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt


def fun(x):
    return 3*x**2 + 6*x - 9


plt.axis([-4, 2, -14, 3])
x = np.linspace(-4, 2, 1000) 
vfun = np.vectorize(fun)
y = vfun(x)

# Add the red dashed line at y = 0
plt.axhline(y=0, color='r', linestyle='--')

plt.xlabel("x")
plt.ylabel("y")
plt.title("Wykres funkcji f'(x)=3*x**2 + 6*x - 9")
plt.grid(True, axis='both')  # włącza siatkę
plt.plot(x, y, '-')
plt.show() 
```

A teraz inny, dokładniejszy sposób wyznaczania ekstremów lokalnych tej samej funkcji.

Opierając się na wzorze analitycznym funkcji, **znajdziemy punkty stacjonarne i następnie zbadamy znak drugiej pochodnej dla każdego z nich**. Jest to znana metoda postępowania. Wykorzystamy komputer, aby "szybko" uzyskać interesujące nas dane. Następnie, mając odpowiednią wiedzę z analizy matematycznej, zinterpretujemy te dane.

Poniżej skrypt w Pythonie. Użyjemy funkcji "solve()" (z bilioteki SymPy) zwracającej pierwiastki równania $f^\prime(x)=0$ w formie listy z danymi.

```python
from __future__ import division
from sympy import *

x = Symbol('x')
f = x**3+3*x**2-9*x-2
print("Funkcja: f(x)=", f)
f_prime = f.diff(x)
f_second = f.diff(x, 2)
print("f'(x)=", f_prime)
print("\n Miejsca zerowe f'(x)=", f_prime, " to: \n", solve(f_prime, x))
root1 = solve(f_prime, x)[0]
root2 = solve(f_prime, x)[1]

print("f''(x)=", f_second)
f2 = lambdify(x, f_second)
print("\n")
print("f''(", root1, ")=", f2(root1))
print("f''(", root2, ")=", f2(root2))
fun = lambdify(x, f)
print("\n Wartości funkcji f: \n") 
print("f(", root1, ")=", fun(root1))
print("f(", root2, ")=", fun(root2))
```

Na podstawie powyższych danych widzać, że:

- punktami stacjonarnymi są: $x=-3$ oraz $x=1$,

- funkcja $f$ ma w punkcie $x=-3$ **maksimum lokalne** (właściwe) ponieważ 

$$
f^{\prime\prime}(-3)<0,
$$

- funkcja $f$ ma w punkcie $x=1$ **minimum lokalne** (właściwe) ponieważ 

$$
f^{\prime\prime}(1)>0.
$$

- dodatkowo widzimy wzory na pierwszą i drugą pochodną funkcji $f$ oraz wartości $y_\max$ i $y_\min$ odpowiednio.

Nie zawsze jednak jest to takie proste. Dla przykładu, poniżej skrypt w Pythonie funkcji mającej trzy ekstrema lokalne. 

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt


def fun(x):
    return (x*(x-1)*(x+3))/((x+1)**2 * (x-2)*(x-4))


plt.axis([-10, 10, -6, 6])
x = np.linspace(-10, 10, 1000) 
vfun = np.vectorize(fun)
y = vfun(x)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Wykres funkcji f(x)=(x*(x-1)*(x+3))/((x+1)**2 *(x-2)*(x-4))")
plt.grid(True, axis='both')
plt.plot(x, y, '-')

# Add the yellow dashed line at y = 0
plt.axhline(y=0, color='y', linestyle='--')
plt.show()
```

Powyższa funkcja 

$$
f(x)=\frac{x(x-1)(x+3)}{(x+1)^2 (x-2)(x-4)}
$$

ma trzy ekstrema, ale jedno z nich nie jest od razu widoczne. Zauważmy ponadto, że Python narysował dodatkowo dwie asymptoty pionowe powyższej funkcji $f$ o równaniach $x=2$ oraz $x=4$, ale **nie narysował trzeciej asymptoty pionowej o równaniu $x=-1$**!

Na poniższym wykresie funkcji widać, gdzie (w przybliżeniu) znajduje się ostatnie ekstremum lokalne tej funkcji.

```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt


def fun(x):
    return (x*(x-1)*(x+3))/((x+1)**2 * (x-2)*(x-4))


plt.axis([-10, -2, -0.1, 0.1])
x = np.linspace(-10, -2, 1000) 
vfun = np.vectorize(fun)
y = vfun(x)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Wykres funkcji f(x)=(x*(x-1)*(x+3))/((x+1)**2 *(x-2)*(x-4))")
plt.grid(True, axis='both')
plt.plot(x, y, '-', label='y=f(x)')

# Add the black solid line at y = 0
plt.axhline(y=0, color='black', linestyle='-')
plt.plot(y, 'k-', label='y=0')

# Add the red dashed line line at y = -0.06
plt.axhline(y=-0.06, color='r', linestyle='--')
plt.plot(y, 'r--', label='y=-0.06')
plt.legend()
plt.show() 
```

Przejdźmy teraz do napisania skryptu w Pythonie obliczającego miejsca zerowe tej funkcji:

$$
f(x)=\frac{x(x-1)(x+3)}{(x+1)^2 (x-2)(x-4)}.
$$


```python
from __future__ import division
from sympy import *


def fun(x):
    return (x*(x-1)*(x+3))/((x+1)**2 * (x-2)*(x-4))


x = Symbol('x')
f2 = (x*(x-1)*(x+3))/((x+1)**2 * (x-2)*(x-4))
df2 = f2.diff(x)

print("Pochodna funkcji: f(x)=", (x*(x-1)*(x+3))/((x+1)**2 * (x-2)*(x-4)), "\n to: \n", df2)
print("\n Rozwiązanie równania f'(x)=0: \n", solve(df2, x))
R2 = solve(df2, x)
print(R2, "\n\n Rozwiązania numeryczne: \n")
print([N(solution) for solution in R2])
```

Przy próbie obliczenia pierwiastków równania $f^\prime(x)=0$ pojawił się problem. Python co prawda pokazał "formułę" na pochodną funkcji, ale nie znalazł pierwiastków równania $f^\prime(x)=0$.

Wygenerował "CRootOf(...)" w zwróconej liście (rozwiązań).

Wyjaśnienie z internetu znaczenia "CRootOf" w kontekście wygenerowania takiego komunikatu przez Pythona:

"CRootOf is a way of symbolically representing roots of polynomials whose roots can't be represented by radicals. If you are only interested in numeric solutions, you can use N on each of the terms as suggested."

Dopiero użycie metody numerycznej do znalezienia rozwiązania równania $f^\prime(x)=0$ zakończyło się sukcesem, co widać na końcu. Python znalazł 3 rozwiązania rzeczywiste i dwa rozwiązania zespolone (są to, de facto, liczby zespolone sprzężone).

Alternatywnie (patrz skrypt poniżej) można użyć metody bisekcji, aby znaleźć przybliżone miejsca zerowe równania $f^\prime(x)=0$. Zanim jednak tej metody użyjemy, musimy najpierw sporządzić wykres funkcji $f$, a następnie znaleźć przedział argumentów tej funkcji, w którym istnieje tylko jedno ekstremum lokalne funkcji.

Czas dokończyć analizę. Wykonajmy test znaku drugiej pochodnej w każdym rzeczywistym punkcie stacjonarnym (ponieważ wykres funkcji rozważamy na płaszczyźnie rzeczywistej).

```python
from __future__ import division
from sympy import *
import scipy.optimize as optimize


def fun(x):
    return (x*(x-1)*(x+53))/((x+1)**2 * (x-2)*(x-4))


x = Symbol('x')
f = (x*(x-1)*(x+3))/((x+1)**2 * (x-2)*(x-4))
df2 = f.diff(x)

f_prime = f.diff(x)
f_second = f.diff(x, 2)
f = lambdify(x, f)
f_p = lambdify(x, f_prime)

R2 = solve(df2, x)
print("\n Funkcja: f(x)=", (x*(x-1)*(x+3))/((x+1)**2 * (x-2)*(x-4)))
print("\n Rozwiązania numeryczne równania f'(x)=0: \n")
print([N(solution) for solution in R2])

r1 = optimize.bisect(f_p, -10, -3)
r2 = optimize.bisect(f_p, 0.25, 0.75)
r3 = optimize.bisect(f_p, 2.25, 3.75)
root1 = round(optimize.bisect(f_p, -10, -3), 3)
root2 = round(optimize.bisect(f_p, 0.25, 0.75), 3)
root3 = round(optimize.bisect(f_p, 2.25, 3.75), 3)
print("\n Wykorzystując algorytm bisekcji można również obliczyć przybliżone miejsca zerowe równania f'(x)=0: \n")
print("Przybliżone miejsce zerowe f'(x):", root1)
print("Przybliżone miejsce zerowe f'(x):", root2)
print("Przybliżone miejsce zerowe f'(x):", root3)
print("\n Sprawdzenie: \nf'(", root1, ")=", f_p(r1), "=", round(f_p(r1), 10), "\n f'(", root2, ")=", f_p(r2), "=", round(f_p(r2), 10), "\n f'(", root3, ")=", f_p(r3), "=", round(f_p(r3), 10))

print("\n f''(x)=", f_second)
f2 = lambdify(x, f_second)
print("\n")
print("f''(", root1, ")=", f2(r1), "=", round(f2(r1), 3))
print("f''(", root2, ")=", f2(r2), "=", round(f2(r2), 3))
print("f''(", root3, ")=", f2(r3), "=", round(f2(r3), 3))
print(' \n Wartości funkcji f w punktach stacjonarnych: \n')
print("f(", root1, ")=", f(r1), "=", round(f(r1), 3))
print("f(", root2, ")=", f(r2), "=", round(f(r2), 3))
print("f(", root3, ")=", f(r3), "=", round(f(r3), 3))
```

Na podstawie powyższych danych widzać, że:

- rzeczywistymi punktami stacjonarnymi funkcji $f$ (w zaokrągleniu do trzech miejsc po przecinku) są $x=-6.398$, $x=0.486$ oraz $x=2.73$;

- funkcja $f$ ma w punkcie $x=-6.398$ **minimum lokalne** (właściwe) ponieważ 

$$
f^{\prime\prime}(-6.398)>0.
$$

Wynosi ono $y_\min=-0.063$;

- funkcja $f$ ma w punkcie $x=0.486$ **minimum lokalne** (właściwe) ponieważ 

$$
f^{\prime\prime}(0.486)>0
$$

Wynosi ono $y_\min=-0.074$;

- funkcja $f$ ma w punkcie $x=2.73$ **maksimum lokalne** (właściwe) ponieważ 

$$
f^{\prime\prime}(2.73)<0.
$$

Wynosi ono $y_\max=-2.098$.

- Dodatkowo widzimy wzór na drugą pochodną funkcji $f$ oraz sprawdzenie, że wyznaczone, **przybliżone punkty stacjonarne** są obliczone z akceptowalną dokładnością.
