## Zbieżność bezwzględna oraz względna szeregów

> __Definicja.__ Mówimy, że szereg  $\sum\limits_{n=1}^\infty a_n$ jest **bezwzględnie zbieżny**, jeśli zbieżny jest szereg  $\sum\limits_{n=1}^\infty |a_n|$. 
>
>Mówimy, że szereg  $\sum\limits_{n=1}^\infty a_n$ jest **warunkowo zbieżny**, jeśli szereg $\sum\limits_{n=1}^\infty a_n$ jest zbieżny, ale nie jest on zbieżny bezwzględnie.


### Przykłady:

1. Przykładem szeregu bezwzględnie zbieżnego jest szereg $\sum\limits_{n=1}^\infty(-1)^{n+1} \frac{1}{n^3}$ ponieważ zbieżnym jest szereg $\sum\limits_{n=1}^\infty \frac{1}{n^3}$.

2. Przykładem szeregu zbieżnego warunkowo jest $\sum\limits_{n=1}^\infty(-1)^{n+1} \frac{1}{n}$.

> __Fakt.__ Jeśli szereg $\sum\limits_{n=1}^\infty a_n$ jest zbieżny bezwzględnie, to jest on zbieżny. Wtedy też $|\sum\limits_{n=1}^\infty a_n|\leq \sum\limits_{n=1}^\infty |a_n|$. 

> __Twierdzenie.__ Szereg  $\sum\limits_{n=1}^\infty a_n$ jest bezwzględnie zbieżny wtedy i tylko wtedy, gdy szereg $\sum\limits_{n=1}^\infty |a_n|$ jest ograniczony.

### Zadanie.

Korzystając z kryterium Leibniza, uzasadnij zbieżność następujących szeregów:

a) $\displaystyle\sum_{n=1}^\infty(-1)^{n+1}\frac{1}{n^3}$

b) $\displaystyle\sum_{n=1}^\infty(-1)^{n+1}\frac{\sqrt{n}}{n+1}$.

Czy szeregi te są zbieżne bezwzględnie?

$\color{green}{Rozwiązanie:}$

a) Mamy: 

$1^0$ $a_n=\frac{1}{n^3}>0$ dla każdego $n\in\BN$.

$2^0$ Zauważmy, że wraz ze wzrostem $n$ wyrażenie $n^3$ rośnie, czyli $\frac{1}{n^3}$ maleje, a zatem ciąg $(\frac{1}{n^3})$  jest malejący.

$3^0$ $\lim\limits_{n\rightarrow\infty} \frac{1}{n^3}=0$.

Na mocy kryterium Leibniza szereg $\displaystyle\sum_{n=1}^\infty(-1)^{n+1}\frac{1}{n^3}$ jest zbieżny.

Zbieżność bezwzględna.

Zauważmy, że szereg $\displaystyle\sum_{n=1}^\infty|(-1)^{n+1}\frac{1}{n^3}|=\sum_{n=1}^\infty\frac{1}{n^3}$ jest zbieżnym szeregiem Dirichleta (z $\alpha=3>1$; formalnie uzasadniamy to w oparciu o kryterium Cauch'ego o zagęszczaniu), a więc szereg $\displaystyle\sum_{n=1}^\infty(-1)^{n+1}\frac{1}{n^3}$ jest zbieżny bezwzględnie.

$$
\color{blue}{\text{Odp. Szereg jest bezwzględnie zbieżny.}}
$$

b) Mamy: 

$1^0$ $a_n=\frac{\sqrt{n}}{n+1}>0$ dla każdego $n\in\BN$.

$2^0$ Sprawdzamy, czy ciąg $(\frac{\sqrt{n}}{n+1})$ jest malejący, tzn. czy $a_{n+1}< a_n$ dla dowolnego $n\in\BN$. Równoważnie: czy  $a_{n+1}- a_n<0$ dla dowolnego $n\in\BN$. Mamy:

$$
\frac{\sqrt{n+1}}{n+2}<\frac{\sqrt{n}}{n+1}\Leftrightarrow \frac{n+1}{(n+2)^2}<\frac{n}{(n+1)^2}\Leftrightarrow (n+1)(n^2+2n+1)<n(n^2+4n+4)\Leftrightarrow
$$

$$
\Leftrightarrow n^3+3n^2+3n+1<n^3+4n^2+4n\Leftrightarrow1<n^2+n.
$$

Ostatnia nierówność jest prawdziwa dla każdego $n\in\BN$, więc prawdziwa jest nierówność $\frac{\sqrt{n+1}}{n+2}<\frac{\sqrt{n}}{n+1}$, czyli ciąg $(\frac{\sqrt{n}}{n+1})$ jest malejący.

$3^0$ Zachodzi $\lim\limits_{n\rightarrow\infty} \frac{\sqrt{n}}{n+1}=\lim\limits_{n\rightarrow\infty} \frac{\frac{1}{\sqrt{n}}}{1+\frac{1}{n}}=0$.

Na mocy kryterium Leibniza szereg $\displaystyle\sum_{n=1}^\infty(-1)^{n+1}\frac{\sqrt{n}}{n+1}$ jest zbieżny.

Sprawdzimy, czy jest on bezwzględnie zbieżny, tzn. czy zbieżny jest szereg

$$
\displaystyle\sum_{n=1}^\infty|(-1)^{n+1}\frac{\sqrt{n}}{n+1}|=\displaystyle\sum_{n=1}^\infty\frac{\sqrt{n}}{n+1}.
$$

Mamy:

$$
\frac{\sqrt{n}}{n+1}\geq \frac{\sqrt{n}}{n+n}=\frac{\sqrt{n}}{2n}=\frac{1}{2\sqrt{n}}=\frac{1}{2n^{1/2}}
$$

dla każdego $n\in\BN$. Oczywiście $\frac{\sqrt{n}}{n+1}>0$. Ponieważ szereg $\sum_{n=1}^\infty \frac{1}{2n^{1/2}}$ jest rozbieżny, na mocy kryterium porównawczego zbieżności szeregów wnosimy, że szereg 

$$
\displaystyle\sum_{n=1}^\infty\frac{\sqrt{n}}{n+1}\text{   jest rozbieżny}.
$$

W konsekwencji, szereg $\displaystyle\sum_{n=1}^\infty(-1)^{n+1}\frac{\sqrt{n}}{n+1}$ nie jest bezwzględnie zbieżny (ale jest warunkowo zbieżny).

$$
\color{blue}{\text{Odp. Szereg nie jest bezwzględnie zbieżny. Jest jednak zbieżny.}}
$$

## Promień i obszar zbieżności szeregu potęgowego

### Przykład 1

Obliczymy promień i koło zbieżności rzeczywistego szeregu potęgowego

$$
\sum_{n=1}^\infty n!(x-5)^n, \hspace{0.5cm} x\in \BR.
$$

Czy szereg ten jest zbieżny dla $x=5$?

Mamy $a_n=n!$ dla każdego $n\in\BN$ oraz $x_0=5$. Na mocy Twierdzenia Cauchy'ego - Hadamarda (patrz wykład), obliczamy

$$
\displaystyle\lambda:=\overline { \lim_{n\to\infty}}\root n\of {\mid a_n\mid}=\limsup_{n\to\infty} \sqrt[n]{n!}=\lim_{n\to\infty} \sqrt[n]{n!}=\infty,
$$

wobec czego przyjmujemy, że promień zbieżności tego szeregu $r$ wynosi $0$ (patrz Twierdzenie Cauchy'ego - Hadamarda).  

Koło zbieżności (czyli zbiór $(5-r,5+r)$ dla $r=0$) jest zbiorem pustym. 

Następnie zauważmy, że dla $x=5$ szereg $\sum\limits_{n=1}^\infty n!(x-5)^n, x\in \BR$, jest zbieżny ponieważ to, de facto, szereg $\sum\limits_{n=1}^\infty 0$ o sumie $0$.

Zatem, pomimo, że w tym przypadku koło zbieżności szeregu potęgowego jest zbiorem pustym, dodatkowa analiza pozwala stwierdzić, że szereg ten jest zbieżny dla $x=5\in\BR$.

### Przykład 2

Obliczymy promień i koło zbieżności rzeczywistego szeregu potęgowego

$$
\sum\limits_{n=1}^\infty \frac{(2n)!}{8^n(n!)^2}(x+7)^n, \hspace{0.5cm} x\in \BR.
$$

Mamy $a_n=\frac{(2n)!}{8^n(n!)^2}>0$ dla wszystkich $n\in\BN$ oraz $x_0=-7$.

Na mocy faktu: 

<p style="text-align: center"><b>
jeżeli istnieje granica $\lim_{n\to\infty}\left|\frac{a_{n+1}}{a_n}\right|$, to granica $\lim_{n\to\infty}\sqrt[n]{|a_n|}$ też istnieje i wynosi tyle samo.
    </b></p>  

wystarczy obliczyć 

$$
\lim_{n\to\infty}\left|\frac{a_{n+1}}{a_n}\right|. 
$$

Ponieważ 

$$
(2(n+1))!=(2n+2)!=(2n)!(2n+1)(2n+2),
$$

otrzymujemy

$$
\lim_{n\to\infty}\left|\frac{a_{n+1}}{a_n}\right|= \lim_{n\to\infty}\left|\frac{\frac{(2(n+1))!}{8\cdot 8^n((n+1)!)^2}}{\frac{(2n)!}{8^n(n!)^2}}\right|=\lim_{n\to\infty}\frac{\frac{(2n)!(2n+1)(2n+2)}{8\cdot 8^n(n!)^2(n+1)^2}}{\frac{(2n)!}{8^n(n!)^2}}=\lim_{n\to\infty}\frac{(2n+1)(2n+2)}{8(n+1)^2}=\lim_{n\to\infty}\frac{2n+1}{4(n+1)}=\frac{1}{2}.
$$ 

Zatem $\lim\limits_{n\to\infty}\sqrt[n]{|a_n|}=\frac{1}{2}$ i w konsekwencji

$$
\displaystyle\lambda:=\overline { \lim_{n\to\infty}}\root n\of {\mid a_n\mid}=\lim_{n\to\infty} \sqrt[n]{\frac{(2n)!}{8^n(n!)^2}}=\frac{1}{2}.
$$

Na mocy Twierdzenia Cauchy'ego - Hadamarda, promień zbieżności szeregu potęgowego definiowany jako

$$
r:=\frac{1}{\lambda},
$$

wynosi w przypadku naszego szeregu $r=2$. Natomiast koło zbieżności rzeczywistego szeregu potęgowego $\sum\limits_{n=1}^\infty \frac{(2n)!}{8^n(n!)^2}(x+7)^n$, definiowane jako

$$
\{x\in\BR\colon |x+7|<2\}=\{x\in\BR\colon -2<x+7<2\},
$$

to przedział:

$$
(-7-2,-7+2)=(-9,-5).
$$

### Przykład 3

Dla zespolonego szeregu potęgowego postaci

$$
\sum_{n=1}^\infty \frac{\left(\frac{3}{2}\right)^n(1+(-1)^n)^n}{n}(z-2)^n,
$$

gdzie $z\in\mathbb{C}$, mamy 

$$
a_n=\frac{\left(\frac{3}{2}\right)^n(1+(-1)^n)^n}{n}, \hspace{0.5cm} n\in\BN\text{  oraz  }z_0=2\hspace{0.2cm} (=2+0\cdot i).
$$

Na mocy Twierdzenia Cauchy'ego - Hadamarda (patrz wykład)

$$
\displaystyle\lambda:=\overline { \lim_{n\to\infty}}\root n\of {\mid a_n\mid}=\limsup_{n\to\infty} \sqrt[n]{\frac{\left(\frac{3}{2}\right)^n(1+(-1)^n)^n}{n}}=\limsup_{n\to\infty} \frac{\frac{3}{2}(1+(-1)^n)}{\sqrt[n]{n}}=3,
$$

więc promień $r$ koła zbieżności tego szeregu, obliczony na mocy Twierdzenia Cauchy'ego - Hadamarda, wynosi:

$$
r:=\frac{1}{\lambda}=\frac{1}{3}.
$$

Obszar (inaczej koło) zbieżności tego szeregu zespolonego to zbiór:

$$
\{z\in\BC\colon |z-2|<\frac{1}{3}\},
$$

czyli wnętrze koła zespolonego w $\mathbb{C}$ o środku w punkcie 

$$
z_0=2=2+0\cdot i
$$ 

i promieniu zbieżności równym $r=\frac{1}{3}$. 

### Uwagi:

1. Promień zbieżności $r$ jest zawsze liczbą rzeczywistą nieujemną bądź symbolem $\infty$.

2. Rzeczywisty szereg potęgowy $\sum_{n=1}^\infty a_n(x-x_0)^n$ jest zbieżny dla każdego $x$ należącego do obszaru zbieżności $(x_0-r,x_0+r)$ oraz rozbieżny dla wszystkich $x$, które należą do zbioru $(-\infty,x_0-r)\cup (x_0+r,\infty)$. Natomiast na brzegu obszaru zbieżności tego szeregu możliwe są różne sytuacje, które rozpatrujemy i rozstrzygamy osobno. Analogicznie dla zespolonych szeregów potęgowych $\sum_{n=1}^\infty z_n(z-z_0)^n$, gdzie $z_0,z_n\in \mathbb{C}$. Dla $z$ z zespolonego koła zbieżności szereg jest zbieżny, dla $z$ położonych na zewnątrz tego koła zbieżności szereg jest rozbieżny, a na brzegu zespolonego koła zbieżności możliwe są różne sytuacje.    

3. Szereg $\sum\limits_{n=1}^\infty \frac{\left(\frac{3}{2}\right)^n(1+(-1)^n)^n}{n}(z-2)^n$ jest bezwzględnie zbieżny na dowolnym zwartym podzbiorze koła zbieżności $\{z\in\BC\colon |z-2|<\frac{1}{3}\}$, tzn. w dowolnym kole:

$$
\{z\in\BC\colon |z-2|\leq\frac{1}{3}-\varepsilon\},
$$

gdzie $\varepsilon\in(0,\frac{1}{3})$.


Wizualizacja Pythona obszaru zbieżności zespolonego szeregu potęgowego z przykładu 3 (uwaga! brzeg koła zespolonego, czyli zbiór $\{z\in\BC\colon |z-2|=\frac{1}{3}\}$, nie należy do rozwiązania!):

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_complex_inequality():
    x = np.linspace(1, 3, 400)
    y = np.linspace(-1, 1, 400)
    X, Y = np.meshgrid(x, y)
    Z = X + 1j * Y
    inequality = np.abs(Z - 2) < (1/3)
    plt.figure(figsize=(8, 6))
    plt.contourf(X, Y, inequality)
    plt.xlabel('Real part')
    plt.ylabel('Imaginary part')
    plt.title('Complex Solution of |Z-2| < 1/3')
    plt.colorbar()
    plt.show()

plot_complex_inequality()
```

### Przykład
Wyliczymy promień zbieżności rzeczywistego szeregu potęgowego postaci:

$$
x+\frac{x^3}{3!}+\frac{x^5}{5!}+\cdots,
$$

czyli szeregu

$$
\sum_{k=1}^\infty \frac{x^{2k-1}}{(2k-1)!}.
$$

Oznaczmy
\begin{equation*}
a_n=
\begin{cases}
0,&\text{gdy $n$ jest parzyste,}\\
\frac{1}{n!},&\text{gdy $n$ jest nieparzyste.}
\end{cases}
\end{equation*}
    
Wyliczymy $\limsup\limits_{n\to\infty}\sqrt[n]{|a_n|}$.  

Niech $b_n=\frac{1}{n!}$ dla $n\in\mathbb{N}$; wówczas $0\le a_n\le b_n$ dla $n\in\mathbb{N}$, więc $0\le\sqrt[n]{|a_n|}\le\sqrt[n]{b_n}$ dla $n\in\mathbb{N}$.  

Dalej, $\lim\limits_{n\to\infty}\frac{b_{n+1}}{b_n}=0$, więc $\lim\limits_{n\to\infty}\sqrt[n]{b_n}=0$.  Na mocy twierdzenia o trzech ciągach otrzymujemy $\limsup\limits_{n\to\infty}\sqrt[n]{|a_n|}=0$, więc $\lambda =0$, $r=\infty$ i dany szereg jest zbieżny dla każdego $x\in\mathbb{R}$.

### Przykład

Wyznaczymy zbiór tych $x\in\mathbb{R}$, dla których szereg 

$$
\sum_{n=1}^\infty n^3(x-7)^n
$$

jest zbieżny. 

Ponieważ $a_n=n^{3}$ dla $n\in\BN$ oraz $x_0=7$, na mocy Twierdzenia Cauchy'ego - Hadamarda

$$
\displaystyle\lambda:=\overline { \lim_{n\to\infty}}\root n\of {\mid a_n\mid}=\limsup_{n\to\infty} \sqrt[n]{n^{3}}=\limsup_{n\to\infty} \left(\sqrt[n]{n}\right)^3=1,
$$

więc promień koła zbieżności tego szeregu wynosi $r:=\frac{1}{\lambda}=1.$

Obszar zbieżności tego szeregu rzeczywistego to zbiór:

$$
\{x\in\BR\colon |x-7|<1\},
$$

czyli przedział $(6,8)$. 

Sprawdźmy jeszcze, czy szereg potęgowy jest zbieżny dla $x=6$ oraz $x=8$. Łatwo zauważyć, że obu tych przypadkach otrzymujemy szeregi liczbowe rozbieżne (wyrazy ogólne tych szeregów liczbowych nie dążą do zera), odpowiednio:

$$
\sum_{n=1}^\infty (-1)^n n^3
$$

oraz 

$$
\sum_{n=1}^\infty n^3.
$$

Zatem szereg potęgowy $\sum\limits_{n=1}^\infty n^3(x-7)^n$ jest zbieżny dla dowolnego $x\in(6,8)$.

## Dodatek

Pokażemy przykład rozwiązania równania różniczkowego zwyczajnego pierwszego rzędu postaci:

$$
y^{\hspace{0.1cm}\prime}=f(x,y)
$$

z warunkiem początkowym

$$
y(x_0)=y_0,
$$

przy pomocy techniki polegającej na przybliżaniu rozwiązania $y$ tego równania różniczkowego szeregiem potęgowym postaci:

$$
y(x) = \sum_{n=0}^\infty a_n(x-x_0)^n.
$$


### Przykład

Rozważmy szczególną postać równania różniczkowego Riccatiego z warunkiem początkowym:

$$
y^{\hspace{0.1cm}\prime}=y^2+x^2, \hspace{0.6cm}y(0)=1.
$$

Dokonajmy przybliżenia funkcji $y$ zmiennej $x$ szeregiem potęgowym:

$$
y(x) = \sum_{n=0}^\infty a_n x^n = a_0 + a_1x + a_2x^2 + a_3x^3 + \cdots \hspace{0.1cm}.
$$

Zauważmy, że warunek początkowy $y(0)=1$ natychmiast daje, że $a_0=1$. 

Obliczamy

$$
(y(x))^2 = (1+a_1x+a_2 x^2 + a_3 x^3+\cdots)^2 = 1 + 2a_1 x + (a_1^2+2a_2)x^2 + (2a_3+2a_2a_1)x^3 + \cdots\hspace{0.1cm}.
$$

Podstawiając do równania różniczkowego dostajemy

$$
a_1+ 2a_2x +3 a_3x^2 + 4a_4x^3 + \cdots  = 1 + 2a_1 x + (a_1^2+2a_2)x^2 + (2a_3+2a_2a_1)x^3 + \cdots + x^2
$$

czyli

$$
a_1+ 2a_2x +3 a_3x^2 + 4a_4x^3 + \cdots  = 1 + 2a_1 x + (1+ a_1^2+2a_2)x^2 + (2a_3+2a_2a_1)x^3 + \cdots \hspace{0.1cm}.
$$

Obliczamy teraz poszczególne współczynniki $a_i$ dla $i=2,3,4,...$, otrzymując:

\begin{array}{ll}
a_1 = 1, & \\
2a_2=2a_1, & a_2=1 \\
3a_3=1+a_1^2+2a_2, & a_3= \frac{4}{3}, \\
4a_4 = 2a_3+2a_2a_1, & a_4 = \frac{7}{6}, \hspace{0.3cm}...\hspace{0.1cm}.
\end{array}

Stąd, dla małych $x$ otrzymaliśmy dobrą aproksymację funkcji $y$, a mianowicie

$$
y(x)=1+x+x^2+\frac{4}{3}x^3+\frac{7}{6}x^4+\mathcal{O}(x^5).
$$
