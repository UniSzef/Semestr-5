## Punkty skupienia zbioru

>__Definicja.__ Element $x_0\in X$ nazywa się punktem skupienia zbioru $A\subset X$ jeżeli w każdej 
kuli otwartej $K(x_0,r)$ $\ (r>0)$ istnieje co najmniej 1 element zbioru $A$ różny od $x_0$:
>
>$$ 
 {\forall}_{r>0} \ {\exists}_{x\in A, \ x\neq x_0} \ \ x\in A\cap K(x_0,r)
$$ 
 
Inaczej mówiąc - $x_0$ jest punktem skupienia zbioru $A$ jeżeli istnieje ciąg $(x_n)\subset A$ taki, 
że $ (x_n)$ jest zbieżny do $x_0$ gdy  ${n\to\infty}$ i $x_n\not= x_0$ dla każdego $n$.


### Przykłady

>* Punktami  skupienia zbioru $A=(0,1)$ są wszystkie liczby należące do  $[0,1]$. Podobnie punktami skupienia zbioru $[0,1]$   są wszystkie liczby należące do  $[0,1]$.

>* Punktami skupienia zbioru $A=[0,1]\cup \{2\}$ są wszyskie liczby należące do $[0,1]$.

>* Zbiór $A=\mathbb{N}$ nie posiada punktów skupienia.

>* Jedynym punktem skupienia zbioru $A=\{\frac{1}{n}:n\in\mathbb{N}\}$ jest $x=0$.

## Granica funkcji w punkcie

Niech $f : X \to {\mathbb{R}}$ i niech $x_0$ niech będzie punktem skupienia zbioru $X$.

>__Definicja.__ (def. Heinego)
>
>Mówimy, że element $y_0$ jest granicą funkcji $f$  w punkcie $x_0$, jeżeli 
dla dowolnego ciągu $(x_n)$ elementów $x_n\in X$,  $x_n\neq x_0$ oraz $x_n \overset{n\to\infty}\to x_0$ 
odpowiedni ciąg $(f(x_n))$ jest zbieżny do $y_0$, czyli $f(x_n) \overset {n\to\infty}\to y_0$.

>__Definicja.__ (def. Cauchy'ego)
>
>Mówimy, że element $y_0$ jest granicą funkcji $f$ w punkcie $x_0$, jeżeli dla dowolnego
$\varepsilon>0$ istnieje taka liczba $\delta >0$, że dla wszystkich elementów $x \not= x_0\in X$ takich, że $|x - x_0| < \delta$ zachodzi $| f(x) - y_0 | < \varepsilon$ 

>Jeżeli $y_0$ jest  granicą funkcji $f$ w punkcie $x_0$, to piszemy
>
>$$
y_0=\lim_{x\to x_0} f(x).
$$

>__Twierdzenie.__  Definicje Heinego i Cauchy'ego granicy funkcji $f$ 
w punkcie $x_0$ dla funkcji _o wartościach rzeczywistych_ są
równoważne tj. element $y_0$ jest granicą funkcji $f$ w sensie definicji Heinego wtedy
i tylko wtedy, gdy jest granicą funkcji $f$  w sensie definicji Cauchy'ego.

### Przykłady

1. Korzystając z definicji (w obydwu formach) pokażemy, że granica funkcji $f\colon\mathbb{R}\to\mathbb{R}$ danej wzorem $f(x)=2x+4$ 
w punkcie $x=3$ jest równa 10.

>* (def. Heinego) Niech $(x_n)$ będzie ciągiem elementów zbioru $\mathbb{R}\setminus\{3\}$ zbieżnym do $3$.
Wtedy 
>
>$$
f(x_n)=2x_n+4\xrightarrow{n\to\infty}10.
$$
>
>Zatem $\displaystyle \lim_{x\to 3}(2x+4)=10.$

>* (def. Cauchy'ego) 
Niech $\varepsilon>0$.
Mamy
>
>$$
|f(x)-10|=|2x+4-10|=2|x-3|.
$$
>
>Zatem jeśli 
>
>$$
|x-3|<\delta:=\frac{\varepsilon}{2}, 
$$
>
>to 
>$$
|f(x)-10|<\varepsilon.
$$
>
> Zatem $\displaystyle \lim_{x\to 3}(2x+4)=10.$

2. Pokażemy, że funkcja $f:(0,\infty)\to\mathbb{R}$ dana wzorem $f(x)=\sin(1/x)$ nie ma granicy w $x_0=0$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Create an array of x values
x = np.linspace(0.001, 1, 10000)

# Calculate y values
y = np.sin(1 / x)

# Plot the function
plt.plot(x, y)
plt.title(r'$y = \sin\left(\frac{1}{x}\right)$')
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1.5, 1.5)  # Set the y-axis limits
plt.grid(True)

# Add more visible axes
plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=2)
plt.show()
```

Niech $(x_n)$ będzie ciągiem danym wzorem 

$$
x_n=\frac{1}{n\pi}.
$$

Oczywiście jest to ciąg zbieżny do $x_0=0$. Ponadto

$$
f(x_n)=\sin(n\pi)=0\xrightarrow{n\to\infty} 0.
$$

Niech $(x'_n)$ będzie ciągiem danym wzorem 

$$
x'_n=\frac{1}{\frac{\pi}{2}+2n\pi}.
$$

Oczywiście jest to ciąg zbieżny do $x_0=0$. Ponadto

$$
f(x'_n)=\sin\left(\frac{\pi}{2}+2n\pi\right)=1\xrightarrow{n\to\infty} 1.
$$

Stąd funkcja $f$ nie ma granicy w $x_0=0$.

3. Pokażemy, że funkcja $f\colon \mathbb{R}\setminus \{0\}\to\mathbb{R}$ dana wzorem $f(x)=\frac{|x|}{x}$
nie ma granicy w $x_0=0$.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x_positive = np.linspace(0.01, 5, 100)  # Avoid x=0
x_negative = np.linspace(-5, -0.01, 100)  # Avoid x=0

# Define the corresponding y values
y_positive = 1
y_negative = -1

# Create the plot
plt.plot(x_positive, y_positive * np.ones(len(x_positive)), label='f(x) = 1 for x > 0')
plt.plot(x_negative, y_negative * np.ones(len(x_negative)), label='f(x) = -1 for x < 0')

# Add more visible axes
plt.axhline(0, color='black', linewidth=2)
plt.axvline(0, color='black', linewidth=2)

# Add labels and legend
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of f(x) = |x|/x')
plt.legend()

# Show the plot
plt.grid()
plt.show()

```

To, że granica funkcja $f$ nie ma granicy w $x_0=0$ wynika z obserwacji:

$$
f(1/n)=1\xrightarrow{n\to\infty}1\quad \text{i} f(-1/n)=-1\xrightarrow{n\to\infty}-1. 
$$

## Arytmetyczne własności granicy

>__Twierdzenie.__ (granica sumy, iloczynu, różnicy i ilorazu)
>     
>Jeżeli funkcje $f$ i $g$ mają granice
w punkcie $x_0$, to funkcje $f+g$, $f-g$ oraz $f\cdot g$ mają też granice w tym punkcie i odpowiednio:
>     
>$$
\lim_{x\to x_0}(f(x)+g(x))=\lim_{x\to x_0}f(x)+\lim_{x\to x_0} g(x)
$$
>     
>$$
\lim_{x\to x_0}(f(x)-g(x))=\lim_{x\to x_0}f(x)-\lim_{x\to x_0} g(x)
$$
>     
>$$
\lim_{x\to x_0}(f(x)\cdot g(x))=\lim_{x\to x_0}f(x)\cdot\lim_{x\to x_0}g(x) 
$$ 
>     
W szczególności dla $f(x)=c=\text {const}$.
>     
>$$
\lim_{x\to x_0}(c\cdot g(x))=c\cdot \lim_{x\to x_0}g(x) .
$$
>
>Ponadto jeśli funkcja $g$ nie ma miejsc zerowych na otoczenu punktu $x_0$ i $\lim_{x\to x_0}g(x)\not=0$, to 
>
>$$
\lim_{x\to x_0}\frac{f(x)}{g(x)}=\frac{\displaystyle\lim_{x\to x_0}f(x)}{\displaystyle\lim_{x\to x_0}g(x)}.
$$ 

### Przykłady

1. $\displaystyle\lim_{x\to 3}\frac{x^2-1}{x-1} = \frac{9-1}{3-1}=\frac{8}{2}=4.$

2. $\displaystyle\lim_{x\to 1}\frac{x^3-1}{x-1} =  \lim_{x\to 1}\frac{(x-1)(x^2+x+1)}{x-1}=\lim_{x\to 1}(x^2+x+1)=1+1+1=3$.

3. $\displaystyle\lim_{x\to 2 }\frac{x^2-5x+6}{x^2-12x+20} = \lim_{x\to 2 }\frac{ (x-3)(x-2)}{(x-2)(x-10)}=\lim_{x\to 2 }\frac{x-3}{x-10}=\frac{-1}{-8}=\frac{1}{8}$.

4.  $\displaystyle \lim_{x\to 0}x\sin\left(\frac{114x+1}{x}\right)=0$, bo $\lim\limits_{x\to 0}x=0$, a $|\sin\left(\frac{114x+1}{x}\right)|\leq 1$ dla każdego $x\in\mathbb{R}$.

### Przykłady do samodzielnego rozwiązania

Oblicz następujące granice:

1. $\displaystyle\lim_{x\to 2}\frac{x-1}{x^2+2}$

2. $\displaystyle\lim_{x\to 2}\frac{3x^2-5x-2}{5x^2-20}$

3. $\displaystyle\lim_{x\to 1}\frac{x^3-1}{x-1}$

4. $\displaystyle\lim_{x\to 2}\frac{x-1}{x^2+2}$

5. $\displaystyle\lim_{x\to 25}\frac{\sqrt{x}-5}{x-25}$


### Ważna granica

$$
\lim_{x\to 0}\frac{\sin x}{x}=1.
$$

### Przykłady

1. $\displaystyle\lim_{x\to 0} \frac{x}{\sin 3x} = \lim_{x\to 0} \frac{1}{\frac{\sin 3x}{x}}=\lim_{x\to 0} \frac{1}{\frac{3\sin 3x}{3x}}=\frac{1}{3}$.

2. $\displaystyle\lim_{x\to 0}\frac{\sin 2x}{\sin 3x} = \lim_{x\to 0}\left(\frac{\sin 2x}{2x}\cdot \frac{2}{3}\cdot \frac{3x}{\sin 3x}\right)=\lim_{x\to 0}\left(\frac{\sin 2x}{2x}\cdot \frac{2}{3}\cdot \frac{1}{\frac{\sin 3x}{3x}}\right)=\frac{2}{3}$.

3. $\displaystyle \lim_{x\to 0} \frac{\sin 5x-\sin 3x}{\sin x} = \lim_{x\to 0} \left(\frac{\sin 5x}{\sin x}-\frac{\sin 3x}{\sin x}\right)
= \lim_{x\to 0} \left(\frac{\sin 5x}{5x}\cdot\frac{5}{\frac{\sin x}{x}}-\frac{\sin 3x}{3x}\cdot \frac{3}{\frac{\sin x}{x} }\right)=5-3=2$.

## Granice jednostronne.

Dla funkcji rzeczywistych $f \ : \ P\to \Bbb R$, gdzie $P$ jest przedziałem można
pojęcie granicy nieco uogólnić.

>__Definicja.__ Niech $x_0$ będzie punktem skupienia przedziału $ P$. Mówimy, że
liczba $g\in \Bbb R$ jest granicą prawostronną funkcji $f$ w punkcie $x_0$ jeżeli: 
>
>(a)  (def. Heinego)
>
>dla dowolnego ciągu $(x_n)$,  $x_n\in P$,  $x_n>x_0$  $x_n\underset {n\to\infty}\longrightarrow x_0$  ciąg $(f(x_n))$ 
jest zbieżny do $g$
>
>
>(b) (def. Cauchy'ego)
>     
>$$
{\forall}_{\varepsilon>0} \  {\exists}_{\delta>0} \ {\forall}_{x\in P} \ \ 0<x-x_0<\delta \ \ \Longrightarrow \ \ 
|f(x)-g | <\varepsilon
$$
>
>Ten fakt oznaczać będziemy:  $g=\lim_{x\to x_0+}f(x)$ (w części literatury: $g=f(x_0+0)$ lub $g=f(x_0+))$.


>Analogicznie definiujemy granicę lewostronną $h$: w (a) bierzemy ciągi $(x_n)$
takie, że $x_n<x_0$, a w (b) $x\in P$ spełniające warunek $0<x_0-x<\delta$. 
>
>Oznaczać ją będziemy $h= \lim_{x\to x_0-}f(x)$ (lub $h=f(x_0-0)=f(x_0-))$. Granice także nazywać będziemy łącznie
__jednostronnymi.__

>__Twierdzenie.__  Jeżeli $\lim_{x\to x_0+}f(x) = \lim_{x\to x_0-}f(x)$, to
istnieje granica funkcji $f$ w punkcie $x_0$ i równa jest wartości tych granic jednostronnych.


>W n i o s e k. Jeżeli $\lim_{x\to x_0+}f(x)\neq \lim_{x\to x_0-}f(x)$, to funkcja $f$ nie posiada granicy w punkcie $x_0$.

### Przykłady

1. Zbadamy granicę funkcji $f\colon\mathbb{R}\to\mathbb{R}$
danej wzorem

$$
f(x)=\begin{cases}2x+2,x\geq 0\\
2x-4, x<0\\
\end{cases}
$$
w punkcie $x_0=0$.


Mamy

$$
\lim_{x\to 0^-}f(x)= \lim_{x\to 0^-}(2x-4)=-4
$$

oraz

$$
\lim_{x\to 0^+}f(x)= \lim_{x\to 0^+}(2x+2)=2.
$$

Zatem $f$ nie ma granicy w $x=0$.

2. Mamy 


$$
\lim_{x\to 0^-}e^\frac{1}{x}=0
$$

i 

$$
\lim_{x\to 0^+}e^\frac{1}{x}=\infty.
$$

3. Mamy 

$$
\lim_{x\to \frac{\pi}{2}^-}\operatorname{tg}x= +\infty.
$$
 
i 

$$
\lim_{x\to \frac{\pi}{2}^+}\operatorname{tg}x= -\infty.
$$

## Granica w $+\infty$ i $-\infty$

Granicę funkcji w $+\infty$ i $-\infty$ analogicznie jak definicję granicy w punkcie (patrz wykład).

Mamy

1. $\displaystyle\lim_{x\to -\infty}x^2  =\color{blue}{\infty}$.
2.  $\displaystyle\lim_{x\to \infty}\frac{x+1}{x-1}  =\lim_{x\to \infty}\frac{1+\frac{1}{x}}{1-\frac{1}{x}}=\color{blue}{1}$, bo $\lim\limits_{x\to \infty}\frac{1}{x}=0$.
3. $\displaystyle\lim_{x\to -\infty}\frac{x^2+x}{3x^2+1} = \lim_{x\to -\infty}\frac{1+\frac{1}{x}}{3+\frac{1}{x^2}}=\color{blue}{\frac{1}{3}}$, bo $\lim\limits_{x\to -\infty}\frac{1}{x^2}=0$.
4. $\displaystyle\lim_{x\to \infty}(\sqrt{x+1}-\sqrt{x})  =\lim_{x\to \infty}\frac{(\sqrt{x+1}-\sqrt{x})(\sqrt{x+1}+\sqrt{x})}{\sqrt{x+1}+\sqrt{x}} =\lim_{x\to \infty}\frac{1}{\sqrt{x+1}+\sqrt{x}}=\color{blue}{0}$.
5. $\displaystyle\lim_{x\to \infty}\frac{x^3+x}{5x^3+2x}  =\lim_{x\to \infty}\frac{1+\frac{1}{x^2}}{5+\frac{2}{x^2}} =\color{blue}{\frac{1}{5}}$.
6. $\displaystyle\lim_{x\to \infty}\frac{\sin x}{x} = \lim_{x\to \infty}\frac{1}{x}\cdot \sin x =\color{blue}{0}$ 
ponieważ $\lim\limits_{x\to \infty}\frac{1}{x}=0$ oraz $|\sin x|\leq 1$ dla wszystkich $x\in\mathbb{R}$.
7. $\displaystyle\lim_{x\to \infty}\frac{2^x+3\cdot 5^x}{5^x-3^x}=\lim_{x\to \infty}\frac{\frac{2^x}{5^x}+3}{1-\frac{3^x}{5^x}}=\lim_{x\to \infty}\frac{\left(\frac{2}{5}\right)^x+3}{1-\left(\frac{3}{5}\right)^x}=\color{blue}{3}$ 
ponieważ
$\lim\limits_{x\to \infty}\left(\frac{2}{5}\right)^x=0$ oraz $\lim\limits_{x\to \infty}\left(\frac{3}{5}\right)^x=0$.
