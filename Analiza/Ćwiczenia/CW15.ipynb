# Całki niewłaściwe

## Cel zajęć

Celem tych zajęć jest badanie zbieżności całek niewłaściwych i omówienie ich  zastosowań w badaniu zbieżności szeregów.

## Obliczanie całek niewłaściwych

### Przykład 1

Obliczymy całkę

$$
\int_0^1\frac{1}{2\sqrt{x}}dx.
$$

Zauważmy, że jedynym punktem osobliwym funkcji $f(x)=\frac{1}{2\sqrt{x}}$ w przedziale $[0,1]$ jest punkt $0$.
Zatem musimy policzyć 

$$
\lim_{t\to 0^+}\int_{t}^1\frac{1}{2\sqrt{x}}dx.
$$
Mamy

$$
\begin{align*}
\lim_{t\to 0^+}\int_t^1\frac{1}{2\sqrt{x}}dx
=\lim_{t\to 0^+}\left(\sqrt{x}\biggr|_t^1\right)
=\lim_{t\to 0^+}\left(1-\sqrt{t}\right)=1.
\end{align*}
$$

Zatem całka 

$$
\int_0^1\frac{1}{2\sqrt{x}}dx
$$
jest zbieżna i 

$$
\int_0^1\frac{1}{2\sqrt{x}}dx=1.
$$


### Przykład 2 

Zbadamy zbieżność całki

$$
\int_0^1\frac{1}{x^2}dx.
$$

Zauważmy, że jedynym punktem osobliwym funkcji $f(x)=\frac{1}{x^2}$ w przedziale $[0,1]$ jest punkt $0$.
Zatem musimy policzyć 

$$
\lim_{t\to 0^+}\int_{t}^1\frac{1}{x^2}dx.
$$

Mamy

$$
\begin{align*}
\lim_{t\to 0^+}\int_t^1\frac{1}{x^2}dx
=\lim_{t\to 0^+} \left(-\frac{1}{x}\biggr|_t^1\right)
=\lim_{t\to 0^+}\left(-1+\frac{1}{t}\right)=\infty.
\end{align*}
$$

Zatem całka 

$$
\int_0^1\frac{1}{x^2}dx
$$

nie jest zbieżna.

### Przykład 3

Obliczymy całkę

$$
\int_{-\infty}^\infty\frac{1}{1+x^2}dx.
$$

Z definicji (zarówno $-\infty$ jak i $\infty$ są punktami osobliwymi, musimy podzielić przedział całkowania tak, by w każdym przedziale był tylko jeden punkt osobliwy) ta całka jest zbieżna wtedy i tylko wtedy, gdy zbieżne są całki

$$
\int_{-\infty}^0\frac{1}{1+x^2}dx\quad
\text{i}\quad \int_0^\infty\frac{1}{1+x^2}dx.
$$

Mamy

$$
\begin{align*}
\int_{-\infty}^0\frac{1}{1+x^2}dx=\lim_{t\to -\infty}\int_{t}^0\frac{1}{1+x^2}dx=\lim_{t\to-\infty}\left(\operatorname{arctg}x\biggr|_t^0\right)=
\lim_{t\to-\infty}-\operatorname{arctg}t=\frac{\pi}{2}.
\end{align*}
$$

Podobnie

$$
\begin{align*}
\int_{0}^\infty\frac{1}{1+x^2}dx=\lim_{t\to \infty}\int_{0}^t\frac{1}{1+x^2}dx=\lim_{t\to\infty}\left(\operatorname{arctg}x\biggr|_0^t\right)=
\lim_{t\to\infty}\operatorname{arctg}t=\frac{\pi}{2}.
\end{align*}
$$

Zatem całka

$$
\int_{-\infty}^\infty\frac{1}{1+x^2}dx
$$

jest zbieżna i 

$$
\int_{-\infty}^\infty\frac{1}{1+x^2}dx=
\int_{-\infty}^0\frac{1}{1+x^2}dx+\int_0^\infty\frac{1}{1+x^2}dx=\pi.
$$

### Przykład 4

Obliczymy całkę

$$
\int_{-\infty}^0xe^xdx.
$$

Zauważmy najpierw, że (proste całkowanie przez części) 

$$
\int xe^xdx=xe^x-e^x+C.
$$

Zatem 

$$
\begin{align*}
\lim_{t\to-\infty} \int_t^0xe^xdx=\lim_{t\to-\infty} 
\left(xe^x-e^x\biggr|_t^0\right)
=\lim_{t\to-\infty} \left(-1-te^t+e^t\right)=-1.
\end{align*} 
$$

Stąd

$$
\int_{-\infty}^0xe^xdx=-1.
$$

### Przykład 5

Obliczymy całkę

$$
\int_0^\infty 2xe^{-x^2}dx.
$$

Zauważmy najpierw,  że (proste całkowanie przez podstawienie)

$$
\int2xe^{-x^2}dx=-e^{-x^2}+C.
$$

Zatem

$$
\lim_{t\to\infty}\int_0^{t}2xe^{-x^2}dx=
\lim_{t\to \infty} -e^{x^2}\biggr|_0^t=
\lim_{t\to\infty} (1-e^{-t^2})=1.
$$

Zatem

$$
\int_0^\infty 2xe^{-x^2}dx=1.
$$

### Przykład 6

Zbadamy teraz całkę

$$
\int_{1}^\infty\frac{2}{x^2-1}dx.
$$

Całka ta jest zbieżna ($1$ i $\infty$ są punktami osobliwymi) wtedy i tylko wtedy, gdy zbieżne są całki 

$$
\int_{1}^2\frac{2}{x^2-1}dx\text{ i } 
\int_{2}^\infty\frac{2}{x^2-1}dx.
$$

Stosując rozkład na ułamki proste otrzymujemy, że 

$$
\frac{2}{x^2-1}=\frac{1}{x-1}-\frac{1}{x+1}.
$$

Stąd

$$
\begin{align*}
\int_{2}^\infty\frac{2}{x^2-1}dx=&
\lim_{t\to\infty}\int_2^t \frac{2}{x^2-1}dx\\
=&\lim_{t\to\infty}\left(\operatorname{ln}|1-x|-\operatorname{ln}|1+x|\biggr|_2^t\right)\\
=&\lim_{t\to\infty}\left(\operatorname{ln}|1-t|-\operatorname{ln}|1+t|-\operatorname{ln}1+\operatorname{ln}3\right)\\
=&\lim_{t\to\infty} \operatorname{ln}\frac{|1-t|}{|1+t|}+\operatorname{ln 3}=\operatorname{ln} 3. 
\end{align*}
$$

Ponadto

$$
\begin{align*}
\int_{1}^2\frac{2}{x^2-1}dx=&
\lim_{t\to 1^+}\int_t^2 \frac{2}{x^2-1}dx\\
=&\lim_{t\to 1^+}\left(\operatorname{ln}|1-x|-\operatorname{ln}|1+x|\biggr|_t^2\right)\\
=&\lim_{t\to 1^+}\left(-\operatorname{ln}3-\operatorname{ln}\frac{|1-t|}{|1+t|}\right)=\infty
. 
\end{align*}
$$

Stąd

$$
\int_{1}^\infty\frac{2}{x^2-1}dx
$$

nie jest zbieżna.

## Uzasadnianie zbieżności całek

Interesowała będzie nas teraz nie dokładna wartość całki niewłaściwej ale tylko to czy jest ona zbieżna. Pomocne nam będą następujące wyniki.

>__Fakt__
>Jeśli całka 
>
>$$
\int_a^b|f(x)|dx
$$
>
>jest zbieżna, to całka
>
>$$
\int_a^bf(x)dx
$$
też jest zbieżna.

>__Fakt__
>
>Niech $b$ będzie jedynym punktem osobliwym funkcji $f$ i $F$ w przedziale $[a,b]$. Jeżeli dla $a\leq x<b$ mamy $|f(x)|\leq F(x)$, 
to  ze zbieżności całki  $\int_a^bF(x)dx$ wynika zbieżność całki
$\int_a^bf(x)dx$. 

### Przykład 7

Uzasadnimy, że zbieżna jest całka

$$
\int_{-\infty}^\infty e^{-x^2}dx.
$$

Można zauważyć, że istnieje stała $C>0$ taka, że 
dla $x\in\mathbb{R}$ mamy 

$$
e^{-x^2}\leq \frac{C}{x^2+1}.
$$

Pokazaliśmy już, że całka

$$
\int_{-\infty}^\infty \frac{1}{x^2+1}dx
$$
jest zbieżna. 

Zatem nasza wyjściowa całka również jest zbieżna.

### Przykład 8

Zbadamy zbieżność całki 

$$
\int_0^\infty e^{-x}\sin xdx.
$$

W tym celu zauważmy, że 
dla wszystkich $x\in\mathbb{R}$ mamy, że 

$$
|e^{-x}\sin x|\leq e^{-x}.
$$

Ponieważ całka

$$
\int_0^\infty e^{-x}dx
$$

jest zbieżna, to całka

$$
\int_0^\infty e^{-x}\sin xdx
$$

też jest.

### Warunkowa zbieżność całki niewłaściwej (materiał dodatkowy)

Jeśli całka 
$\int_a^bf(x)dx$ jest zbieżna, ale całka $\int_a^b|f(x)|dx$
nie jest zbieżna, to mówimy, że całka $\int_a^bf(x)dx$ jest zbieżna warunkowo.
>__Twierdzenie}__ Niech punkt $b$ będzie jedynym punktem osobliwym
iloczynu funkcji $fg$ w przedziale $[a,b]$. Jeżeli funkcja 
$\displaystyle F(x)=\int_a^xf(x)dx$ istnieje i jest ograniczona na przedziale $[a,b)$, a funkcja $g(x)$ dąży monotonicznie do zera, gdy $x$ dąży do $b$, to całka  
>
>
>$$
\int_a^bf(x)g(x)dx
$$
>
>jest zbieżna.

#### Przykład 9 

Pokażemy, że całka 

$$
\int_1^\infty\frac{\cos x}{x}dx
$$

jest warunkowo zbieżna.

Najpierw pokażemy, że całka 

$$
\int_1^\infty\frac{\cos x}{x}dx
$$

jest zbieżna.
W tym celu zastosujemy powyższe twierdzenie.
Zauważmy najpierw, że dla każdego $x\geq 1$ mamy
$$
\begin{align*}
\left|\int_1^x\cos x dx\right|=\left|\sin x\biggr|_1^x\right|=
\left|\sin x-\sin 1\right|\leq 2.
\end{align*}
$$

Ponadto funkcja $\frac{1}{x}$ monotonicznie dąży do zera, gdy $x$ dąży do $\infty$. Zatem na mocy powyższego twierdzenia całka 

$$
\int_1^\infty\frac{\cos x}{x}dx
$$

jest zbieżna. Pokażemy teraz, że nie jest ona bezwzględnie zbieżna.
W tym celu musimy pokazać, że
nie jest zbieżna całka


$$
\int_1^\infty\left|\frac{\cos x}{x}\right|dx.
$$

W tym celu zauważmy, że dla każdego $n\in\mathbb{N}$ jeśli 

$$
x\in [-\pi/3+2n\pi, \pi/3+2n\pi],\text{ to }\cos x\geq \frac{1}{2}.
$$
Zatem

$$
\begin{align*}
\int_1^\infty\left|\frac{\cos x}{x}\right|dx&\geq 
\sum_{n=1}^\infty\int_{-\pi/3+2n\pi}^{\pi/3+2n\pi}\left|\frac{\cos x}{x}\right|dx
\\
&\geq \frac{1}{2}\sum_{n=1}^\infty\int_{-\pi/3+2n\pi}^{\pi/3+2n\pi}\frac{1}{x}dx\\
&\geq \frac{1}{2}\sum_{n=1}^\infty\int_{-\pi/3+2n\pi}^{\pi/3+2n\pi}\frac{1}{\pi/3+2n\pi}dx\\
&=\frac{1}{2}\cdot\frac{2\pi}{3}\sum_{n=1}^\infty\frac{1}{\pi/3+2n\pi}=\infty.
\end{align*}
$$

## Całkowe kryterium zbieżności szeregów

>__Twierdzenie__ Niech $f$ będzie funkcją dodatnią i nierosnącą na zbiorze $[1,\infty)$ oraz niech $f(n)=a_n$ dla $n\in\mathbb{N}$.
Wtedy szereg $\displaystyle \sum_{n=1}^\infty a_n$ jest zbieżny wtedy i tylko wtedy, gdy całka $\displaystyle\int_1^\infty f(x)dx$ jest zbieżna.

### Przykład 10
Zbadamy zbieżność szeregu 

$$
\sum_{n=1}^\infty \frac{1}{n^s}.
$$

Gdy $s\leq0$, to wyraz ogólny tego szeregu nie dąży do zera, zatem szereg ten nie jest zbieżny. Załóżmy więc teraz, że $s>0$.
Funkcja $f$ dana wzorem

$$
f(x)=\frac{1}{x^s}
$$
jest malejąca na zbiorze $[1,\infty)$.
Całka

$$
\int_1^\infty \frac{1}{x^s}
$$

jest zbieżna tylko wtedy, gdy $s>1$.
Zatem  szereg

$$
\sum_{n=1}^\infty \frac{1}{n^s}
$$

jest zbieżny tylko dla $s>1$.

### Przykład do samodzielnego rozwiązania

Korzystając z całkowego kryterium zbieżności szeregów pokazać, że szereg 

$$
\sum_{n=1}^\infty \frac{1}{n^2+1}
$$

jest zbieżny.
