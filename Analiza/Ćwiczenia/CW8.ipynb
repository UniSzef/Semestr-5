## Cel zajęć

Wprowadzenie definicji pochodnej funkcji jednej zmiennej w punkcie, jej interpretacji geometrycznej i fizycznej oraz przedstawienie różnych sposobów obliczania pochodnych funkcji, w tym twierdzenia o pochodnej funkcji złożonej.

## Wstęp

W szkole średniej uczyliśmy się, że średnią prędkość z jaką porusza się punkt materialny
oblicza się ze wzoru

$$
\frac{\text{przebyta droga}}{\text{czas}}.
$$

Postawmy teraz przed sobą  ambitne zadanie precyzyjnego zdefiniowania pojęcia prędkości chwilowej, to znaczy  wyjaśnienia co mamy na myśli mówiąc, że w danej chwili punkt materialny porusza się z daną prędkością.

Niech zatem funkcja $f\colon[a,b]\to\mathbb{R}$ opisuje drogę przebytą przez punkt materialny:
to znaczy  niech dla $x\in[a,b]$ liczba $f(x)$ będzie równa drodze przebytej przez punkt materialny do chwili $x$. Naszym celem jest zdefiniowanie prędkości naszego punktu w chwili $x_0\in[a,b]$.
W tym celu zauważmy, że dla dowolnego $h>0$
liczba 

$$
f(x_0+h)-f(x_0)
$$

jest równa drodze przebytej przez punkt materialny w czasie od $x_0$ do $x_0+h$ (czas, który upłynął‚ jest równy $h$).
Stąd liczba 

$$
\frac{f(x_0+h)-f(x_0)}{h}
$$

jest równa średniej prędkości punku materialnego w czasie od $x_0$ do $x_0+h$. Intuicja  podpowiada nam, że dla bardzo małych liczb $h$ średnia prędkość punku materialnego w czasie od $x_0$ do $x_0+h$ powinna być w przybliżeniu równa prędkości naszego punktu w chwili $x_0$ (w tej chwili nadal nie mamy precyzyjnej definicji tego pojęcia).

Z powyższych rozważań, wynika (zwłaszcza dla czytelnika  obytego z teorią granic), że  naturalną definicją prędkości w chwili $x_0$ jest wartość granicy (o ile ona istnieje)

$$
\lim_{h\to 0}\frac{f(x_0+h)-f(x_0)}{h}.
$$

Powyższe rozumowanie daje też dobrą intuicję dlaczego pochodne są tak bardzo pomocne przy badaniu szeroko pojętego przebiegu zmienności funkcji (bo prędkość dobrze opisuje zmianę przebytej drogi).

## Definicja pochodnej funkcji w punkcie

> __Definicja.__ Niech $f\colon (a,b)\to\mathbb{R}$ i niech $x_0\in (a,b)$. Granicę ilorazu różnicowego 
>
>$$
\lim\limits_{h\tow 0}\frac{f(x_0+h)-f(x_0)}{h}
$$
>
>nazywamy **pochodną funkcji $f$ w punkcie $x_0$** i oznaczamy $f^\prime(x_0)$.
>
>Funkcję $f$ nazywamy **różniczkowalną w punkcie $x_0$**, jeżeli istnieje pochodna funkcji w tym punkcie.
>
>Funkcję $f$ nazywamy **różniczkowalną**, gdy jest ona różniczkowalna w dowolnym punkcie $x_0\in(a,b)$, tj. w dowolnym punkcie dziedziny funkcji $f$. 

### Uwagi:

1. Inne oznaczenia pochodnej funkcji $f$ zmiennej $x$ w punkcie $x_0$ to: 

$$
\frac{d y}{d x}|_{x=x_0},\hspace{0.5cm} \frac{d}{d x}f(x_0),\hspace{0.5cm} y^{\hspace{0.05cm}\prime}(x_0).
$$

2. Powyższa definicja pokazuje, że aby obliczyć pochodną z definicji trzeba wiedzieć jak  oblicza się  granice funkcji.

3. Jeżeli funkcja $f$ jest określona na przedziale domkniętym $[a,b]$
to mówimy, że $f$ 

- ma pochodną prawostronną w punkcie $x_0=a$, gdy istnieje (skończona) granica 

$$
\lim_{h\to 0^+}\frac{f(a+h)-f(a)}{h}
$$ 

i wartość tej granicy oznaczamy symbolem $f_+^{\prime}(a)$. 
-  ma pochodną lewostronną w punkcie $x_0=b$, gdy istnieje (skończona) granica 

$$
\lim_{h\to 0^-}\frac{f(b+h)-f(b)}{h}
$$

i wartość tej granicy oznaczamy symbolem $f_-^{\prime}(b)$.
- jest różniczkowalna w $[a,b]$, gdy:

$1^0$ jest różniczkowalna w każdym punkcie przedziału otwartego $(a,b)$, 

$2^0$ ma pochodną prawostronną w punkcie $a$, tzn. $f_+^{\prime}(a):=\lim\limits_{h\to 0^+}\frac{f(a+h)-f(a)}{h}$ istnieje i jest skończona,

$3^0$ posiada pochodną lewostronną w punkcie $b$, tzn. $f_-^{\prime}(b):=\lim\limits_{h\to 0^-}\frac{f(b+h)-f(b)}{h}$ istnieje i jest skończona.

> ### Fakt
>Niech $f\colon (a,b)\to\mathbb{R}$ i niech $x_0\in (a,b)$.
Funkcja $f$ jest różniczkowalna w punkcie $x_0$ wtedy i tylko wtedy, gdy  
>
>$$
f^\prime_+(x_0)=f^\prime_-(x_0).
$$
> 
>Piszemy wówczas $f^\prime(x)=d$, gdzie liczba skończona $d$ oznacza wspólną wartość obu pochodnych jednostronnych.  

### Przykład 1
Niech $f(x)=x^2$ i niech $x\in\mathbb{R}$. Obliczymy pochodną funkcji $f$ w dowolnym punkcie $x\in\mathbb{R}$ korzystając z definicji.

Mamy

$$
\lim_{h\to 0}\frac{f(x+h)-f(x)}{h}=\lim_{h\to 0}\frac{(x+h)^2-(x)^2}{h}=\lim_{h\to 0}\frac{2xh+h^2}{h}
=\lim_{h\to 0}(2x+h)=2x.
$$

Zatem $f$ ma pochodną w dowolnym punkcie $x\in\mathbb{R}$ oraz 

$$
f'(x)=2x.
$$

Zatem dla np. argumentu $x=3$ wartość pochodnej $f'$ w punkcie 3 wynosi $f'(3)=2\cdot 3=6$.

### Przykład 2
Niech $f(x)=\sqrt{x}$. Dziedziną funkcji $f$ jest zbiór $[0,\infty)$.

Niech  $x>0$.
Mamy

$$
\lim_{h\to 0}\frac{f(x+h)-f(x)}{h}=
\lim_{h\to 0}\frac{\sqrt{x+h}-\sqrt{x}}{h}
=\lim_{h\to 0}\frac{\sqrt{x+h}-\sqrt{x}}
{h}\cdot \frac{\sqrt{x+h}+\sqrt{x}}{\sqrt{x+h}+\sqrt{x}}=
\lim_{h\to 0}\frac{h}{h(\sqrt{x+h}+\sqrt{x})}=\lim_{h\to 0}\frac{1}{\sqrt{x+h}+\sqrt{x}}=\frac{1}{2\sqrt{x}}.
$$

Zatem $f$ ma pochodną w dowolnym punkcie $x\in(0,\infty)$ i 

$$
f'(x)=\frac{1}{2\sqrt{x}}.
$$

Zauważmy teraz, że 

$$
\frac{f(0+h)-f(0)}{h}=\frac{\sqrt{h}}{h}=\frac{1}{\sqrt{h}}.
$$

Wyrażenie to nie ma skończonej granicy, gdy $h\to 0^+$. Zatem $f$ nie ma pochodnej w punkcie $x=0$.

### Przykład 3
Niech 

$$
f(x)=|x|.
$$

Pokażemy, że $f$ nie ma pochodnej w punkcie $x=0$.
W tym celu zauważmy, że 

$$
\frac{f(0+h)-f(0)}{h}=
\frac{|h|}{h}.
$$

Wyrażenie to nie ma granicy, gdy $h\to 0$ ponieważ 

$$
\lim_{h\to 0^+}\frac{|h|}{h}=
\lim_{h\to 0^+}\frac{h}{h}=1
\quad
\text{i}
\quad
\lim_{h\to 0^-}\frac{|h|}{h}=
\lim_{h\to 0^-}\frac{-h}{h}=-1.
$$

Zatem funkcja $f$ nie jest różniczkowalna w punkcie $x=0$.

Na koniec przykładu odnotujmy jeszcze, że funkcja $y=|x|$ jest różniczkowalna w $\BR \backslash \{0\}$ (łatwe rachunki).

### Przykład 4
Zbadamy istnienie pochodnej funkcji

$$
f(x)=\begin{cases}
4x+2, & x<1\\
2x^2+4, & x\geq 1
\end{cases}
$$

w punkcie $x_0=1$, a następnie sprawdzimy, czy funkcja jest różniczkowalna (wszędzie na $\mathbb{R}$). 

Z uwagi na fakt, iż funkcja $f$ zdefiniowana jest innym wzorem dla argumentów mniejszych od $1$, a innym dla argumentów większych od $1$ (przy przejściu z lewostronnego otczenia punktu $x_0=1$ do prawostronnego otoczenia punktu $x_0=1$ zmienia się wzór funkcji $f$), zaczynamy od wyliczenia pochodnych jednostronnych tej funkcji w punkcie $x_0=1$. 

Mamy

$$
f^{\prime}_{-}(1)=\lim\limits_{h\tow 0^-}\frac{f(1+h)-f(1)}{h}=\lim\limits_{h\tow 0^-}\frac{4(1+h)+2-6}{h}=\lim\limits_{h\tow 0^-}\frac{4+4h-4}{h}=\lim\limits_{h\tow 0^-}\frac{4h}{h}=\lim\limits_{h\tow 0^-} 4=4
$$

oraz 

$$
f^{\prime}_{+}(1)=\lim\limits_{h\tow 0^+}\frac{f(1+h)-f(1)}{h}=\lim\limits_{h\tow 0^+}\frac{2(1+h)^2+4-6}{h}=\lim\limits_{h\tow 0^+}\frac{4h+2h^2}{h}=\lim\limits_{h\tow 0^+}(4+2h)=4.
$$

Zatem

$$
f^{\prime}_{-}(1)=4=f^{\prime}_{+}(1),
$$

tzn., że istnieje pochodna funkcji $f$ w punkcie $x_0=1$ i wynosi $f^{\prime}(1)=4$.

Sprawdzimy teraz, czy funkcja $f$ jest różniczkowalna poza punktem $1$. Mamy

$$
f^{\prime}(x)=4, \text{ gdy }x<1
$$

oraz 

$$
f^{\prime}(x)=4x, \text{ gdy }x>1,
$$

czyli funkcja $f$ jest różniczkowalna (w każdym punkcie swojej dziedziny), a jej pochodna wyrażona jest wzorem:

$$
f^{\prime}(x)=
\begin{cases}
4, \text{ gdy }x\leq 1,\\
4x, \text{ gdy }x>1.
\end{cases}
$$

Powyższe wzory można otrzymać obliczając pochodne z definicji, bądź stosując wzory na pochodne funkcji elementarnych i twierdzenie o działaniach arytmetycznych na pochodnych (patrz poniżej).

Wniosek dodatkowy: funkcja $f$ jest ciągła w swojej dziedzinie (ponieważ różniczkowalność implikuje ciągłość funkcji - patrz twierdzenie poniżej).

Związek różniczkowalności z ciągłością funkcji $f$ podaje następujące

>__Twierdzenie.__
>Każda funkcja różniczkowalna w $x_0$ jest w tym punkcie ciągła. 

### Uwaga
Twierdzenie odwrotne do powyższego nie jest prawdziwe, co ilustruje poniższy  

### Przykład
Funkcja $f(x)=|x|$ jest ciągła (a nawet jednostajnie ciągła na $\BR$), ale 

$$
f^\prime_+(0)=1\neq-1=f^\prime_-(0)
$$

ponieważ

$$
\lim_{h\to 0^+}\frac{|h|}{h}=
\lim_{h\to 0^+}\frac{h}{h}=1,
$$

a

$$
\lim_{h\to 0^-}\frac{|h|}{h}=
\lim_{h\to 0^-}\frac{-h}{h}=-1.
$$

Kolejny przykład pokazuje, że można wykorzystać brak ciągłości funkcji w punkcie, aby stwierdzić, że dana funkcja nie jest różniczkowalna w tym punkcie.  
### Przykład
Rozważmy funkcję:

$$
f(x)=\begin{cases} 
2x^3-1,&\text{ gdy }x\leq 0 \\ 
4x-3,&\text{ gdy }x>0.\end{cases}
$$

Funkcja ta nie jest różniczkowalna w punkcie $x_0=0$ (tym samym nie jest różniczkowalna na $\BR$), bo nie jest ciągła w tym punkcie.

Gdybyśmy jednak zapomnieli o związku różniczkowalności funkcji jednej zmiennej z ciągłością, to uzyskamy ten sam wniosek stosując definicje pochodnych jednostronnych $f^\prime_-(0)$ oraz $f^\prime_+(0)$, mianowicie

$$
f^\prime_-(0):=\lim\limits_{h\tow 0^-}\frac{f(0+h)-f(0)}{h}=\lim\limits_{h\tow 0^-}\frac{2 h^3-1-(-1)}{h}=\lim\limits_{h\tow 0^-}2(h)^2=0,
$$

oraz

$$
f^\prime_+(0):=\lim\limits_{h\tow 0^+}\frac{f(0+h)-f(0)}{h}=\lim\limits_{h\tow 0^+}\frac{4 h-3-(-1)}{h}=\lim\limits_{h\tow 0^+}\left(4-\frac{2}{h}\right)=-\infty,
$$

czyli $f^\prime_-(0)\neq f^\prime_+(0)$.

## Geometryczna interpretacja pochodnej: prosta styczna

Niech $f\colon (a,b)\to\mathbb{R}$ i niech $x_0\in(a,b)$. Jeżeli  $h\not=0$ jest takie, że 
$x_0+h\in(a,b)$, to punkt $(x_0,f(x_0))$ oraz punkt $(x_0+h,f(x_0+h))$
należą do wykresu funkcji $f$.
Prosta dana wzorem

$$
y=\frac{f(x_0+h)-f(x_0)}{h}(x-x_0)+f(x_0)
$$

przechodzi przez te dwa punkty i nazywana jest sieczną do wykresu funkcji $f$ przechodzącą przez te punkty. 
W aplikacji 

<center><a href="https://www.geogebra.org/m/fct2nuba">Sieczna do wykresu</a></center> 

można obejrzeć wykres siecznej dla różnych parametrów $x_0$ oraz $h$.
Zauważmy, że dla bardzo małych wartości parametru $h$ sieczna 
coraz lepiej przybliża wykres funkcji $f$ w otoczeniu punktu $x_0$.
W "granicy" sieczna staje się prostą, która nazywa się prostą styczną do wykresu funkcji $f$.

Podamy teraz definicję prostej stycznej i prostej normalnej do wykresu funkcji $f$.

>__Definicja.__ Niech $f\colon (a,b)\to\mathbb{R}$ i niech $x_0\in (a,b)$.
>Jeżeli $f$ ma pochodną w punkcie $x_0$, to prostą daną wzorem
>
>$$
y=f'(x_0)(x-x_0)+f(x_0)
$$
>
>nazywamy **prostą styczną** do wykresu funkcji $f$ w punkcie $x_0$.  
>
>Jeżeli dodatkowo $f^\prime(x_0)\neq 0$, to prostą daną wzorem
>
>$$
y=-\frac{1}{f^\prime(x_0)}(x-x_0)+f(x_0).
$$
>
>nazywamy **prostą normalną** do $f$ w punkcie $x_0$.
>
>Prosta normalna to prosta, która jest prostopadła do prostej stycznej do $f$ w punkcie $x_0$.


### Uwagi:
1. Zarówno prosta styczna jak i prosta normalna przechodzą przez wspólny punkt o współrzędnych $(x_0, f(x_0))$. Współczynnik kierunkowy prostej stycznej jest równy $f'(x_0)$, a współczynnik kierunkowy prostej normalnej wynosi $-\frac{1}{f^\prime(x_0)}$.

2. W pewnym sensie prosta styczna "dobrze przybliża" wykres funkcji $f$ w otoczeniu punktu $x_0$.


W aplikacji 

<center><a href="http://www.geogebra.org/m/ktwk6hfy">Styczna do wykresu</a></center> 

można zobaczyć wizualizację stycznej do wykresu funkcji.

### Zadania do samodzielnego rozwiązania

**Zadanie 1**. Znajdź prostą styczną oraz prostą normalną do wykresu funkcji 

$$
f(x)=x^2+3
$$

w punkcie $x_0=1$.

**Zadanie 2**. Wyznacz równania stycznych do wykresu funkcji 

$$
f(x)=\frac{x^3}{3}
$$ 

równoległych do prostej $y=4x+10$.

**Wskazówka.** Proste równoległe mają równe współczynniki kierunkowe.

Odpowiedzi:

1. Prosta styczna: $y=2x+2$; prosta normalna: $y=-\frac{1}{2}x+4\frac{1}{2}$;

2. Proste styczne to: $y=4x-5\frac{1}{3}$ oraz $y=4x+5\frac{1}{3}$.

## Interpretacja fizyczna pochodnej funkcji w punkcie

Niech dany będzie punkt materialny, którego droga poruszania się $s=s(t)$ jest funkcją zależną od czasu $t$. Prędkość w chwili czasu $t_0$ tego punktu materialnego jest pochodną funkcji $s$ w punkcie $t_0$, tzn.

$$
v(t_0)=\lim\limits_{\Delta t\rightarrow 0}\frac{s(t_0+\Delta t)-s(t_0)}{\Delta t} =:s^{\prime}(t_0),
$$

gdzie $\Delta t$ oznacza przyrost czasu.

### Uwaga
Prędkość średnia w przedziale czasowym $[t_0,t_0+\Delta t]$ odpowiada ilorazowi różnicowemu funkcji drogi $s$ o przyroście argumentu $\Delta t$, czyli $\frac{\Delta s}{\Delta t}$.

>### Dla zainteresowanych
>Możliwa jest również inna (pochodząca z termodynamiki) interpretacja fizyczna pojęcia pochodnej.
>
>Przypuśćmy, że $Q(t)$ oznacza ilość ciepła (energii) potrzebną, by podgrzać jednostkę objętości pewnego ciała do temperatury $t$. Pochodna $Q^{\prime}(t)$ jest tzw. ciepłem właściwym ciała w temperaturze $t$. Jest to ilość ciepła potrzebna, aby jednostkę objętości ciała pozostającego w temperaturze $t$ podgrzać o jeden stopień.  

### Zadanie do samodzielnego rozwiązania

Drogę przebytą przez punkt materialny opisuje funkcja $s:[0,\infty]\to[0,\infty]$, 
$s(t)=t^2+2t$. 
Wyznacz prędkość tego punktu w chwili $t_0=1$.

**Wskazówka.** Zastosuj odpowiednie techniki obliczania pochodnych podane poniżej.

Odpowiedź. Prędkość punktu materialnego w chwili $t_0=1$ wynosi $V(1)=s^\prime(1)= 4$.

## Wzory na pochodne wybranych funkcji elementarnych

Poniższe twierdzenie prezentuje wzory na pochodne wybranych funkcji elementarnych. Znajomość tych wzorów jest kluczowa do obliczania pochodnych "bardziej skomplikowanych" funkcji.

>__Twierdzenie.__
>Prawdziwe są następujące wzory na pochodne.
>
>1. Niech $c\in\mathbb{R}$. Funkcja stała $f(x)=c$ ma pochodną w dowolnym punkcie i $f'(x)=0$.
>
>2. Funkcja $f(x)=x$ ma pochodną w dowolnym punkcie i $f'(x)=1$.
>
>3. Niech $\alpha\in\mathbb{R}\setminus\{0,1\}$ i niech $f$ będzie funkcją daną wzorem $f(x)=x^\alpha$. Wówczas:
>- Funkcja $f$ ma pochodną w dowolnym punkcie $x\not=0$ należącym do jej dziedziny i  $f'(x)=\alpha x^{\alpha-1}$.
>
>- Jeżeli $\alpha>1$, to $f$ ma pochodną dla $x=0$  i $f'(0)=0$.
>
>- Jeżeli $\alpha<1$, to albo $0$ nie należy do dziedziny $f$ albo  $f$ nie ma pochodnej w $0$.
>
>4. Funkcja $f(x)=e^x$ ma pochodną w dowolnym punkcie i $f'(x)=e^x$.
>
>5. Funkcja $f(x)=\ln x$ ma pochodną dla każdego $x>0$ i $f'(x)=\frac{1}{x}$.
>
>6. Funkcja $f(x)=\sin x$ ma pochodną w dowolnym punkcie i $f'(x)=\cos x$. 
>
>7. Funkcja $f(x)=\cos x$ ma pochodną w dowolnym punkcie i $f'(x)=-\sin x$.
>
>8. Funkcja $f(x)=\operatorname{tg}x$ ma pochodną w dowolnym punkcie swojej dziedziny i $f'(x)=\frac{1}{\cos^2x}$.
>
>9. Funkcja $f(x)=\operatorname{ctg}x$ ma pochodną w dowolnym punkcie swojej dziedziny i $f'(x)=-\frac{1}{\sin^2x}$.
>
>10. Funkcja $f(x)=\operatorname{arctg}x$ ma pochodną w dowolnym punkcie swojej dziedziny (czyli w $\BR$) i $f'(x)=\frac{1}{1+x^2}$.
>
>11. Funkcja $f(x)=\operatorname{arcsin}x$ ma pochodną w dowolnym punkcie swojej dziedziny i $f'(x)=\frac{1}{\sqrt{1-x^2}}$.

### Uwaga

>Można również spotkać następujące dodatkowe dwa wzory:
>
>12. Funkcja $f(x)=\operatorname{arcctg}x$ ma pochodną w dowolnym punkcie swojej dziedziny (czyli w $\BR$) i $f'(x)=\frac{-1}{1+x^2}$.
>
>13. Funkcja $f(x)=\operatorname{arccos}x$ ma pochodną w dowolnym punkcie swojej dziedziny i $f'(x)=\frac{-1}{\sqrt{1-x^2}}$.

### Przykład

Omówimy stosowanie wzoru $3.$. 

- Niech $f(x)=x^2$. Dla każdego $x\in\mathbb{R}$ mamy $f'(x)=2x$.

- Niech $f(x)=x^5$. Dla każdego $x\in\mathbb{R}$ mamy $f'(x)=5x^4$.

- Niech $f(x)=\frac{1}{x^5}$. Dziedziną funkcji $f$ jest zbiór $\mathbb{R}\setminus \{0\}$. Zauważmy, że $f(x)=x^{-5}$. Dla każdego $x\not=0$ mamy 

$$
f'(x)=-5x^{-6}=-\frac{5}{x^6}.
$$

- Niech $f(x)=\sqrt{x}$. Dziedziną funkcji $f$ jest zbiór $[0,\infty)$. Zauważmy, że $f(x)=x^{\frac{1}{2}}$. Dla każdego $x>0$ mamy

$$
f'(x)=\frac{1}{2}x^{-\frac{1}{2}}=
\frac{1}{2\sqrt{x}}.
$$

Wzór ten warto pamiętać!

- Niech $f(x)=\frac{1}{\sqrt[3]{x}}$. Dziedziną funkcji $f$ jest zbiór $\mathbb{R}\setminus \{0\}$. Zauważmy, że $f(x)=x^{-\frac{1}{3}}$. Dla każdego $x\not=0$ mamy

$$
f'(x)=-\frac{1}{3}x^{-\frac{4}{3}}=
-\frac{1}{3\sqrt[3]{x^4}}.
$$

## Twierdzenie o działaniach arytmetycznych na pochodnych

>__Twierdzenie__
>
>Niech funkcje $f$ i $g$, określone w pewnym otoczeniu punktu $x_0\in\BR$ i o wartościach rzeczywistych, będą różniczkowalne w punkcie $x_0$. Wtedy funkcje $f+g$, $f-g$ i $f\cdot g$ też są różniczkowalne w punkcie $x_0$ oraz 
>
>$$
1^0\hspace{0.5cm} (f+g)^\prime(x_0)=f^\prime(x_0)+g^\prime(x_0),
$$
>
>$$
2^0\hspace{0.5cm}(f-g)^\prime(x_0)=f^\prime(x_0)-g^\prime(x_0),
$$
>
>$$
3^0\hspace{0.5cm} (f\cdot g)^\prime(x_0)=f^\prime(x_0)\cdot g(x_0)+f(x_0)\cdot g^\prime(x_0),
$$ 
>
>w szczególności, 
>
>$$
(cf)^\prime(x_0)=c\cdot f^\prime(x_0)
$$
>
>dla dowolnej liczby rzeczywistej $c$. 
>
>Jeżeli ponadto $g(x_0)\neq 0$, to funkcja $\frac{f}{g}$ jest też różniczkowalna w punkcie $x_0$ oraz
>
>$$
4^0\hspace{0.5cm} \left(\frac{f}{g}\right)^\prime(x_0)=\frac{f^\prime(x_0)\cdot g(x_0)-f(x_0)\cdot g^\prime(x_0)}{(g(x_0))^2}.
$$

### Uwaga
Ze wzoru na pochodną sumy dwóch funkcji bardzo łatwo wynika, że pochodna sumy skończonej ilości funkcji jest równa sumie pochodnych tych funkcji. 

Rozważmy teraz kilka przykładów obliczania pochodnych różnych funkcji.

### Przykład 1
Niech $h(x)=x^6+x^4$. Widzimy, że $h$ jest sumą funkcji $f(x)=x^6$ i funkcji $g(x)=x^4$.
Mamy $f'(x)=6x^5$ i $g'(x)=4x^3$. Zatem ze wzoru na pochodną sumy wynika, że 

$$
h'(x)=6x^5+4x^3.
$$

### Przykład 2
Niech $f(x)=\sin x -2\ln x$. Wtedy 

$$
f'(x)=\cos x -\frac{2}{x}.
$$

### Przykład 3
Niech $f(x)=\operatorname{tg}(x)-\frac{2}{x}+\sqrt{x}$. Wtedy 

$$
f'(x)=\frac{1}{\cos^2x}+\frac{2}{x^2}+\frac{1}{2\sqrt{x}}.
$$

### Przykład 4
Wiemy, że pochodną funkcji $f(x)=x^2$ jest funkcja $f'(x)=2x$.
Zatem pochodną funkcji $g(x)=2x^2$ jest funkcja 

$$
g'(x)=(2f)'(x)=2f'(x)=2\cdot 2x=4x.
$$

### Przykład 5
Wiemy, że pochodną funkcji $f(x)=\ln x$ jest funkcja $f'(x)=\frac{1}{x}$. Zatem pochodną funkcji $g(x)=-\sqrt{3}\ln x$ jest funkcja 

$$
g'(x)=-\frac{\sqrt{3}}{x}.
$$

### Przykład 6
Niech $h(x)=x^2\ln x$. Funkcja $h$ jest iloczynem funkcji $f(x)=x^2$ i funkcji $g(x)=\ln x$.
Mamy $f'(x)=2x$ oraz $g'(x)=\frac{1}{x}$. Stąd ze wzoru na pochodną iloczynu dwóch funkcji otrzymujemy, że 

$$
h'(x)=(f\cdot g)'(x)=f'(x)g(x)+f(x)g'(x)=2x\ln x+x^2\cdot\frac{1}{x}=2x\ln x+x=x(2\ln x+1).
$$

### Przykład 7
Niech $h(x)=\sin x\cdot \arcsin x$. Widzimy, że funkcja $h$ jest iloczynem funkcji $f(x)=\sin x$ i funkcji $g(x)=\arcsin x$. Ponieważ $f'(x)=\cos x$ oraz $g'(x)=\frac{1}{\sqrt{1-x^2}}$, to ze wzoru na pochodna iloczynu dwóch funkcji otrzymujemy, że 

$$
h'(x)=(f\cdot g)'(x)=f'(x)g(x)+f(x)g'(x)=
\cos x\cdot\arcsin x +\frac{\sin x}{\sqrt{1-x^2}}.
$$

### Przykład 8
Niech $h(x)=(3x^2+x)(2+\ln x)$. Funkcja $h$ jest iloczynem 
funkcji $f(x)=3x^2+x$ i funkcji $g(x)=2+\ln x$. Wiemy, że  $f'(x)=6x+1$ oraz $g'(x)=\frac{1}{x}$.
Zatem

$$
h'(x)=(f\cdot g)'(x)=f'(x)g(x)+f(x)g'(x)=(6x+1)(2+\ln x)+(3x^2+x)\cdot\frac{1}{x}=6x\ln x+\ln x+15x+3.
$$

### Przykład 9
Niech $h(x)=x^6\cos x\ln x$. By poradzić sobie z tym przykładem (funkcja $h$ jest iloczynem trzech prostszych funkcji) zauważmy, że funkcja $h$ jest iloczynem funkcji $f(x)=x^6$ i funkcji $g(x)=\cos x\ln x$.

Wiemy, że $f'(x)=6x^5$, a ze wzoru na pochodną iloczynu wynika, że 

$$
g'(x)=-\sin x\ln x+\frac{\cos x}{x}.
$$

Zatem 

$$
h'(x)=(f\cdot g)'(x)=f'(x)g(x)+f(x)g'(x)=6x^5\cos x\ln x+x^6\left(-\sin x\ln x+\frac{\cos x}{x}\right).
$$

### Przykład 10
a) Niech $h(x)=\frac{x^2}{\sin x}$.
Funkcja $h$ jest ilorazem funkcji $f(x)=x^2$ i funkcji $g(x)=\sin x$.
Mamy $f'(x)=2x$ oraz $g'(x)=\cos x$.
Stąd ze wzoru na pochodną ilorazu dwóch funkcji otrzymujemy, że 

$$
h'(x)=\left(\frac{f}{g}\right)'(x)=\frac{f'(x)g(x)-f(x)g'(x)}{g^2(x)}=\frac{2x\sin x-x^2\cos x}{\sin^2x}.
$$

b) Niech $h(x)=\frac{3x^2+x}{2+\ln x}$.
Funkcja $h$ jest ilorazem 
funkcji $f(x)=3x^2+x$ i funkcji $g(x)=2+\ln x$.
Wiemy, że  $f'(x)=6x+1$ oraz $g'(x)=\frac{1}{x}$.
Stąd 

$$
h'(x)=\left(\frac{f}{g}\right)'(x)=
\frac{f'(x)g(x)-f(x)g'(x)}{g^2(x)}
=\frac{(6x+1)(2+\ln x)-(3x^2+x)\cdot\frac{1}{x}}{(2+\ln x)^2}=\frac{9x+1+(6x+1)\ln x}{(2+\ln x)^2}.
$$

c) Niech $h(x)=\frac{x^2}{\sin x}$.
Funkcja $h$ jest ilorazem funkcji $f(x)=x^2$ i funkcji $g(x)=\sin x$.
Mamy $f'(x)=2x$ oraz $g'(x)=\cos x$.
Stąd 

$$
h'(x)=\left(\frac{f}{g}\right)'(x)=\frac{f'(x)g(x)-f(x)g'(x)}{g^2(x)}=\frac{2x\sin x-x^2\cos x}{\sin^2x}.
$$

## Twierdzenie o pochodnej funkcji złożonej

> __Twierdzenie__
>
>Niech funkcja $g$ będzie różniczkowalna w punkcie $x_0$ i niech funkcja $f$ będzie różniczkowalna w punkcie $u_0=g(x_0)$. Wtedy funkcja złożona $f\circ g$ jest różniczkowalna w  punkcie $x_0$ i zachodzi wzór:
>
>$$
(f\circ g)^\prime(x_0)=f^\prime(g(x_0))\cdot g^\prime(x_0).
$$

Zanim pokażemy jak stosować to twierdzenie, rozpoczniemy od przypomnienia operacji superpozycji (złożenia).

__Przypomnienie.__
Niech $X,Y,Z,V$ będą niepustymi podzbiorami zbioru liczb rzeczywistych $\BR$, $f\colon X\rightarrow Y$, $Y\subseteq Z$ i niech $g\colon Z\rightarrow V$. Wówczas funkcja $f\circ g\colon X\rightarrow V$ (czyt. $g$ złożone z $f$) określona jest wzorem

$$
(f\circ g)(x)=f(g(x)) \text{ dla każdego }x\in X.
$$

Funkcję $g$ nazywamy funkcją **wewnętrzną**, a funkcję $f$ nazywamy funkcją **zewnętrzną**. 

Dla przykładu:

- jeżeli $f(x)=e^x$ oraz $g(x)=x^2$, to 

$$
(f\circ g)(x)=f(g(x))=e^{x^2}.
$$

- jeżeli $f(x)=\ln x$ oraz $g(x)=\sin x +1$, to 

$$
(f\circ g)(x)=f(g(x))=\ln(\sin x+1).
$$

- jeżeli $f(x)=x^7$ oraz $g(x)=x^2+4$, to 

$$ 
(f\circ g)(x)=f(g(x))=(x^2+4)^7.
$$

Przejdźmy teraz do przykładów ilustrujących zastosowanie twierdzenia o pochodnej funkcji złożonej.

### Przykłady: 
a) Niech $h(x)=\ln (\sin x+1)$. Mamy $h=f\circ g$, gdzie $f(x)=\ln x$ a $g(x)=\sin x+1$. Ponieważ
$f'(x)=\frac{1}{x}$ a $g'(x)=\cos x$, to ze wzoru na pochodną funkcji złożonej wynika, że

$$
h'(x)=(f\circ g)'(x)=f'(g(x))\cdot g'(x)
=\frac{\cos x}{\sin x+1}.
$$

b) Niech $h(x)=(x^2+4)^7$. Mamy $h=f\circ g$, gdzie $f(x)=x^7$ a $g(x)=x^2+4$. Ponieważ
$f'(x)=7x^6$ a $g'(x)=2x$, to ze wzoru na pochodną funkcji złożonej wynika, że

$$
h'(x)=(f\circ g)'(x)=f'(g(x))\cdot g'(x)
=7(x^2+4)^6\cdot 2x.
$$

c) Niech $h(x)=e^{x^2}$. Mamy
$h=f\circ g$, gdzie $f(x)=e^x$ a $g(x)=x^2$. Ponieważ
$f'(x)=e^x$ a $g'(x)=2x$, to ze wzoru na pochodną funkcji złożonej wynika, że

$$
h'(x)=(f\circ g)'(x)=f'(g(x))\cdot g'(x)
=e^{x^2}\cdot 2x=2x e^{x^2}.
$$

Policzymy teraz pochodne bardziej skomplikowanych funkcji.

## Różniczkowanie $f(x)^{g(x)}$

Dla $a>0$ prawdziwy jest wzór

$$ 
a=e^{\ln a}.
$$

Zatem, w celu obliczenia pochodnej $f(x)^{g(x)}$, możemy zastosować przekształcenie

$$
f(x)^{g(x)}=e^{\ln f(x)^{g(x)}}=e^{g(x)\cdot \ln f(x)},
$$

o ile stwierdzimy, że $f(x)^{g(x)}>0$ dla odpowiednich $x$, a następnie obliczamy pochodną funkcji złożonej $\displaystyle e^{g(x)\cdot \ln f(x)}$, czyli

$$\displaystyle \left(e^{g(x)\cdot \ln f(x)}\right)^\prime.$$


### Przykłady:
a) Funkcja $h(x)=x^x$ jest określona dla wszystkich $x>0$. Obliczymy jej pochodną. Z własności logarytmów mamy, że

$$
h(x)=x^x=e^{\ln x^x}=e^{x\ln x}.
$$

Zatem $h=f\circ g$, gdzie 
$f(x)=e^x$ i $g(x)=x\ln x$.
Wiemy, że 
$f'(x)=e^x$ i $g'(x)=\ln x +1$.
Zatem ze wzoru na pochodną funkcji złożonej wynika, że  

$$
h'(x)=f'(g(x))\cdot g'(x)=e^{x\ln x}\cdot (\ln x +1)=
x^x(\ln x+1).
$$

b) Funkcja $h(x)=(x^2+1)^{\sin x}$ jest określona dla wszystkich $x$ rzeczywistych.  Mamy, że

$$
h(x)=(x^2+1)^{\sin x}=e^{\sin x \ln(x^2+1)}.
$$

Zatem $h=f\circ g$, gdzie 
$f(x)=e^x$ i $g(x)=\sin x\ln(x^2+1)$.
Ponieważ
$f'(x)=e^x$ i 

$$
g'(x)=\cos x\ln (x^2+1)+\sin x\frac{2x}{x^2+1},
$$

to ze wzoru na pochodną funkcji złożonej wynika, że  

$$
h'(x)=e^{\sin x \ln(x^2+1)}\left(\cos x\ln (x^2+1)+\frac{2x\sin x}{x^2+1}\right)=(x^2+1)^{\sin x}\left(\cos x\ln (x^2+1)+\frac{2x\sin x}{x^2+1}\right).
$$

### Zadanie do samodzielnego rozwiązania

b) $\displaystyle h(x)=x^{e^x}$, gdy $x>0$,

b) $h^{\prime}(x)= x^{e^x}\cdot e^x\left(\ln x+\frac{1}{x}\right)$;

c) $h^{\prime}(x)= (\sin x)^{\sin x}(\cos x \ln(\sin x)+\cos x)$.

## Pochodna funkcji odwrotnej

Przypomnijmy, że jeżeli funkcja $f\colon X\to Y$ jest bijekcją (tzn. jest różnowartościowa i suriektywna), to funkcję
$f^{-1}\colon Y\to X$ określoną regułą

$$
f^{-1}(y)=x \text{ wtedy i tylko wtedy, gdy } f(x)=y
$$

nazywamy funkcją odwrotną do funkcji $f$.

>### Twierdzenie
>Niech $f:(a,b)\to (c,d)$ będzie bijekcją i niech $y\in (c,d)$. Jeżeli funkcja $f$ ma pochodną w punkcie $x\in(a,b)$ i $f'(x)\not=0$, to funkcja $f^{-1}$ ma pochodną w punkcie $y=f(x)$ oraz prawdziwy jest wzór
>
>$$
(f^{-1})'(y)=\frac{1}{f'(x)}.
$$



### Przykład 

Funkcja $f\colon (0,2)\to (0,10)$ dana wzorem $f(x)=x+x^3$ jest bijekcją.

Uzasadnimy, że funkcja do niej odwrotna ma pochodną w punkcie $y=2$ i obliczymy jej wartość.

W tym celu zauważmy, że $f(1)=2$.
Ponieważ $f'(x)=1+3x^2$, a w szczególności $f'(1)=4\not=0$, to 
z twierdzenia o pochodnej funkcji odwrotnej wynika, że $f^{-1}$ ma pochodną w punkcie $y=2$ oraz 

$$
(f^{-1})'(2)=\frac{1}{f'(1)}=\frac{1}{4}.
$$

### Uwaga
Zauważmy, że policzyliśmy pochodną funkcji odwrotnej (w pewnym punkcie) pomimo, że nie znamy jej wzoru. 

### Przykład 
Funkcja $f\colon \left(-\frac{\pi}{2},\frac{\pi}{2}\right)\to (-1,1)$ dana wzorem $f(x)=\sin x$ jest bijekcją.

Przypomnijmy, że funkcję do niej odwrotną nazywamy funkcją arcus sinus i oznaczamy symbolem $\arcsin$.
Zatem

$$
\arcsin\colon (-1,1)\to \left(-\frac{\pi}{2},\frac{\pi}{2}\right)
$$

i 

$$
\arcsin (y)=x\quad\text{wtedy i tylko wtedy, gdy}\quad \sin x=y.
$$

Wyprowadzimy teraz wzór na pochodną tej funkcji. 

Zauważmy najpierw, że $f'(x)=\cos x$.
Zatem dla $x\in \left(-\frac{\pi}{2},\frac{\pi}{2}\right)$
mamy, że $f'(x)\not=0$.
Niech $y\in(-1,1)$  i niech $\arcsin(y)=x$ (co oznacza, że $\sin x =y$ ).
Z twierdzenia o funkcji odwrotnej wynika, że 

$$
(\arcsin)'(y)=\frac{1}{\cos x}.
$$

Z kolei z jedynki trygonometrycznej wynika, że (bo dla $x\in \left(-\frac{\pi}{2},\frac{\pi}{2}\right)$ mamy $\cos x>0$) 

$$
\cos x =\sqrt{1-\sin^2x}=\frac{1}{\sqrt{1-y^2}}.
$$

Zatem

$$
(\arcsin)'(y)=\frac{1}{\cos x}=\frac{1}{\sqrt{1-y^2}}.     
$$
